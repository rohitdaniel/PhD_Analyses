```python
logger = logging.getLogger(__name__)

def read_csv_file(path: Path) -> pd.DataFrame:
    try:
        df = pd.read_csv(path, encoding="utf-8")
        logger.info(f"Successfully loaded data from {path}")
        return df
    except FileNotFoundError:
        logger.error(f"File not found: {path}")
        raise
    except pd.errors.ParserError as e:
        logger.error(f"Parsing error in file {path}: {e}")
        raise
    except Exception as e:
        logger.error(f"Unexpected error reading {path}: {e}")
        raise

df_preobs_2024 = read_csv_file(PREOBS_2024_FILE_PATH)
df_preobs_2025 = read_csv_file(PREOBS_2025_FILE_PATH)
df_kari_dataset = read_csv_file(KARI_DATASET_PATH)
df_pilot = read_csv_file(PILOT_FILE_PATH)
```

### Prepare datasets for analyses

```python
# Define latent attributions mapping to item columns (success/failure valence)
latent_map = {
    "eff_1": ("amp_1", "amn_1"),
    "eff_2": ("amp_7", "amn_7"),
    "eff_3": ("amp_15", "amn_15"),
    "ab_1": ("amp_2", "amn_2"),
    "ab_2": ("amp_5", "amn_5"),
    "ab_3": ("amp_9", "amn_9"),
    "luck_1": ("amp_3", "amn_3"),
    "luck_2": ("amp_11", "amn_11"),
    "luck_3": ("amp_14", "amn_14"),
    "task_1": ("amp_4", "amn_4"),
    "task_2": ("amp_8", "amn_8"),
    "task_3": ("amp_12", "amn_12"),
    "int_1": ("amp_6", "amn_6"),
    "int_2": ("amp_10", "amn_10"),
    "int_3": ("amp_13", "amn_13"),
}

# Coalesce success ('amp_*') and failure ('amn_*') columns into unified attribution columns
def coalesce_attributions(df, latent_map):
    df = df.copy()
    for new_col, (amp_col, amn_col) in latent_map.items():
        amp = df[amp_col] if amp_col in df.columns else None
        amn = df[amn_col] if amn_col in df.columns else None
        if amp is not None and amn is not None:
            df[new_col] = amp.combine_first(amn)
        elif amp is not None:
            df[new_col] = amp
        elif amn is not None:
            df[new_col] = amn
        drop = [c for c in [amp_col, amn_col] if c in df.columns]
        df.drop(columns=drop, inplace=True)
    return df

df_preobs_2024  = coalesce_attributions(df_preobs_2024, latent_map)
df_preobs_2025  = coalesce_attributions(df_preobs_2025, latent_map)
df_pilot        = coalesce_attributions(df_pilot, latent_map)
df_kari_dataset = coalesce_attributions(df_kari_dataset, latent_map)

# Add valence category based on scoring in preobs dataset
def assign_valence_preobs(score):
    if score in [3, 4, 5]:
        return "success"
    elif score in [1, 2]:
        return "failure"
    else:
        return None

df_preobs_2024["valence"] = df_preobs_2024["sc"].apply(assign_valence_preobs).astype("category")
df_preobs_2025["valence"] = df_preobs_2025["sc"].apply(assign_valence_preobs).astype("category")
df_kari_dataset["valence"] = df_kari_dataset["sc_1"].apply(assign_valence_preobs).astype("category")

# Add valence category based on scoring in pilot dataset
def assign_valence_pilot(score):
    if score in ["Ganske godt", "Svært godt"]:
        return "success"
    elif score in ["Ganske dårleg", "Svært dårleg"]:
        return "failure"
    else:
        return None

df_pilot["valence"] = df_pilot["sc"].apply(assign_valence_pilot).astype("category")

# Drop rows with missing values in any of the attribution columns
df_preobs_2024.dropna(subset=list(latent_map.keys()), inplace=True)
df_preobs_2025.dropna(subset=list(latent_map.keys()), inplace=True)
df_pilot.dropna(subset=list(latent_map.keys()), inplace=True)

# Filter out patterned responses based on flags in both datasets
df_preobs_2024 = df_preobs_2024[~((df_preobs_2024["attr_patterned"] == 1) & (df_preobs_2024["total_patterned"] > 2))]
df_preobs_2025 = df_preobs_2025[~((df_preobs_2025["attr_patterned"] == 1) & (df_preobs_2025["total_patterned"] > 2))]
df_pilot = df_pilot[~((df_pilot["attr_patterned"] == 1) | (df_pilot["time_flag"] == 1))]

# Reorder columns: keep identifiers and valence first, then latent attribution columns
item_cols = ['eff_1', 'ab_1', 'luck_1', 'task_1', 'int_1', 'eff_2', 'ab_2', 'luck_2', 'task_2', 'int_2', 'eff_3', 'ab_3', 'luck_3', 'task_3', 'int_3']
item_cols_kari = ['eff_1', 'ab_1', 'luck_1', 'task_1', 'int_1', 'eff_2', 'ab_2', 'task_2', 'eff_3', 'ab_3', 'luck_3', 'int_3']

columns_order = ["id_unique", "valence"] + item_cols
columns_order_kari = ["id_unique", "valence"] + item_cols_kari

def safe_select(df, columns):
    existing_cols = [c for c in columns if c in df.columns]
    return df[existing_cols].copy()

preobs_2024_attr_data = safe_select(df_preobs_2024, columns_order)
preobs_2025_attr_data = safe_select(df_preobs_2025, columns_order)
kari_attr_data = safe_select(df_kari_dataset, columns_order_kari)
pilot_attr_data = safe_select(df_pilot, columns_order)
```

```python
# # Convert attribution columns to nullable integer type for both datasets
# def safe_convert_int64(df, cols):
#     existing_cols = [c for c in cols if c in df.columns]
#     df.loc[:, existing_cols] = df.loc[:, existing_cols].astype('Int64')

# safe_convert_int64(preobs_2024_attr_data, item_cols)
# safe_convert_int64(preobs_2025_attr_data, item_cols)
# safe_convert_int64(pilot_attr_data, item_cols)
# safe_convert_int64(kari_attr_data, item_cols_kari)

# Separate success and failure subsets from preobs dataset
def safe_valence_subset(df, valence, cols):
    existing_cols = [c for c in cols if c in df.columns]
    return df.loc[df['valence'] == valence, existing_cols]

# Preobs datasets
preobs_2024_success = safe_valence_subset(preobs_2024_attr_data, 'success', item_cols)
preobs_2024_failure = safe_valence_subset(preobs_2024_attr_data, 'failure', item_cols)
preobs_2025_success = safe_valence_subset(preobs_2025_attr_data, 'success', item_cols)
preobs_2025_failure = safe_valence_subset(preobs_2025_attr_data, 'failure', item_cols)

# Pilot dataset
pilot_success = safe_valence_subset(pilot_attr_data, 'success', item_cols)
pilot_failure = safe_valence_subset(pilot_attr_data, 'failure', item_cols)

# Kari dataset
kari_success = safe_valence_subset(kari_attr_data, 'success', item_cols_kari)
kari_failure = safe_valence_subset(kari_attr_data, 'failure', item_cols_kari)

# Define ordered categorical dtype with categories 1 through 6 (1 through 11 in Kari's data)
ordered_cat_type = pd.CategoricalDtype(categories=list(range(1, 7)), ordered=True)
ordered_cat_type_kari = pd.CategoricalDtype(categories=list(range(1, 12)), ordered=True)

# Convert attribution columns in main datasets to ordered categorical
def safe_convert_ordered(df, cols, cat_type):
    existing_cols = [c for c in cols if c in df.columns]
    df.loc[:, existing_cols] = df.loc[:, existing_cols].astype(cat_type)

safe_convert_ordered(preobs_2024_attr_data, item_cols, ordered_cat_type)
safe_convert_ordered(preobs_2025_attr_data, item_cols, ordered_cat_type)
safe_convert_ordered(pilot_attr_data, item_cols, ordered_cat_type)
safe_convert_ordered(kari_attr_data, item_cols, ordered_cat_type_kari)

# Convert separated success/failure subsets to ordered categorical dtype
preobs_2024_success = preobs_2024_success.astype(ordered_cat_type)
preobs_2024_failure = preobs_2024_failure.astype(ordered_cat_type)
preobs_2025_success = preobs_2025_success.astype(ordered_cat_type)
preobs_2025_failure = preobs_2025_failure.astype(ordered_cat_type)
pilot_success = pilot_success.astype(ordered_cat_type)
pilot_failure = pilot_failure.astype(ordered_cat_type)
kari_success = kari_success.astype(ordered_cat_type_kari)
kari_failure = kari_failure.astype(ordered_cat_type_kari)
```

```python
# Convert preobs attribution data to JAX arrays (zero-indexed)
preobs_2024_data_array = jnp.asarray(preobs_2024_attr_data[item_cols].to_numpy(), dtype=jnp.int32) - 1
preobs_2025_data_array = jnp.asarray(preobs_2025_attr_data[item_cols].to_numpy(), dtype=jnp.int32) - 1

# Convert preobs success/failure subsets to JAX arrays (zero-indexed)
preobs_2024_success_array = jnp.asarray(preobs_2024_success.to_numpy(), dtype=jnp.int32) - 1
preobs_2024_failure_array = jnp.asarray(preobs_2024_failure.to_numpy(), dtype=jnp.int32) - 1
preobs_2025_success_array = jnp.asarray(preobs_2025_success.to_numpy(), dtype=jnp.int32) - 1
preobs_2025_failure_array = jnp.asarray(preobs_2025_failure.to_numpy(), dtype=jnp.int32) - 1

# Convert Kari's attribution data to JAX arrays (zero-indexed)
kari_data_array = jnp.asarray(kari_attr_data[item_cols_kari].to_numpy(), dtype=jnp.int32) - 1

# Convert Kari's success/failure subsets to JAX arrays (zero-indexed)
kari_success_array = jnp.asarray(kari_success.to_numpy(), dtype=jnp.int32) - 1
kari_failure_array = jnp.asarray(kari_failure.to_numpy(), dtype=jnp.int32) - 1

# Convert pilot attribution data to JAX arrays (zero-indexed)
pilot_data_array = jnp.asarray(pilot_attr_data[item_cols].to_numpy(), dtype=jnp.int32) - 1

# Convert pilot success/failure subsets to JAX arrays (zero-indexed)
pilot_success_array = jnp.asarray(pilot_success.to_numpy(), dtype=jnp.int32) - 1
pilot_failure_array = jnp.asarray(pilot_failure.to_numpy(), dtype=jnp.int32) - 1
```

```python
# Initialize LabelEncoder to convert text labels into numeric codes
le = LabelEncoder()

le.fit(["failure", "success"])

# Encode 'valence' in preobs dataset and store as int32
preobs_2024_valence = le.transform(preobs_2024_attr_data['valence']).astype(np.int32)
preobs_2024_attr_data.loc[:, 'valence_encoded'] = preobs_2024_valence
preobs_2025_valence = le.transform(preobs_2025_attr_data['valence']).astype(np.int32)
preobs_2025_attr_data.loc[:, 'valence_encoded'] = preobs_2025_valence

# Encode 'valence' in pilot dataset using the same encoder to keep labels consistent
pilot_valence = le.transform(pilot_attr_data['valence']).astype(np.int32)
pilot_attr_data.loc[:, 'valence_encoded'] = pilot_valence

# Encode 'valence' in Kari's dataset using the same encoder to keep labels consistent
kari_valence = le.transform(kari_attr_data['valence']).astype(np.int32)
kari_attr_data.loc[:, 'valence_encoded'] = kari_valence

def create_group_indices(group_labels: np.ndarray) -> list[np.ndarray]:
    """Create list of index arrays, one per unique group label."""
    unique_groups = np.unique(group_labels)
    return [np.where(group_labels == g)[0] for g in unique_groups]

# Create index lists for preobs and pilot valence groups
preobs_2024_valence_indices = create_group_indices(preobs_2024_valence)
preobs_2025_valence_indices = create_group_indices(preobs_2025_valence)
pilot_valence_indices = create_group_indices(pilot_valence)
kari_valence_indices = create_group_indices(kari_valence)

# Print mapping of encoded groups to original labels
print("Group labels:", le.classes_)
```

### Descriptive statistics: Preobs 2024

```python
# Categories (e.g., 1 to 6)
categories = range(1, 7)

# Set seaborn style
sns.set(style="whitegrid")

# 1. Frequency distribution barplots per item
for col in item_cols:
    counts = preobs_2024_attr_data[col].value_counts().reindex(categories, fill_value=0)
    plt.figure(figsize=(8, 4))
    sns.barplot(x=list(categories), y=counts.values, palette="muted")
    plt.title(f"Frequency Distribution: {col}")
    plt.xlabel("Category")
    plt.ylabel("Count")
    plt.xticks(ticks=list(categories), labels=list(categories))
    plt.show()

# 2. Stacked bar plot of category proportions across all items
prop_df = pd.DataFrame({
    col: preobs_2024_attr_data[col].value_counts(normalize=True).reindex(categories, fill_value=0)
    for col in item_cols
}).T

prop_df.plot(kind='bar', stacked=True, colormap='Spectral', figsize=(12,6))
plt.title("Stacked Bar Plot of Category Proportions by Item")
plt.xlabel("Items")
plt.ylabel("Proportion")
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 3. Cumulative distribution function (CDF) plots per item
for col in item_cols:
    data = preobs_2024_attr_data[col].dropna().astype(int)
    cdf = data.value_counts(normalize=True).sort_index().cumsum()
    plt.figure(figsize=(8, 4))
    sns.lineplot(x=cdf.index, y=cdf.values, marker="o")
    plt.title(f"Cumulative Distribution (CDF) of {col}")
    plt.xlabel("Category")
    plt.ylabel("Cumulative Proportion")
    plt.xticks(ticks=list(categories), labels=list(categories))
    plt.ylim(0, 1)
    plt.grid(True)
    plt.show()

# 4. Median and mode per item (descriptive summary)
summary_stats = []
for col in item_cols:
    data = preobs_2024_attr_data[col].dropna().astype(int)
    median_val = data.median()
    mode_val = data.mode().iloc[0] if not data.mode().empty else np.nan
    summary_stats.append({
        "Item": col,
        "Median": median_val,
        "Mode": mode_val,
        "Count": len(data)
    })

summary_df = pd.DataFrame(summary_stats)
print(summary_df)

# 5. Nonparametric group comparisons (example: valence groups)
import scipy.stats as stats

# Assuming 'valence' is categorical with two groups: 'success' and 'failure'
group1 = preobs_2024_attr_data[preobs_2024_attr_data['valence'] == 'success']
group2 = preobs_2024_attr_data[preobs_2024_attr_data['valence'] == 'failure']

group_comp_results = []
for col in item_cols:
    data1 = group1[col].dropna().astype(int)
    data2 = group2[col].dropna().astype(int)
    # Mann-Whitney U test (nonparametric test for independent samples)
    if len(data1) > 0 and len(data2) > 0:
        stat, p_val = stats.mannwhitneyu(data1, data2, alternative='two-sided')
    else:
        stat, p_val = np.nan, np.nan
    group_comp_results.append({
        "Item": col,
        "Mann-Whitney U stat": stat,
        "p-value": p_val
    })

group_comp_df = pd.DataFrame(group_comp_results)
print(group_comp_df)

# Optional: adjust p-values for multiple testing (e.g., Benjamini-Hochberg FDR)
from statsmodels.stats.multitest import multipletests
pvals = group_comp_df["p-value"].values
_, pvals_adj, _, _ = multipletests(pvals, method='fdr_bh')
group_comp_df["p-value_adj"] = pvals_adj

print(group_comp_df)
```

Exploratory Cross-Group EFA Structure Comparison: Success vs. Failure
Overview
Prior to multigroup confirmatory factor analysis (MG-CFA), an exploratory comparison of the posterior factor structures was conducted between the success (n=213) and failure (n=45) groups using Bayesian Tucker's congruence coefficients (φ). The purpose of this analysis was to assess whether the five-factor attribution structure identified in the combined exploratory factor analysis (EFA) manifests similarly across valence conditions, and to inform the appropriate level of measurement invariance to pursue in subsequent CFA modelling.
Factor correspondence across groups was evaluated using per-draw Hungarian matching over 5,000 independent cross-pairs sampled from the product posterior p(Λ_success) × p(Λ_failure), following the approach described by Conti et al. (2014). Matrix-level similarity was assessed using the RV coefficient (Robert & Escoufier, 1976). Congruence thresholds follow Lorenzo-Seva and ten Berge (2006), who define φ ≥ 0.95 as indicating good factor replication and φ ≥ 0.85 as fair replication.

Factor Correspondence
The joint permutation posterior assigned 96.5% of draw pairs to the identity permutation (Effort→Effort, Ability→Ability, Luck→Luck, Task→Task, Interest→Interest), confirming unambiguous factor correspondence across groups. The small residual probability mass on alternative permutations (Ability↔Task at 1.2%, all others < 1%) is consistent with sampling noise rather than genuine factor ordering ambiguity. Factor labelling across groups is therefore not in question.

No factor achieved good congruence (φ ≥ 0.95). Interest was the most structurally stable factor across groups (φ = 0.84, P(φ>0.85) = 0.42), while Effort and Luck showed the greatest divergence (φ = 0.61 and 0.64 respectively), with effectively zero posterior probability of fair or good replication. The RV coefficient, measuring matrix-level similarity, was 0.670 (90% HDI [0.579, 0.762]), with the entire interval well below 0.90, confirming that the loading matrices differ substantially as a whole.

Theoretical Interpretation
The pattern of congruence is theoretically interpretable within Weiner's (1985) attribution framework. Study items were designed with mirrored wording across valence conditions — for example, "I succeeded because I worked hard" (success) and "I failed because I did not work hard" (failure) — so that item content was held constant while allowing attribution patterns to vary. Under a purely symmetric inversion model, factor structures would be expected to replicate closely. The observed divergence therefore reflects genuine psychological asymmetry in attribution rather than item content differences.
Effort and Luck, the most divergent factors, correspond to the internal-controllable and external-uncontrollable dimensions of Weiner's model respectively. Their low congruence is consistent with self-serving attribution bias (Mezulis et al., 2004), whereby students invoke effort as a prominent explanation for success but show a structurally different pattern when explaining failure, with luck attributions similarly asymmetric across valence conditions. Interest, the most stable factor, is consistent with its status as a relatively stable trait attribution that is less susceptible to self-serving motivational distortion. The borderline congruence of Ability and Task difficulty suggests partial but incomplete structural similarity, potentially reflecting the complexity of students' competence-related attributions across success and failure contexts.

Sample Size Caveat
These findings must be interpreted with considerable caution. The failure group comprised only 45 participants, well below the minimum of 100–200 typically recommended for stable factor solutions (MacCallum et al., 1999), and yielding approximately three observations per free parameter in a five-factor model. At this sample size, the posterior distribution over loadings in the failure group is substantially wider than in the success group, meaning that apparent structural differences may partly reflect estimation uncertainty rather than genuine population-level divergence. The posterior factor correlations in the failure model showed a notably uniform pattern (all pairs 0.38–0.69) with high posterior standard deviations (0.19–0.26), further indicating that the failure model is not fully identified by the available data. Consequently, the phi values reported here should be treated as preliminary and exploratory. Replication with a larger failure sample is necessary before drawing firm conclusions about structural non-equivalence.
Additionally, the mirrored item wording may introduce a framing artefact in the failure condition. Items phrased as negations of effort or ability (e.g., "I failed because I did not work hard") carry implicit self-critical content that may activate social desirability responding independently of genuine attribution patterns, potentially contributing to Effort factor divergence beyond the self-serving bias mechanism.

Next Steps
The results of this exploratory comparison inform the following analytical decisions:
1. Combined-sample EFA as the primary measurement model. Given the insufficient failure sample size for reliable group-specific factor estimation, combined EFA across both valence conditions is used to establish the measurement model for subsequent CFA. This approach maximises the stability of the estimated factor structure while preserving the full sample.
2. Configural MG-CFA as the primary invariance model. The phi results do not support proceeding directly to metric or scalar invariance testing. A configural model — specifying the same factor pattern across groups with all parameters freely estimated — is the appropriate starting point (Vandenberg & Lance, 2000). Adequate configural fit in each group separately is a prerequisite for any subsequent invariance testing (Byrne et al., 1989).
3. Partial invariance testing if configural fit is adequate. Should configural CFA achieve acceptable fit in both groups, partial metric invariance testing (Byrne et al., 1989) may be pursued to identify specific loadings that are invariant across conditions. Given the theoretical expectation that Interest attributions are most stable, these items are the most promising candidates for invariant loadings.
4. Pooling of additional data. To address the sample size limitation, data from additional collection waves should be pooled to increase the failure group sample prior to formal invariance testing. The current failure n=45 is insufficient to support the asymptotic assumptions underlying multigroup CFA fit indices and standard errors, even in a Bayesian framework where regularisation partially mitigates small-sample instability.
5. Bayesian multigroup ordinal CFA. All subsequent confirmatory modelling will use Bayesian ordinal MG-CFA implemented in NumPyro, consistent with the EFA stage. This approach handles the imbalanced group sizes more gracefully than frequentist MG-CFA by regularising failure group parameters through priors, and provides full posterior uncertainty quantification rather than potentially unreliable asymptotic standard errors.

## Helper functions

```python
import numpy as np
import pandas as pd
import arviz as az

def extract_latent_means(
    idata,
    factor_names,
    group_names,
    hdi_prob=0.95,
):
    da = idata.posterior["latent_means"]

    # Convert to numpy: (chain, draw, group, factor)
    values = da.values

    C, D, G, F = values.shape

    if F != len(factor_names):
        raise ValueError(
            f"Model has {F} latent factors, but {len(factor_names)} names were provided."
        )

    if G != len(group_names):
        raise ValueError(
            f"Model has {G} groups, but {len(group_names)} names were provided."
        )

    rows = []

    for g in range(G):
        for f in range(F):
            samples = values[:, :, g, f].reshape(-1)

            hdi = az.hdi(samples, hdi_prob)

            rows.append({
                "Group": group_names[g],
                "Factor": factor_names[f],
                "Mean": float(samples.mean()),
                "Median": float(np.median(samples)),
                f"{int(hdi_prob*100)}% HDI Lower": float(hdi[0]),
                f"{int(hdi_prob*100)}% HDI Upper": float(hdi[1]),
            })

    return pd.DataFrame(rows)
```

```python
def plot_ppc_ordinal_item(obs, ppc_samples, group_vec, model, item_idx, 
                          group_names=("Success", "Failure"), num_pp_samples=100):
    n_cats = model.n_cats
    fig, axes = plt.subplots(1, len(group_names), figsize=(14, 5), sharey=True)
    
    for g, (ax, gname) in enumerate(zip(axes, group_names)):
        g_mask = np.asarray(group_vec) == g
        obs_g = obs[g_mask]
        ppc_g = ppc_samples[:num_pp_samples, g_mask, :]
        n_obs_g = obs_g.shape[0]
        
        obs_counts = np.bincount(obs_g[:, item_idx], minlength=n_cats) / n_obs_g
        ppc_cat_props = np.array([
            np.bincount(s[:, item_idx], minlength=n_cats) / n_obs_g
            for s in ppc_g
        ])
        
        ppc_mean  = ppc_cat_props.mean(axis=0)
        ppc_lower = np.percentile(ppc_cat_props, 5,  axis=0)
        ppc_upper = np.percentile(ppc_cat_props, 95, axis=0)
        categories = np.arange(n_cats)
        
        for sp in ppc_cat_props:
            ax.plot(categories, sp, color="gray", alpha=0.15, lw=1)
        ax.plot(categories, ppc_mean, color="blue", lw=3, label="PP mean")
        ax.fill_between(categories, ppc_lower, ppc_upper, color="blue", 
                        alpha=0.2, label="90% CI")
        ax.stem(categories, obs_counts, linefmt="r-", markerfmt="ro", 
                basefmt="k-", label="Observed")
        ax.set_title(f"Item {item_idx} — {gname} (n={n_obs_g})")
        ax.set_xlabel("Category")
        ax.legend()
        ax.grid(True)
    
    axes[0].set_ylabel("Proportion")
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import arviz as az
import pandas as pd


def inspect_loadings(
    idata,
    n_items,
    n_factors,
    n_groups,
    item_names,
    invariance_type="configural",
    partial_invariance=None,
    standardize=True,
    return_intervals=True,
    hdi_prob=0.95,
    compute_reliability=True,
    compact_ci=True,
    summary_stat="mean",
):
    """
    Inspect Bayesian ordinal CFA loadings with posterior standardization and
    McDonald's omega reliability.

    Parameters
    ----------
    idata : InferenceData
        ArviZ InferenceData from NumPyro MCMC sampling. Must contain
        posterior variables 'loadings' and 'factor_cov_group'.
    n_items : int
    n_factors : int
    n_groups : int
    item_names : list[str]
        Length must equal n_items.
    invariance_type : str
        "configural", "metric", or "scalar". Controls whether loadings tables
        are shown per group or shared.
    partial_invariance : dict or None
        Keys: 'loadings_free_factors' (list[int]). If provided and non-empty
        under metric/scalar invariance, separate tables are shown per group.
    standardize : bool
        If True, loadings are standardized using posterior factor and residual
        variances.
    return_intervals : bool
        If True, HDI intervals are computed and included in output tables.
    hdi_prob : float
        HDI probability (default 0.95).
    compute_reliability : bool
        If True, McDonald's omega is computed per factor per group.
    compact_ci : bool
        If True, mean and CI are combined into a single string column.
    summary_stat : str
        "mean" or "median" for the point estimate of loadings.

    Returns
    -------
    loadings_tables : dict[str, pd.DataFrame]
    reliability_tables : dict[str, pd.DataFrame] or None
    """

    if partial_invariance is None:
        partial_invariance = {"loadings_free_factors": []}

    partial_invariance.setdefault("loadings_free_factors", [])

    # ----------------------------------------------------------------
    # 1. Extract loadings: (samples, groups, items, factors)
    # ----------------------------------------------------------------
    lam_stacked = idata.posterior["loadings"].stack(sample=("chain", "draw"))
    # dims after stack: some permutation of (group, item, factor, sample)
    # Use .values after sorting axes by known sizes rather than name inference
    # to avoid brittle dim-name matching on square covariance dimensions.
    lam_np = lam_stacked.values  # shape depends on xarray dim order

    # Identify axis positions by size
    shape = lam_np.shape
    # sample axis is always last after stack
    # remaining axes: find which is group/item/factor by size
    non_sample_shape = shape[:-1]  # excludes sample (last)

    def _find_axis(target_size, exclude=()):
        for ax, s in enumerate(non_sample_shape):
            if s == target_size and ax not in exclude:
                return ax
        raise ValueError(
            f"Could not find axis with size {target_size} in shape {non_sample_shape}. "
            f"Check n_groups={n_groups}, n_items={n_items}, n_factors={n_factors}."
        )

    ax_g = _find_axis(n_groups)
    ax_i = _find_axis(n_items, exclude=(ax_g,))
    ax_f = _find_axis(n_factors, exclude=(ax_g, ax_i))
    ax_s = len(non_sample_shape)  # last axis

    # Transpose to (samples, groups, items, factors)
    lam_np = np.transpose(lam_np, (ax_s, ax_g, ax_i, ax_f))
    n_samples, G, I, F = lam_np.shape

    # ----------------------------------------------------------------
    # 2. Extract factor covariance: (samples, groups, factors, factors)
    # ----------------------------------------------------------------
    cov_stacked = idata.posterior["factor_cov_group"].stack(sample=("chain", "draw"))
    cov_np = cov_stacked.values  # shape: (..., sample) with two factor axes of size F

    cov_shape = cov_np.shape
    non_sample_cov = cov_shape[:-1]

    ax_cg = _find_axis_cov(n_groups, non_sample_cov)
    # The two factor axes both have size F; find them in order
    f_axes = [ax for ax, s in enumerate(non_sample_cov) if s == n_factors]
    if len(f_axes) < 2:
        raise ValueError(
            f"Expected two factor axes of size {n_factors} in factor_cov_group, "
            f"got shape {non_sample_cov}."
        )
    ax_cf1, ax_cf2 = f_axes[0], f_axes[1]
    ax_cs = len(non_sample_cov)

    # Transpose to (samples, groups, factors, factors)
    cov_np = np.transpose(cov_np, (ax_cs, ax_cg, ax_cf1, ax_cf2))
    # shape: (n_samples, G, F, F)

    # Factor variances: diagonal of covariance
    factor_vars = np.diagonal(cov_np, axis1=2, axis2=3)  # (n_samples, G, F)

    # ----------------------------------------------------------------
    # 3. Residual variances: 1 - communality per item
    #    communality_i = sum_f ( lambda_if^2 * phi_ff )
    #    This is the diagonal of Lambda Phi Lambda^T
    # ----------------------------------------------------------------
    residual_vars = np.zeros((n_samples, G, I))
    for s in range(n_samples):
        for g in range(G):
            # communality: sum over factors of lambda^2 * factor_variance
            communality = np.sum(lam_np[s, g] ** 2 * factor_vars[s, g], axis=1)  # (I,)
            residual_vars[s, g] = np.clip(1.0 - communality, 1e-6, None)

    # ----------------------------------------------------------------
    # 4. Standardize loadings (FIX: correct formula)
    #    Standardized loading: lambda*_if = lambda_if * sqrt(phi_ff) / sqrt(Var(y_i))
    #    Var(y_i) = sum_f lambda_if^2 * phi_ff + psi_i
    #             = communality_i + residual_i
    # ----------------------------------------------------------------
    if standardize:
        lam_std = np.zeros_like(lam_np)
        for s in range(n_samples):
            for g in range(G):
                for i in range(I):
                    item_var = (
                        np.sum(lam_np[s, g, i] ** 2 * factor_vars[s, g])
                        + residual_vars[s, g, i]
                    )
                    sqrt_item_var = np.sqrt(np.clip(item_var, 1e-12, None))
                    for f in range(F):
                        sqrt_phi = np.sqrt(np.clip(factor_vars[s, g, f], 1e-12, None))
                        lam_std[s, g, i, f] = (
                            lam_np[s, g, i, f] * sqrt_phi / sqrt_item_var
                        )
    else:
        lam_std = lam_np.copy()

    # ----------------------------------------------------------------
    # 5. Posterior summary
    # ----------------------------------------------------------------
    if summary_stat == "mean":
        lam_center = lam_std.mean(axis=0)        # (G, I, F)
    elif summary_stat == "median":
        lam_center = np.median(lam_std, axis=0)
    else:
        raise ValueError("summary_stat must be 'mean' or 'median'.")

    # ----------------------------------------------------------------
    # 6. HDI intervals
    # ----------------------------------------------------------------
    hdi_lo = np.zeros((G, I, F))
    hdi_hi = np.zeros((G, I, F))
    if return_intervals:
        for g in range(G):
            for i in range(I):
                for f in range(F):
                    h = az.hdi(lam_std[:, g, i, f], hdi_prob)
                    hdi_lo[g, i, f] = float(h[0])
                    hdi_hi[g, i, f] = float(h[1])

    def fmt(mu, lo, hi):
        return f"{mu:.3f} [{lo:.3f}, {hi:.3f}]"

    # ----------------------------------------------------------------
    # 7. Build loadings tables
    # ----------------------------------------------------------------
    shared = invariance_type in ("metric", "scalar")
    free_factors = partial_invariance.get("loadings_free_factors", [])
    loadings_tables = {}

    def _make_table(g_idx, label):
        data = {}
        for f in range(F):
            col = f"Factor {f + 1}"
            if return_intervals and compact_ci:
                col += f" ({int(hdi_prob * 100)}% CI)"
                data[col] = [
                    fmt(lam_center[g_idx, i, f], hdi_lo[g_idx, i, f], hdi_hi[g_idx, i, f])
                    for i in range(I)
                ]
            else:
                data[col] = lam_center[g_idx, :, f].tolist()
        return pd.DataFrame(data, index=item_names)

    if shared:
        if free_factors:
            # Partial invariance: show per-group tables
            for g in range(G):
                loadings_tables[f"Group {g} (partial invariance)"] = _make_table(g, g)
        else:
            # Fully shared: show group 0 only (identical across groups by invariance)
            loadings_tables["Shared loadings"] = _make_table(0, 0)
    else:
        # Configural: separate table per group
        for g in range(G):
            loadings_tables[f"Group {g}"] = _make_table(g, g)

    # ----------------------------------------------------------------
    # 8. McDonald's omega reliability (FIX: correct formula)
    #
    #    omega_f = (sum_i lambda_if)^2 * phi_ff
    #              -----------------------------------
    #              (sum_i lambda_if)^2 * phi_ff + sum_i psi_i
    #
    #    where the sum over items is restricted to items loading on factor f
    #    (non-zero loadings only, i.e. primary items).
    # ----------------------------------------------------------------
    reliability_tables = None
    if compute_reliability:

        def omega_per_factor(lam_g, factor_var_g, resid_var_g):
            """
            lam_g:         (I, F)
            factor_var_g:  (F,)
            resid_var_g:   (I,)
            Returns list of F omega values.
            """
            omegas = []
            for f in range(F):
                lam_f = lam_g[:, f]
                # Restrict to items with non-negligible primary loading
                primary = np.abs(lam_f) > 1e-6
                if primary.sum() == 0:
                    omegas.append(np.nan)
                    continue
                lam_f_primary = lam_f[primary]
                psi_primary   = resid_var_g[primary]
                phi_f         = factor_var_g[f]

                numerator   = (np.sum(lam_f_primary) ** 2) * phi_f
                denominator = numerator + np.sum(psi_primary)
                omegas.append(numerator / denominator if denominator > 1e-12 else np.nan)
            return omegas

        def _reliability_table(g_idx):
            omegas = np.zeros((n_samples, F))
            for s in range(n_samples):
                omegas[s] = omega_per_factor(
                    lam_np[s, g_idx],       # unstandardized for omega
                    factor_vars[s, g_idx],
                    residual_vars[s, g_idx],
                )
            omega_mean   = np.nanmean(omegas, axis=0)
            omega_median = np.nanmedian(omegas, axis=0)
            omega_hdi    = np.array([az.hdi(omegas[:, f], hdi_prob) for f in range(F)])
            return pd.DataFrame(
                {
                    "Omega Mean":                     omega_mean,
                    "Omega Median":                   omega_median,
                    f"{int(hdi_prob * 100)}% HDI Lower": omega_hdi[:, 0],
                    f"{int(hdi_prob * 100)}% HDI Upper": omega_hdi[:, 1],
                },
                index=[f"Factor {f + 1}" for f in range(F)],
            )

        reliability_tables = {}
        if shared and free_factors:
            for g in range(G):
                reliability_tables[f"Group {g} (partial invariance)"] = _reliability_table(g)
        elif shared:
            reliability_tables["Shared reliability"] = _reliability_table(0)
        else:
            for g in range(G):
                reliability_tables[f"Group {g}"] = _reliability_table(g)

    return loadings_tables, reliability_tables


# ----------------------------------------------------------------
# Helper used internally for covariance axis inference
# (separate from _find_axis to avoid closure over non_sample_shape)
# ----------------------------------------------------------------
def _find_axis_cov(target_size, shape_tuple):
    for ax, s in enumerate(shape_tuple):
        if s == target_size:
            return ax
    raise ValueError(
        f"Could not find axis with size {target_size} in shape {shape_tuple}."
    )
```

```python
import numpy as np
import arviz as az
import pandas as pd


def factor_correlation_table(
    idata,
    summary="mean",
    include_ci=True,
    ci_prob=0.95,
    factor_names=None,
    by_group=True,
    group_names=None,
):
    """
    Create tables of factor correlations from Bayesian (multi-group) CFA posteriors.

    Supports configural, metric, scalar, and partial invariance models.
    With corr_invariance=True (the default in BayesianOrdinalMGCFA), the shared
    correlation matrix is broadcast to all groups — returned tables will be
    identical across groups, which is correct and expected.

    Parameters
    ----------
    idata : az.InferenceData
        Must contain posterior variable 'factor_corr_group' with dims
        (chain, draw, group, factor, factor).
    summary : str
        "mean" or "median" point estimate.
    include_ci : bool
        If True, HDI intervals are included.
    ci_prob : float
        HDI probability (default 0.95).
    factor_names : list[str] or None
        Names for each factor. Defaults to ["Factor 1", "Factor 2", ...].
    by_group : bool
        If True, returns dict of per-group DataFrames.
        If False, returns a single pooled DataFrame with a 'group' column.
    group_names : list[str] or None
        Names for each group. Defaults to ["Group 0", "Group 1", ...].

    Returns
    -------
    dict[str, pd.DataFrame]   if by_group=True
    pd.DataFrame              if by_group=False
    """

    # ----------------------------------------------------------------
    # 1. Stack chain + draw → sample; shape becomes (..., sample) last
    # ----------------------------------------------------------------
    da_stacked = idata.posterior["factor_corr_group"].stack(sample=("chain", "draw"))

    # Move sample axis to last position explicitly — do not assume xarray order
    sample_ax = list(da_stacked.dims).index("sample")
    values = np.moveaxis(da_stacked.values, sample_ax, -1)
    # values shape: (group, F, F, S) or (F, F, S) for single-group models

    # ----------------------------------------------------------------
    # 2. Ensure group axis is present
    # ----------------------------------------------------------------
    if values.ndim == 3:
        # Single-group model — add dummy group axis
        values = values[None, ...]   # (1, F, F, S)

    if values.ndim != 4:
        raise ValueError(
            f"Unexpected shape after stacking: {values.shape}. "
            "Expected (group, F, F, S)."
        )

    G, F, _, S = values.shape

    # ----------------------------------------------------------------
    # 3. Naming
    # ----------------------------------------------------------------
    if factor_names is None:
        factor_names = [f"Factor {i + 1}" for i in range(F)]
    if len(factor_names) != F:
        raise ValueError(
            f"factor_names has {len(factor_names)} entries but model has {F} factors."
        )

    if group_names is None:
        group_names = [f"Group {g}" for g in range(G)]
    if len(group_names) != G:
        raise ValueError(
            f"group_names has {len(group_names)} entries but model has {G} groups."
        )

    # ----------------------------------------------------------------
    # 4. Build per-group tables
    # ----------------------------------------------------------------
    tables = {}
    for g in range(G):
        rows = []
        for i in range(F):
            for j in range(i + 1, F):
                samples = values[g, i, j, :]   # (S,)

                if summary == "mean":
                    point = float(samples.mean())
                elif summary == "median":
                    point = float(np.median(samples))
                else:
                    raise ValueError("summary must be 'mean' or 'median'.")

                row = {
                    "factor_1": factor_names[i],
                    "factor_2": factor_names[j],
                    "r":        point,
                }

                if include_ci:
                    # FIX: use HDI instead of symmetric quantiles —
                    # correlations are bounded and often skewed
                    h = az.hdi(samples, ci_prob)
                    row[f"r_lower_{int(ci_prob * 100)}hdi"] = float(h[0])
                    row[f"r_upper_{int(ci_prob * 100)}hdi"] = float(h[1])

                rows.append(row)

        tables[group_names[g]] = pd.DataFrame(rows)

    # ----------------------------------------------------------------
    # 5. Return
    # ----------------------------------------------------------------
    if by_group:
        return tables

    pooled = pd.concat(
        tables.values(),
        keys=tables.keys(),
        names=["group"],
    ).reset_index(level="group")
    return pooled
```

```python
import jax.numpy as jnp
import numpy as np
import numpyro
import numpyro.distributions as dist
from numpyro import plate


class BayesianOrdinalLCFA:
    """
    Bayesian ordinal CFA for longitudinal measurement invariance testing.

    Invariance hierarchy
    --------------------
    configural : loadings free per time, thresholds free per time
    metric     : loadings equal across time, thresholds free per time
                 (thresholds shrunk toward shared base via hierarchical prior)
    scalar     : loadings equal across time, thresholds equal across time

    Identification
    --------------
    Factor scale
        z ~ N(0,1) + LKJCholesky pins the factor *correlation* matrix.
        The product lambda * eta is scale-identified by fixing one anchor
        loading per factor to a positive constant (anchor_loading_value,
        default 1.0). All other loadings are free with a Normal prior.
        This is the standard CFA scale-fixing approach and eliminates the
        ridge between loading magnitude and latent factor variance.

        Note: with large loading_scale, non-anchor loadings may still have
        diffuse posteriors. Monitor ESS for anchored factors and keep
        loading_scale modest (default 0.3 is appropriate for most uses).

    Rotational
        Sparse CFA pattern (one primary factor per item) plus the fixed
        anchor loading removes rotational indeterminacy.

    Threshold location
        First threshold fixed to 0 at every time point per item.

    Metric threshold note
    ---------------------
    Classical metric invariance requires only that thresholds are *free*
    (not constrained equal). The Bayesian extension here shrinks time-specific
    thresholds toward a shared base via a hierarchical prior in log-increment
    space, which is strictly stronger than the classical definition. Users who
    prefer the classical definition should use configural thresholds with metric
    loadings — that is formally equivalent.

    One-item factors
    ----------------
    Factors with only one assigned item are statistically identified (the anchor
    loading is fixed) but contribute nothing to the factor covariance structure
    and are not interpretable as latent constructs. Avoid them in practice.

    NUTS cost note
    --------------
    eta is sampled explicitly as (N, F) latent variables. NUTS cost is O(N*F).
    True marginalization is analytically intractable for ordinal probit.
    For N > ~1000 consider SVI with a mean-field guide over eta.

    LOO warning
    -----------
    Observations within a person are NOT conditionally independent.
    Naive az.loo() over observation-level log-likelihoods is INVALID.
    Correct procedure:
        ll = idata.log_likelihood["y"]     # (chain, draw, N, J)
        ll_obs = ll.sum("y_dim_1")         # sum over J items -> (chain, draw, N)
        # group by pid, sum within person, then call az.loo()
    This is equivalent to leave-one-person-out cross-validation.

    Parameters
    ----------
    J                    : number of items
    F                    : number of factors
    T                    : number of time points
    n_persons            : number of unique persons (required, must be > 0)
    n_thr                : number of thresholds per item (= n_categories - 1).
                           Must satisfy n_thr + 1 == number of observed categories.
    item_to_factor       : array-like of int, length J; maps item -> factor (0-based)
    invariance           : "configural", "metric", or "scalar"
    anchor_items         : list of int, length F; one anchor item per factor.
                           Defaults to the first item assigned to each factor.
    anchor_loading_value : positive float; the fixed loading value for each
                           anchor item. Default 1.0. Identifies the latent scale.
    loading_loc          : prior mean for non-anchor loadings
    loading_scale        : prior SD for non-anchor loadings
    threshold_base_sd    : prior SD for log base-threshold increments
    threshold_dev_scale  : prior SD for time-specific log-increment deviations
                           (metric branch only)
    delta_min            : minimum positive spacing added after exp()
    sigma_e_prior        : scale for HalfNormal prior on occasion noise SD
    lkj_eta              : LKJCholesky concentration parameter
    """

    def __init__(
        self,
        J: int,
        F: int,
        T: int,
        n_persons: int,
        n_thr: int,
        item_to_factor,
        invariance: str = "metric",
        anchor_items=None,
        anchor_loading_value: float = 1.0,
        loading_loc: float = 0.7,
        loading_scale: float = 0.3,
        threshold_base_sd: float = 0.5,
        threshold_dev_scale: float = 0.2,
        delta_min: float = 1e-3,
        sigma_e_prior: float = 0.5,
        lkj_eta: float = 2.0,
    ):
        assert invariance in ("configural", "metric", "scalar"), (
            f"invariance must be one of configural/metric/scalar, got {invariance!r}"
        )
        assert isinstance(n_persons, int) and n_persons > 0, (
            "n_persons must be a positive integer"
        )
        assert len(item_to_factor) == J, (
            f"item_to_factor length {len(item_to_factor)} != J={J}"
        )
        assert n_thr >= 1, "n_thr must be >= 1 (requires at least 2 ordinal categories)"
        assert anchor_loading_value > 0, "anchor_loading_value must be positive"

        self.J = J
        self.F = F
        self.T = T
        self.P = n_persons
        self.n_thr = n_thr
        self.item_to_factor = np.array(item_to_factor, dtype=int)
        self.invariance = invariance
        self.anchor_loading_value = float(anchor_loading_value)
        self.loading_loc = loading_loc
        self.loading_scale = loading_scale
        self.threshold_base_sd = threshold_base_sd
        self.threshold_dev_scale = threshold_dev_scale
        self.delta_min = delta_min
        self.sigma_e_prior = sigma_e_prior
        self.lkj_eta = lkj_eta

        # Factor -> item groupings (plain NumPy, JIT-safe)
        self.items_per_factor = [
            np.where(self.item_to_factor == f)[0].tolist() for f in range(F)
        ]
        for f, items in enumerate(self.items_per_factor):
            assert len(items) > 0, f"Factor {f} has no items assigned in item_to_factor"

        # Anchor items: first item per factor by default
        if anchor_items is None:
            self.anchor_items = [items[0] for items in self.items_per_factor]
        else:
            anchor_items = list(anchor_items)
            assert len(anchor_items) == F, (
                f"anchor_items length {len(anchor_items)} != F={F}"
            )
            for f, idx in enumerate(anchor_items):
                assert self.item_to_factor[idx] == f, (
                    f"anchor_items[{f}]={idx} is not assigned to factor {f}"
                )
            self.anchor_items = anchor_items

    # ------------------------------------------------------------------
    # Loadings  ->  (T, J, F)
    # ------------------------------------------------------------------
    def _sample_loadings(self) -> jnp.ndarray:
        """
        Sparse loading matrix with scale identification via fixed anchor loadings.

        For each factor f:
          - The anchor item loading is fixed to +anchor_loading_value (a constant,
            not sampled). This pins the latent scale and eliminates the ridge
            between loading magnitude and factor variance.
          - All other items on factor f have freely sampled loadings with
            Normal(loading_loc, loading_scale) prior.

        configural : per-time loadings for non-anchor items
        metric     : time-invariant loadings for non-anchor items (broadcast)
        scalar     : identical to metric for loadings
        """
        lam_full = jnp.zeros((self.T, self.J, self.F))

        for f, items in enumerate(self.items_per_factor):
            items_np = np.array(items, dtype=int)
            anchor_idx = self.anchor_items[f]
            anchor_local_pos = np.where(items_np == anchor_idx)[0][0]

            non_anchor_items = np.array(
                [i for i in items if i != anchor_idx], dtype=int
            )
            non_anchor_local_pos = np.array(
                [k for k, i in enumerate(items_np) if i != anchor_idx], dtype=int
            )
            n_free = len(non_anchor_items)

            # Anchor loading: fixed positive constant, not sampled
            block = jnp.zeros((self.T, len(items_np)))
            block = block.at[:, anchor_local_pos].set(self.anchor_loading_value)

            if n_free > 0:
                if self.invariance == "configural":
                    free_lam = numpyro.sample(
                        f"lambda_{f}",
                        dist.Normal(self.loading_loc, self.loading_scale).expand(
                            [self.T, n_free]
                        ),
                    )                                                 # (T, n_free)
                else:  # metric or scalar: time-invariant
                    free_lam_base = numpyro.sample(
                        f"lambda_{f}",
                        dist.Normal(self.loading_loc, self.loading_scale).expand(
                            [n_free]
                        ),
                    )                                                 # (n_free,)
                    free_lam = jnp.broadcast_to(
                        free_lam_base[None, :], (self.T, n_free)
                    )                                                 # (T, n_free)

                block = block.at[:, non_anchor_local_pos].set(free_lam)

            lam_full = lam_full.at[:, items_np, f].set(block)

        return lam_full  # (T, J, F)

    # ------------------------------------------------------------------
    # Thresholds  ->  (T, J, n_thr)
    # ------------------------------------------------------------------
    def _sample_thresholds(self) -> jnp.ndarray:
        """
        Ordered thresholds with first threshold fixed to 0 at all time points.
        Ordering guaranteed in all branches via cumsum of strictly positive
        increments (exp(raw) + delta_min > 0).

        scalar
            Shared base thresholds broadcast across T.

        metric
            Hierarchical log-increment model: deviations added to raw_base in
            log space before exponentiation. Since exp(raw_base + raw_dev) > 0
            for any raw_dev, cumsum preserves ordering at every time point.
            raw_dev ~ Normal(0, threshold_dev_scale) shrinks time-specific
            thresholds toward the shared base (Bayesian extension of classical
            metric invariance; see class docstring).

        configural
            Fully independent ordered thresholds per time point.
        """
        n_thr = self.n_thr  # K - 1

        if self.invariance == "scalar":
            raw_base = numpyro.sample(
                "raw_base",
                dist.Normal(0.0, self.threshold_base_sd).expand([self.J, n_thr - 1]),
            )                                                         # (J, n_thr-1)
            delta_base = jnp.exp(raw_base) + self.delta_min          # (J, n_thr-1), > 0
            thr_base = jnp.concatenate(
                [jnp.zeros((self.J, 1)), jnp.cumsum(delta_base, axis=-1)],
                axis=-1,
            )                                                         # (J, n_thr)
            thresholds = jnp.broadcast_to(
                thr_base[None, :, :], (self.T, self.J, n_thr)
            )                                                         # (T, J, n_thr)

        elif self.invariance == "metric":
            raw_base = numpyro.sample(
                "raw_base",
                dist.Normal(0.0, self.threshold_base_sd).expand([self.J, n_thr - 1]),
            )                                                         # (J, n_thr-1)
            raw_dev = numpyro.sample(
                "raw_dev",
                dist.Normal(0.0, self.threshold_dev_scale).expand(
                    [self.T, self.J, n_thr - 1]
                ),
            )                                                         # (T, J, n_thr-1)
            # Deviations in log-increment space -> ordering preserved by construction
            log_increments = raw_base[None, :, :] + raw_dev          # (T, J, n_thr-1)
            delta_t = jnp.exp(log_increments) + self.delta_min       # (T, J, n_thr-1), > 0
            thr_t = jnp.cumsum(delta_t, axis=-1)                     # (T, J, n_thr-1)
            thresholds = jnp.concatenate(
                [jnp.zeros((self.T, self.J, 1)), thr_t],
                axis=-1,
            )                                                         # (T, J, n_thr)

        else:  # configural
            raw_dev = numpyro.sample(
                "raw_dev",
                dist.Normal(0.0, self.threshold_base_sd).expand(
                    [self.T, self.J, n_thr - 1]
                ),
            )                                                         # (T, J, n_thr-1)
            delta_t = jnp.exp(raw_dev) + self.delta_min              # (T, J, n_thr-1), > 0
            thr_t = jnp.cumsum(delta_t, axis=-1)                     # (T, J, n_thr-1)
            thresholds = jnp.concatenate(
                [jnp.zeros((self.T, self.J, 1)), thr_t],
                axis=-1,
            )                                                         # (T, J, n_thr)

        return thresholds  # (T, J, n_thr)

    # ------------------------------------------------------------------
    # Model
    # ------------------------------------------------------------------
    def model(self, y, pid, time_idx):
        """
        Parameters
        ----------
        y        : ndarray (N, J)   observed ordinal responses, 0-indexed
        pid      : ndarray (N,)     person index per observation, 0..P-1
        time_idx : ndarray (N,)     time index per observation, 0..T-1
        """
        N = y.shape[0]

        # Runtime check: n_thr must match the categories present in y.
        # y is 0-indexed so max valid category is n_thr (= n_categories - 1).

        assert y.shape[1] == self.J, (
            f"y has {y.shape[1]} items but model was configured with J={self.J}"
        )
        observed_max = int(np.asarray(y).max())
        assert observed_max <= self.n_thr, (
            f"y contains category {observed_max} but n_thr={self.n_thr} implies "
            f"a maximum category of {self.n_thr} (n_categories={self.n_thr + 1}). "
            f"Set n_thr = n_categories - 1."
        )

        # ── Factor correlation (Cholesky of correlation matrix) ───────────────
        # LKJCholesky gives the correlation matrix (unit diagonal).
        # Factor *scale* is identified by fixed anchor loadings, not by
        # constraining factor variances.
        L_T = numpyro.sample("L_T", dist.LKJCholesky(self.F, self.lkj_eta))

        # ── Person-level latent traits  (P, F) ───────────────────────────────
        # plate declares conditional independence across persons.
        # Matrix multiply is OUTSIDE the plate to avoid per-element tracing.
        with plate("persons", self.P):
            z_person = numpyro.sample(
                "z_person",
                dist.Normal(0.0, 1.0).expand([self.F]).to_event(1),
            )                                                         # (P, F)
        eta_person = z_person @ L_T.T                                 # (P, F)
        numpyro.deterministic("eta_person", eta_person)
        # Post-hoc stability: use idata.posterior["eta_person"] (chain, draw, P, F)
        # and align persons across time points using pid + time_idx externally.

        # ── Occasion noise  (diagonal, estimated) ────────────────────────────
        # Identifiable: HalfNormal penalises large values; anchor loadings pin
        # the total response scale so sigma_e is estimated relative to a fixed
        # reference.
        # NOTE: eta is sampled explicitly -> O(N x F) latent variables in NUTS.
        # True marginalization is analytically intractable for ordinal probit.
        # For N > ~1000 consider SVI with a mean-field guide over eta.
        sigma_e = numpyro.sample(
            "sigma_e",
            dist.HalfNormal(self.sigma_e_prior).expand([self.F]),
        )                                                             # (F,)
        eta = numpyro.sample(
            "eta",
            dist.Normal(eta_person[pid], sigma_e).to_event(1),
        )                                                             # (N, F)

        # ── Loadings and thresholds ───────────────────────────────────────────
        lam_full = self._sample_loadings()                            # (T, J, F)
        thresholds = self._sample_thresholds()                        # (T, J, n_thr)

        lam_row = lam_full[time_idx]                                  # (N, J, F)
        thr_row = thresholds[time_idx]                                # (N, J, n_thr)

        mu = jnp.einsum("nf,njf->nj", eta, lam_row)                  # (N, J)

        # ── Vectorised ordinal probit likelihood ─────────────────────────────
        numpyro.sample("y", dist.OrderedProbit(mu, thr_row), obs=y)

        # ── Diagnostics ───────────────────────────────────────────────────────
        numpyro.deterministic("factor_corr", L_T @ L_T.T)
        # factor_corr: within-occasion factor correlation matrix.
        # Cross-time trait stability must be computed post-hoc from eta_person.

        # LOO WARNING ─────────────────────────────────────────────────────────
        # Observations within a person are NOT conditionally independent.
        # Naive az.loo() over observation-level log-likelihoods is INVALID.
        # Correct procedure:
        #   ll = idata.log_likelihood["y"]          # (chain, draw, N, J)
        #   ll_obs = ll.sum("y_dim_1")              # -> (chain, draw, N)
        #   # group ll_obs by pid, sum within person, then call az.loo()
        # This is equivalent to leave-one-person-out cross-validation.
```

```python
# ============================================================
# FULL UPDATED SCRIPT (NumPyro / JAX) — v7
# Longitudinal Bayesian ordinal Trait–Occasion CFA (TRUE ordinal)
# Invariance series: configural / metric / scalar (+ partial options)
#
# v7 changes (fixing the three issues you flagged):
# 1) compare_by_cached_loo_with_se no longer recomputes LOO if you pass diag=
#    - diag: Optional[pd.DataFrame] parameter added.
# 2) attach_loglik_personlevel now accepts explicit n_person
#    - avoids undercount if some persons have no rows (e.g., dropout/complete missingness at wave 2)
# 3) _targeted_rhat_ess uses az.rhat(method="rank") with a compatibility fallback
#    - if older ArviZ doesn't support method=, it retries without it.
#
# Other behavior unchanged from v6:
# - row-level LOO default (person-wave row)
# - cell-level LOO uses observed-only cells (filters missing)
# - person-level LOO provided but PSIS often fails (use as sensitivity only)
# - structural convergence gating excludes z_trait / z_occasion by default
# - cached LOO objects stored in results (side effect documented)
# - SE(delta elpd) computed from pointwise loo_i differences (ddof=0)
# ============================================================

from __future__ import annotations
from dataclasses import dataclass
from typing import Sequence, Tuple, Optional, Dict, Any, List, Set

import numpy as np
import pandas as pd
import jax
import jax.numpy as jnp
from jax.nn import softplus
from jax.scipy.special import log_ndtr

import numpyro
import numpyro.distributions as dist
from numpyro.distributions import Distribution, constraints
from numpyro.handlers import mask as mask_handler
from numpyro.infer import NUTS, MCMC
from numpyro.infer.initialization import init_to_value
from numpyro.infer.util import log_likelihood

import arviz as az
import xarray as xr


# ----------------------------
# Numerics helpers
# ----------------------------
def _log1mexp(x: jnp.ndarray) -> jnp.ndarray:
    log_half = jnp.asarray(-0.6931471805599453)
    return jnp.where(x < log_half, jnp.log1p(-jnp.exp(x)), jnp.log(-jnp.expm1(x)))


def _sample_truncnorm_pos(name: str, loc: float, scale: float, shape: int | tuple[int, ...]):
    shape_tup = (shape,) if isinstance(shape, int) else tuple(shape)
    try:
        d = dist.TruncatedNormal(loc=loc, scale=scale, low=0.0)
    except TypeError:
        try:
            d = dist.TruncatedNormal(loc, scale, 0.0, jnp.inf)
        except TypeError as e:
            raise TypeError("NumPyro TruncatedNormal API mismatch; pin/upgrade numpyro.") from e
    return numpyro.sample(name, d.expand(shape_tup))


# ----------------------------
# Ordered Probit distribution
# ----------------------------
class OrderedProbit(Distribution):
    arg_constraints = {"mu": constraints.real, "thresholds": constraints.ordered_vector}

    def __init__(self, mu, thresholds, validate_args=None):
        self.mu = jnp.asarray(mu)
        thr = jnp.asarray(thresholds)

        if thr.ndim < 1:
            raise ValueError("thresholds must have at least 1 dimension (last dim = n_thr).")

        self.n_thr = int(thr.shape[-1])
        self.n_cats = self.n_thr + 1

        if thr.ndim == 1:
            thr = thr.reshape((1,) * self.mu.ndim + (self.n_thr,))
        elif thr.ndim == 2:
            n_items = self.mu.shape[-1]
            if thr.shape[0] != n_items:
                raise ValueError(f"thresholds shape mismatch: got {thr.shape}, expected [{n_items},{self.n_thr}]")
            thr = thr.reshape((1,) * (self.mu.ndim - 1) + (n_items, self.n_thr))

        self.thresholds = jnp.broadcast_to(thr, self.mu.shape + (self.n_thr,))
        super().__init__(batch_shape=self.mu.shape, event_shape=(), validate_args=validate_args)

    @property
    def support(self):
        return constraints.integer_interval(0, self.n_cats - 1)

    def log_prob(self, value):
        y = jnp.asarray(value).astype(jnp.int32)
        in_range = (y >= 0) & (y < self.n_cats)

        y_idx = jnp.clip(y, 0, self.n_cats - 1)
        y_exp = y_idx[..., None]

        t = self.thresholds - self.mu[..., None]
        neg_inf = jnp.full(self.mu.shape + (1,), -jnp.inf)
        pos_inf = jnp.full(self.mu.shape + (1,),  jnp.inf)
        t_full = jnp.concatenate([neg_inf, t, pos_inf], axis=-1)

        lower = jnp.take_along_axis(t_full[..., :-1], y_exp, axis=-1)[..., 0]
        upper = jnp.take_along_axis(t_full[...,  1:], y_exp, axis=-1)[..., 0]

        ordered = upper > lower
        log_cdf_u = log_ndtr(upper)
        log_cdf_l = log_ndtr(lower)

        d = jnp.minimum(log_cdf_l - log_cdf_u, 0.0)
        eps = jnp.asarray(1e-12)
        d_safe = jnp.where(d > -eps, -eps, d)

        log_p_ordered = log_cdf_u + _log1mexp(d_safe)
        log_p = jnp.where(ordered, log_p_ordered, -jnp.inf)
        log_p = jnp.where(in_range, log_p, -jnp.inf)
        return log_p


# ----------------------------
# Long data container
# ----------------------------
@dataclass(frozen=True)
class LongData:
    y: jnp.ndarray
    pid: jnp.ndarray
    tid: jnp.ndarray
    obs_mask: jnp.ndarray


# ============================================================
# MODEL CLASS
# ============================================================
class BayesianOrdinalTraitOccasionCFA:
    def __init__(
        self,
        n_factors: int,
        n_cats: int,
        n_times: int,
        n_person: int,
        item_to_factor_indices: Sequence[Sequence[int]],
        marker_items: Optional[Sequence[int]] = None,
        invariance: str = "metric",
        free_loading_pairs: Optional[Sequence[Tuple[int, int]]] = None,
        free_threshold_items: Optional[Sequence[int]] = None,
        loading_dev_scale: float = 0.15,
        threshold_loc_dev_scale: float = 0.10,
        cfa_loading_loc: float = 0.7,
        loadings_scale: float = 0.5,
        allow_negative_loadings: bool = False,
        lkj_eta_T: float = 2.0,
        lkj_eta_O: float = 2.0,
        alpha_sd: float = 1.0,
        trait_df: float = 7.0,
        occas_df: float = 7.0,
        sigO_sd: float = 0.5,
        delta_sd: float = 0.3,
        min_delta: float = 1e-3,
        threshold_location: str = "fixed0",
        t0_sd: float = 1.0,
    ):
        self.F = int(n_factors)
        self.K = int(n_cats)
        self.T = int(n_times)
        self.P = int(n_person)

        self.item_to_factor_indices = [list(map(int, lst)) for lst in item_to_factor_indices]

        self.invariance = invariance.lower()
        if self.invariance not in ("configural", "metric", "scalar"):
            raise ValueError("invariance must be one of: configural, metric, scalar")

        self.free_loading_pairs = None if free_loading_pairs is None else [tuple(map(int, p)) for p in free_loading_pairs]
        self.free_threshold_items = None if free_threshold_items is None else [int(i) for i in free_threshold_items]

        self.loading_dev_scale = float(loading_dev_scale)
        self.threshold_loc_dev_scale = float(threshold_loc_dev_scale)

        self.cfa_loading_loc = float(cfa_loading_loc)
        self.loadings_scale = float(loadings_scale)
        self.allow_negative_loadings = bool(allow_negative_loadings)

        self.lkj_eta_T = float(lkj_eta_T)
        self.lkj_eta_O = float(lkj_eta_O)

        self.alpha_sd = float(alpha_sd)
        self.trait_df = float(trait_df)
        self.occas_df = float(occas_df)
        self.sigO_sd = float(sigO_sd)

        self.delta_sd = float(delta_sd)
        self.min_delta = float(min_delta)
        self.threshold_location = threshold_location.lower()
        if self.threshold_location not in ("fixed0", "free_t0"):
            raise ValueError("threshold_location must be fixed0 or free_t0")
        self.t0_sd = float(t0_sd)

        if marker_items is None:
            marker_items = [lst[0] for lst in self.item_to_factor_indices]
        if len(marker_items) != self.F:
            raise ValueError("marker_items must have length n_factors")
        self.marker_items = np.asarray([int(i) for i in marker_items], dtype=int)

        pairs = sorted({(i, f) for f, items in enumerate(self.item_to_factor_indices) for i in items})
        self._pairs = pairs
        self._pair_items = np.asarray([p[0] for p in pairs], dtype=np.int32)
        self._pair_facts = np.asarray([p[1] for p in pairs], dtype=np.int32)
        self._n_pairs = len(pairs)

        marker_pairs = {(int(self.marker_items[f]), int(f)) for f in range(self.F)}
        is_marker = np.asarray([(p in marker_pairs) for p in pairs], dtype=bool)
        self._marker_pos = np.where(is_marker)[0].astype(np.int32)
        self._nonmarker_pos = np.where(~is_marker)[0].astype(np.int32)

        self._free_loading_set: Optional[Set[Tuple[int, int]]] = None if self.free_loading_pairs is None else set(self.free_loading_pairs)
        self._free_threshold_item_set: Optional[Set[int]] = None if self.free_threshold_items is None else set(self.free_threshold_items)

        if self.invariance == "configural":
            dev_pos = self._nonmarker_pos
        else:
            if self._free_loading_set is None:
                dev_pos = np.asarray([], dtype=np.int32)
            else:
                dev_pos = np.asarray(
                    [i for i in self._nonmarker_pos.tolist() if self._pairs[i] in self._free_loading_set],
                    dtype=np.int32,
                )
        self._dev_pos = dev_pos

        if self.invariance in ("configural", "metric"):
            self._thr_free_items: Optional[np.ndarray] = None
        else:
            if self._free_threshold_item_set is None or len(self._free_threshold_item_set) == 0:
                self._thr_free_items = np.asarray([], dtype=np.int32)
            else:
                self._thr_free_items = np.asarray(sorted(self._free_threshold_item_set), dtype=np.int32)

    def _sample_thresholds(self, n_items: int, time_varying: bool):
        n_thr = self.K - 1
        if self.K == 2:
            return jnp.zeros((self.T, n_items, 1)) if time_varying else jnp.zeros((n_items, 1))

        if self.threshold_location == "fixed0":
            t0 = jnp.zeros((n_items,))
        else:
            t0 = numpyro.sample("t0", dist.Normal(0.0, self.t0_sd).expand([n_items]))

        raw_delta = numpyro.sample("raw_delta", dist.Normal(0.0, self.delta_sd).expand([n_items, n_thr - 1]))
        delta = softplus(raw_delta) + self.min_delta
        t_rest = jnp.cumsum(delta, axis=1)
        thr0 = jnp.concatenate([t0[:, None], t0[:, None] + t_rest], axis=1)

        if not time_varying:
            return thr0

        if self.invariance in ("configural", "metric"):
            free_items = jnp.arange(n_items, dtype=jnp.int32)
        else:
            free_items = jnp.asarray(self._thr_free_items, dtype=jnp.int32)

        thr_t = jnp.broadcast_to(thr0[None, :, :], (self.T, n_items, n_thr))
        n_free = int(free_items.shape[0])

        if self.T > 1 and n_free > 0:
            thr_loc_dev = numpyro.sample(
                "thr_loc_dev",
                dist.Normal(0.0, self.threshold_loc_dev_scale).expand([self.T - 1, n_free]),
            )
            thr_t = thr_t.at[1:, free_items, :].add(thr_loc_dev[:, :, None])

        return thr_t

    def model(self, long_data: LongData):
        y = long_data.y.astype(jnp.int32)
        pid = long_data.pid.astype(jnp.int32)
        tid = long_data.tid.astype(jnp.int32)
        obs_mask = long_data.obs_mask.astype(jnp.bool_)

        n_rows, n_items = y.shape
        F, T = self.F, self.T

        L_T = numpyro.sample("L_T", dist.LKJCholesky(F, self.lkj_eta_T))
        numpyro.deterministic("factor_corr_T", L_T @ L_T.T)

        L_O = numpyro.sample("L_O", dist.LKJCholesky(F, self.lkj_eta_O))
        numpyro.deterministic("factor_corr_O", L_O @ L_O.T)

        if T == 1:
            alpha = jnp.zeros((1, F))
        else:
            alpha_dev = numpyro.sample("alpha_dev", dist.Normal(0.0, self.alpha_sd).expand([T - 1, F]))
            alpha = jnp.concatenate([jnp.zeros((1, F)), alpha_dev], axis=0)
        numpyro.deterministic("alpha", alpha)

        zT = numpyro.sample("z_trait", dist.StudentT(self.trait_df, 0.0, 1.0).expand([self.P, F]))
        T_lat = zT @ L_T.T

        sigO = numpyro.sample("sig_O", dist.HalfNormal(self.sigO_sd).expand([F]))
        numpyro.deterministic("ICC_marginal_per_factor", 1.0 / (1.0 + sigO**2))

        zO = numpyro.sample("z_occasion", dist.StudentT(self.occas_df, 0.0, 1.0).expand([n_rows, F]))
        O_lat = (zO * sigO[None, :]) @ L_O.T

        eta = alpha[tid, :] + T_lat[pid, :] + O_lat
        numpyro.deterministic("eta", eta)

        n_pairs = self._n_pairs
        n_nonmarker = int(self._nonmarker_pos.shape[0])

        if n_nonmarker > 0:
            if self.allow_negative_loadings:
                lam_nm = numpyro.sample(
                    "lam_pairs_nonmarker",
                    dist.Normal(self.cfa_loading_loc, self.loadings_scale).expand([n_nonmarker]),
                )
            else:
                lam_nm = _sample_truncnorm_pos(
                    "lam_pairs_nonmarker",
                    loc=self.cfa_loading_loc,
                    scale=self.loadings_scale,
                    shape=n_nonmarker,
                )
        else:
            lam_nm = jnp.zeros((0,))

        lam_pairs0 = jnp.zeros((n_pairs,))
        if self._marker_pos.size > 0:
            lam_pairs0 = lam_pairs0.at[jnp.asarray(self._marker_pos)].set(1.0)
        lam_pairs0 = lam_pairs0.at[jnp.asarray(self._nonmarker_pos)].set(lam_nm)

        dev_pos = self._dev_pos
        n_dev = int(dev_pos.shape[0])

        lam_pairs_t = jnp.broadcast_to(lam_pairs0[None, :], (T, n_pairs))
        if T > 1 and n_dev > 0:
            z_lam_dev = numpyro.sample("z_lam_dev", dist.Normal(0.0, 1.0).expand([T - 1, n_dev]))
            lam_pairs_t = lam_pairs_t.at[1:, jnp.asarray(dev_pos)].add(z_lam_dev * self.loading_dev_scale)

        lam_t = jnp.zeros((T, n_items, F))
        lam_t = lam_t.at[:, jnp.asarray(self._pair_items), jnp.asarray(self._pair_facts)].set(lam_pairs_t)

        for f in range(F):
            lam_t = lam_t.at[:, self.marker_items[f], f].set(1.0)
        numpyro.deterministic("loadings_t", lam_t)

        if self.invariance in ("configural", "metric"):
            thresholds_t = self._sample_thresholds(n_items, time_varying=True)
            thresholds_row = thresholds_t[tid, :, :]
            numpyro.deterministic("thresholds_t", thresholds_t)
        else:
            time_varying = bool(self._free_threshold_item_set) if self._free_threshold_item_set is not None else False
            if time_varying:
                thresholds_t = self._sample_thresholds(n_items, time_varying=True)
                thresholds_row = thresholds_t[tid, :, :]
                numpyro.deterministic("thresholds_t", thresholds_t)
            else:
                thresholds = self._sample_thresholds(n_items, time_varying=False)
                thresholds_row = jnp.broadcast_to(thresholds[None, :, :], (n_rows, n_items, thresholds.shape[-1]))
                numpyro.deterministic("thresholds", thresholds)

        lam_row = lam_t[tid, :, :]
        mu = jnp.einsum("rf,rif->ri", eta, lam_row)

        with mask_handler(mask=obs_mask):
            numpyro.sample("obs", OrderedProbit(mu, thresholds_row, validate_args=False), obs=y)


# ============================================================
# FITTING / CONVERGENCE META
# ============================================================
def _run_with_extras_compat(mcmc: MCMC, key, *, model_kwargs: dict, extra_fields: Tuple[str, ...]):
    try:
        return mcmc.run(key, **model_kwargs, extra_fields=extra_fields)
    except TypeError:
        return mcmc.run(key, **model_kwargs)


def _targeted_rhat_ess(
    idata: az.InferenceData,
    targeted_vars: List[str],
) -> Tuple[float, float]:
    present = [v for v in targeted_vars if v in idata.posterior.data_vars]
    if not present:
        return float("nan"), float("nan")

    # ArviZ compatibility: method="rank" may not exist in older versions
    try:
        rhat_ds = az.rhat(idata, var_names=present, method="rank")
    except TypeError:
        rhat_ds = az.rhat(idata, var_names=present)

    ess_ds = az.ess(idata, var_names=present, method="bulk")

    max_rhat = float(np.nanmax([np.nanmax(v.to_numpy()) for v in rhat_ds.data_vars.values()]))
    min_ess = float(np.nanmin([np.nanmin(v.to_numpy()) for v in ess_ds.data_vars.values()]))
    return max_rhat, min_ess


def run_mcmc_with_meta(
    model_obj: BayesianOrdinalTraitOccasionCFA,
    long_data: LongData,
    seed: int,
    *,
    num_warmup: int = 2000,
    num_samples: int = 4000,
    num_chains: int = 4,
    chain_method: str = "sequential",
    target_accept_prob: float = 0.95,
    max_tree_depth: int = 12,
    init_values: Optional[Dict[str, Any]] = None,
    rhat_gate: float = 1.05,
    ess_gate: float = 100.0,
    targeted_vars: Optional[List[str]] = None,
    print_diagnostics: bool = True,
) -> Tuple[MCMC, NUTS, az.InferenceData, Dict[str, Any]]:
    nuts = NUTS(
        model_obj.model,
        target_accept_prob=target_accept_prob,
        max_tree_depth=max_tree_depth,
        init_strategy=init_to_value(values=init_values) if init_values is not None else None,
    )

    mcmc = MCMC(
        nuts,
        num_warmup=num_warmup,
        num_samples=num_samples,
        num_chains=num_chains,
        chain_method=chain_method,
        progress_bar=True,
    )

    if chain_method == "sequential":
        key = jax.random.PRNGKey(seed)
    else:
        key = jax.random.split(jax.random.PRNGKey(seed), num_chains)

    _run_with_extras_compat(
        mcmc,
        key,
        model_kwargs={"long_data": long_data},
        extra_fields=("diverging", "num_steps", "tree_depth"),
    )

    n_div = None
    max_steps = None
    hit_td = None
    try:
        extra = mcmc.get_extra_fields()
        div = np.asarray(extra.get("diverging", 0))
        steps = np.asarray(extra.get("num_steps", 0))
        td = np.asarray(extra.get("tree_depth", np.nan))

        n_div = int(div.sum()) if np.ndim(div) else int(div)
        max_steps = int(steps.max()) if np.ndim(steps) else int(steps)
        hit_td = int(np.sum(td >= max_tree_depth)) if np.ndim(td) else int(td >= max_tree_depth)

        if print_diagnostics:
            print(f"Sampler: divergences={n_div}, max_num_steps={max_steps}, tree_depth_hit={hit_td}")
    except Exception as e:
        if print_diagnostics:
            print(f"Sampler: could not read extra_fields ({repr(e)})")

    idata = az.from_numpyro(mcmc)

    if targeted_vars is None:
        targeted_vars = [
            "L_T", "L_O", "alpha_dev", "sig_O",
            "lam_pairs_nonmarker", "z_lam_dev",
            "raw_delta", "thr_loc_dev", "t0",
        ]

    max_rhat_struct, min_ess_struct = np.nan, np.nan
    converged_struct = False
    try:
        max_rhat_struct, min_ess_struct = _targeted_rhat_ess(idata, targeted_vars=targeted_vars)
        converged_struct = (
            (np.isfinite(max_rhat_struct) and max_rhat_struct <= rhat_gate)
            and (np.isfinite(min_ess_struct) and min_ess_struct >= ess_gate)
            and (n_div == 0 if n_div is not None else True)
        )
        if print_diagnostics:
            print(f"Convergence(struct): max_Rhat={max_rhat_struct:.3f}, min_ESS_bulk={min_ess_struct:.0f}, ok={converged_struct}")
    except Exception as e:
        if print_diagnostics:
            print(f"Convergence(struct): failed ({repr(e)})")

    meta = {
        "divergences": n_div,
        "max_num_steps": max_steps,
        "max_tree_depth": int(max_tree_depth),
        "tree_depth_hit": hit_td,
        "max_rhat_struct": float(max_rhat_struct) if np.isfinite(max_rhat_struct) else None,
        "min_ess_struct": float(min_ess_struct) if np.isfinite(min_ess_struct) else None,
        "converged_struct": bool(converged_struct),
        "targeted_vars": targeted_vars,
    }
    return mcmc, nuts, idata, meta


# ============================================================
# MASKED LOG-LIKELIHOOD EXTRACTION
# ============================================================
def masked_loglik_cells(nuts_model, samples_flat: Dict[str, jnp.ndarray], long_data: LongData) -> np.ndarray:
    ll_raw = log_likelihood(nuts_model, samples_flat, long_data=long_data)["obs"]
    ll_raw_np = np.asarray(ll_raw)
    mask_np = np.asarray(long_data.obs_mask).astype(bool)
    return np.where(mask_np[None, :, :], ll_raw_np, 0.0)


# ============================================================
# ATTACH LOG-LIKELIHOOD TO IDATA
# ============================================================
def _assign_loglik(idata: az.InferenceData, name: str, ll_xr: xr.DataArray) -> az.InferenceData:
    existing_ll = getattr(idata, "log_likelihood", None)
    if existing_ll is None or (hasattr(existing_ll, "data_vars") and len(existing_ll.data_vars) == 0):
        new_ll = xr.Dataset({name: ll_xr})
    else:
        new_ll = existing_ll.assign({name: ll_xr})
    return idata.assign(log_likelihood=new_ll)


def attach_loglik_rowlevel(
    idata: az.InferenceData,
    ll_masked_cells: np.ndarray,
    *,
    num_chains: int,
    num_samples: int,
    var_name: str = "obs_row",
) -> az.InferenceData:
    n_draws, n_rows, _ = ll_masked_cells.shape
    if n_draws != num_chains * num_samples:
        raise ValueError(f"ll first dim {n_draws} != num_chains*num_samples {num_chains*num_samples}")
    ll_row = ll_masked_cells.sum(axis=-1).reshape(num_chains, num_samples, n_rows)
    ll_xr = xr.DataArray(ll_row, dims=["chain", "draw", "obs_id"], name=var_name)
    return _assign_loglik(idata, var_name, ll_xr)


def attach_loglik_celllevel_observed_only(
    idata: az.InferenceData,
    ll_masked_cells: np.ndarray,
    long_data: LongData,
    *,
    num_chains: int,
    num_samples: int,
    var_name: str = "obs_cell",
) -> Tuple[az.InferenceData, np.ndarray]:
    n_draws, n_rows, n_items = ll_masked_cells.shape
    if n_draws != num_chains * num_samples:
        raise ValueError(f"ll first dim {n_draws} != num_chains*num_samples {num_chains*num_samples}")

    mask_flat = np.asarray(long_data.obs_mask).astype(bool).ravel()
    obs_indices = np.where(mask_flat)[0]

    ll_flat = ll_masked_cells.reshape(n_draws, n_rows * n_items)[:, obs_indices]
    ll_cell = ll_flat.reshape(num_chains, num_samples, ll_flat.shape[1])

    ll_xr = xr.DataArray(ll_cell, dims=["chain", "draw", "obs_id"], name=var_name)
    return _assign_loglik(idata, var_name, ll_xr), obs_indices


def attach_loglik_personlevel(
    idata: az.InferenceData,
    ll_masked_cells: np.ndarray,
    long_data: LongData,
    *,
    n_person: int,                 # <-- explicit (fixes undercount)
    num_chains: int,
    num_samples: int,
    var_name: str = "obs_person",
) -> az.InferenceData:
    n_draws, n_rows, _ = ll_masked_cells.shape
    if n_draws != num_chains * num_samples:
        raise ValueError(f"ll first dim {n_draws} != num_chains*num_samples {num_chains*num_samples}")

    pid_np = np.asarray(long_data.pid).astype(int)
    if pid_np.max() >= n_person:
        raise ValueError(f"pid contains value {pid_np.max()} but n_person={n_person}")

    ll_row = ll_masked_cells.sum(axis=-1)  # [n_draws, n_rows]

    # one-hot [n_rows, n_person] -> fast BLAS multiply
    onehot = np.zeros((n_rows, n_person), dtype=ll_row.dtype)
    onehot[np.arange(n_rows), pid_np] = 1.0

    ll_person = ll_row @ onehot  # [n_draws, n_person]
    ll_person = ll_person.reshape(num_chains, num_samples, n_person)

    ll_xr = xr.DataArray(ll_person, dims=["chain", "draw", "obs_id"], name=var_name)
    return _assign_loglik(idata, var_name, ll_xr)


# ============================================================
# CACHED LOO DIAGNOSTICS + COMPARISON
# ============================================================
def loo_diagnostics_cached(
    results: Dict[str, Dict[str, Any]],
    *,
    var_name: str = "obs_row",
    pointwise: bool = True,
    scale: str = "log",
) -> pd.DataFrame:
    """
    Computes PSIS-LOO per model and caches loo objects into results[model][f"loo_{var_name}"].
    NOTE: MUTATES results (documented by design).
    """
    if scale not in ("log", "negative_log", "deviance"):
        raise ValueError('scale must be one of: "log", "negative_log", "deviance"')

    rows = []
    for name, res in results.items():
        idata = res.get("idata")
        meta = res.get("meta", {})
        converged = bool(meta.get("converged_struct", False))

        ok_ll = (
            (idata is not None)
            and (getattr(idata, "log_likelihood", None) is not None)
            and (hasattr(idata.log_likelihood, "__contains__"))
            and (var_name in idata.log_likelihood)
        )
        if not ok_ll:
            rows.append({
                "model": name,
                "converged_struct": converged,
                "elpd_loo": np.nan,
                "se": np.nan,
                "p_loo": np.nan,
                "p_loo_ratio": np.nan,
                "pareto_k_max": np.nan,
                "pareto_k_p99": np.nan,
                "n_k_gt_0p7": np.nan,
                "warning": True,
                "error": f"missing log_likelihood[{var_name}]",
            })
            continue

        try:
            loo_res = az.loo(idata, var_name=var_name, pointwise=pointwise, scale=scale)
            res[f"loo_{var_name}"] = loo_res  # cache

            pk = loo_res.pareto_k
            pk_np = np.asarray(pk.values if hasattr(pk, "values") else pk)
            n_obs = pk_np.size

            se_val = float(getattr(loo_res, "se", None) or getattr(loo_res, "elpd_loo_se", np.nan))

            rows.append({
                "model": name,
                "converged_struct": converged,
                "elpd_loo": float(loo_res.elpd_loo),
                "se": se_val,
                "p_loo": float(loo_res.p_loo),
                "p_loo_ratio": float(loo_res.p_loo) / float(n_obs) if n_obs > 0 else np.nan,
                "pareto_k_max": float(np.nanmax(pk_np)) if n_obs else np.nan,
                "pareto_k_p99": float(np.nanpercentile(pk_np, 99)) if n_obs else np.nan,
                "n_k_gt_0p7": int(np.sum(pk_np > 0.7)) if n_obs else np.nan,
                "warning": bool(getattr(loo_res, "warning", False)),
                "error": "",
            })
        except Exception as e:
            rows.append({
                "model": name,
                "converged_struct": converged,
                "elpd_loo": np.nan,
                "se": np.nan,
                "p_loo": np.nan,
                "p_loo_ratio": np.nan,
                "pareto_k_max": np.nan,
                "pareto_k_p99": np.nan,
                "n_k_gt_0p7": np.nan,
                "warning": True,
                "error": repr(e),
            })

    df = pd.DataFrame(rows).set_index("model")
    return df.sort_values(["converged_struct", "elpd_loo"], ascending=[False, False])


def compare_by_cached_loo_with_se(
    results: Dict[str, Dict[str, Any]],
    *,
    var_name: str = "obs_row",
    diag: Optional[pd.DataFrame] = None,          # <-- fix: accept precomputed diag
    pareto_k_gate: float = 0.7,
    require_converged: bool = True,
) -> pd.DataFrame:
    """
    Compare models using cached pointwise loo_i (no az.compare recompute).
    If diag is None, it will compute/cached LOO via loo_diagnostics_cached.
    """
    if diag is None:
        diag = loo_diagnostics_cached(results, var_name=var_name, pointwise=True, scale="log")

    kept = diag.copy()

    if require_converged:
        kept = kept[kept["converged_struct"] == True]

    kept = kept[np.isfinite(kept["elpd_loo"])]
    kept = kept[np.isfinite(kept["pareto_k_max"])]
    kept = kept[kept["pareto_k_max"] <= pareto_k_gate]

    if kept.shape[0] < 2:
        raise RuntimeError("No models available after filtering (loglik/convergence/Pareto-k gate).")

    best_name = kept["elpd_loo"].idxmax()
    best_elpd = float(kept.loc[best_name, "elpd_loo"])

    loo_best = results[best_name].get(f"loo_{var_name}", None)
    if loo_best is None:
        raise RuntimeError(f"Best model {best_name} has no cached loo_{var_name} object.")

    loo_i_best = np.asarray(loo_best.loo_i.values if hasattr(loo_best.loo_i, "values") else loo_best.loo_i)
    n = loo_i_best.size

    rows = []
    for name in kept.index:
        loo_obj = results[name].get(f"loo_{var_name}", None)
        if loo_obj is None:
            continue
        loo_i = np.asarray(loo_obj.loo_i.values if hasattr(loo_obj.loo_i, "values") else loo_obj.loo_i)
        if loo_i.size != n:
            raise ValueError(f"Pointwise size mismatch: {name} has {loo_i.size}, best has {n}")

        elpd = float(kept.loc[name, "elpd_loo"])
        delta = elpd - best_elpd

        if name == best_name:
            se_delta = 0.0
            z = 0.0
        else:
            diff_i = loo_i - loo_i_best
            se_delta = float(np.sqrt(n * np.var(diff_i, ddof=0))) if n > 1 else np.nan
            z = delta / se_delta if (np.isfinite(se_delta) and se_delta > 0) else np.nan

        rows.append({
            "model": name,
            "elpd_loo": elpd,
            "delta_elpd": delta,
            "se_elpd": float(kept.loc[name, "se"]) if np.isfinite(kept.loc[name, "se"]) else np.nan,
            "se_delta_elpd": se_delta,
            "z_delta": z,
            "p_loo": float(kept.loc[name, "p_loo"]) if np.isfinite(kept.loc[name, "p_loo"]) else np.nan,
            "p_loo_ratio": float(kept.loc[name, "p_loo_ratio"]) if np.isfinite(kept.loc[name, "p_loo_ratio"]) else np.nan,
            "pareto_k_max": float(kept.loc[name, "pareto_k_max"]),
            "n_k_gt_0p7": int(kept.loc[name, "n_k_gt_0p7"]) if np.isfinite(kept.loc[name, "n_k_gt_0p7"]) else np.nan,
            "converged_struct": bool(kept.loc[name, "converged_struct"]),
        })

    return pd.DataFrame(rows).sort_values("elpd_loo", ascending=False).reset_index(drop=True)


# ============================================================
# INVARIANCE SERIES FIT
# ============================================================
def fit_invariance_series(
    long_data: LongData,
    *,
    n_factors: int,
    n_cats: int,
    n_times: int,
    n_person: int,
    item_to_factor_indices: Sequence[Sequence[int]],
    marker_items: Optional[Sequence[int]] = None,
    free_loading_pairs: Optional[Sequence[Tuple[int, int]]] = None,
    free_threshold_items: Optional[Sequence[int]] = None,
    invariances: Sequence[str] = ("configural", "metric", "scalar"),
    seed: int = 42,
    chain_method: str = "sequential",
    num_warmup: int = 2000,
    num_samples: int = 4000,
    num_chains: int = 4,
    target_accept_prob: float = 0.95,
    max_tree_depth: int = 12,
    attach_celllevel_ll: bool = False,
    attach_personlevel_ll: bool = False,
    rhat_gate: float = 1.05,
    ess_gate: float = 100.0,
) -> Dict[str, Dict[str, Any]]:
    y_np = np.asarray(long_data.y)
    if y_np.min() < 0:
        raise ValueError("Negative codes in y: use obs_mask for missing, do not code missing as -1.")
    if y_np.max() >= n_cats:
        raise ValueError(f"y has values >= n_cats ({n_cats}). max={y_np.max()}")

    results: Dict[str, Dict[str, Any]] = {}
    n_rows, n_items = map(int, long_data.y.shape)

    for i, inv in enumerate(invariances):
        model = BayesianOrdinalTraitOccasionCFA(
            n_factors=n_factors,
            n_cats=n_cats,
            n_times=n_times,
            n_person=n_person,
            item_to_factor_indices=item_to_factor_indices,
            marker_items=marker_items,
            invariance=inv,
            free_loading_pairs=free_loading_pairs,
            free_threshold_items=free_threshold_items,
        )

        init_vals: Dict[str, Any] = {
            "L_T": jnp.eye(n_factors),
            "L_O": jnp.eye(n_factors),
            "sig_O": jnp.full((n_factors,), 0.3),
            "z_trait": jnp.zeros((n_person, n_factors)),
            "z_occasion": jnp.zeros((n_rows, n_factors)),
            "lam_pairs_nonmarker": jnp.full((len(model._nonmarker_pos),), model.cfa_loading_loc),
        }
        if n_times > 1:
            init_vals["alpha_dev"] = jnp.zeros((n_times - 1, n_factors))
        if n_cats > 2 and model.threshold_location == "free_t0":
            init_vals["t0"] = jnp.zeros((n_items,))
        if n_cats > 2:
            init_vals["raw_delta"] = jnp.zeros((n_items, n_cats - 2))

        n_dev = int(model._dev_pos.shape[0])
        if n_times > 1 and n_dev > 0:
            init_vals["z_lam_dev"] = jnp.zeros((n_times - 1, n_dev))

        thr_loc_is_sampled = (inv in ("configural", "metric")) or (inv == "scalar" and (free_threshold_items or []))
        if n_times > 1 and thr_loc_is_sampled:
            n_free_thr = n_items if inv in ("configural", "metric") else len(free_threshold_items or [])
            init_vals["thr_loc_dev"] = jnp.zeros((n_times - 1, int(n_free_thr)))

        mcmc, nuts, idata, meta = run_mcmc_with_meta(
            model_obj=model,
            long_data=long_data,
            seed=seed + 10 * i,
            num_warmup=num_warmup,
            num_samples=num_samples,
            num_chains=num_chains,
            chain_method=chain_method,
            target_accept_prob=target_accept_prob,
            max_tree_depth=max_tree_depth,
            init_values=init_vals,
            rhat_gate=rhat_gate,
            ess_gate=ess_gate,
            print_diagnostics=True,
        )

        samples_flat = mcmc.get_samples(group_by_chain=False)
        ll_cells = masked_loglik_cells(nuts.model, samples_flat, long_data)

        num_chains_eff = int(mcmc.num_chains)
        num_samples_eff = int(mcmc.num_samples)

        idata = attach_loglik_rowlevel(
            idata,
            ll_cells,
            num_chains=num_chains_eff,
            num_samples=num_samples_eff,
            var_name="obs_row",
        )

        cell_obs_indices = None
        if attach_celllevel_ll:
            idata, cell_obs_indices = attach_loglik_celllevel_observed_only(
                idata,
                ll_cells,
                long_data,
                num_chains=num_chains_eff,
                num_samples=num_samples_eff,
                var_name="obs_cell",
            )

        if attach_personlevel_ll:
            idata = attach_loglik_personlevel(
                idata,
                ll_cells,
                long_data,
                n_person=n_person,              # <-- explicit
                num_chains=num_chains_eff,
                num_samples=num_samples_eff,
                var_name="obs_person",
            )

        results[inv] = {
            "model": model,
            "mcmc": mcmc,
            "nuts": nuts,
            "idata": idata,
            "meta": meta,
            "ll_cells": ll_cells,
            "cell_obs_indices": cell_obs_indices,
        }

    return results


# ============================================================
# USAGE EXAMPLE
# ============================================================
# results = fit_invariance_series(..., attach_celllevel_ll=True)
# diag_row = loo_diagnostics_cached(results, var_name="obs_row")
# cmp_row  = compare_by_cached_loo_with_se(results, var_name="obs_row", diag=diag_row)
# ============================================================
```
