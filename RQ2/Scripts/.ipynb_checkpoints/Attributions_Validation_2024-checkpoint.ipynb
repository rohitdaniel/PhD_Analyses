```python
logger = logging.getLogger(__name__)

def read_csv_file(path: Path) -> pd.DataFrame:
    try:
        df = pd.read_csv(path, encoding="utf-8")
        logger.info(f"Successfully loaded data from {path}")
        return df
    except FileNotFoundError:
        logger.error(f"File not found: {path}")
        raise
    except pd.errors.ParserError as e:
        logger.error(f"Parsing error in file {path}: {e}")
        raise
    except Exception as e:
        logger.error(f"Unexpected error reading {path}: {e}")
        raise

df_preobs_2024 = read_csv_file(PREOBS_2024_FILE_PATH)
df_preobs_2025 = read_csv_file(PREOBS_2025_FILE_PATH)
df_pilot = read_csv_file(PILOT_FILE_PATH)
```

### Prepare datasets for analyses

```python
# Define latent attributions mapping to item columns (success/failure valence)
latent_map = {
    "eff_1": ("amp_1", "amn_1"),
    "eff_2": ("amp_7", "amn_7"),
    "eff_3": ("amp_15", "amn_15"),
    "ab_1": ("amp_2", "amn_2"),
    "ab_2": ("amp_5", "amn_5"),
    "ab_3": ("amp_9", "amn_9"),
    "luck_1": ("amp_3", "amn_3"),
    "luck_2": ("amp_11", "amn_11"),
    "luck_3": ("amp_14", "amn_14"),
    "task_1": ("amp_4", "amn_4"),
    "task_2": ("amp_8", "amn_8"),
    "task_3": ("amp_12", "amn_12"),
    "int_1": ("amp_6", "amn_6"),
    "int_2": ("amp_10", "amn_10"),
    "int_3": ("amp_13", "amn_13"),
}

# Coalesce success ('amp_*') and failure ('amn_*') columns into unified attribution columns
for new_col, (amp_col, amn_col) in latent_map.items():
    df_preobs_2024[new_col] = df_preobs_2024[amp_col].combine_first(df_preobs_2024[amn_col])
    df_preobs_2025[new_col] = df_preobs_2025[amp_col].combine_first(df_preobs_2025[amn_col])
    df_pilot[new_col] = df_pilot[amp_col].combine_first(df_pilot[amn_col])
    # Drop original columns to avoid redundancy
    df_preobs_2024.drop(columns=[amp_col, amn_col], inplace=True)
    df_preobs_2025.drop(columns=[amp_col, amn_col], inplace=True)
    df_pilot.drop(columns=[amp_col, amn_col], inplace=True)

# Add valence category based on scoring in preobs dataset
def assign_valence_preobs(score):
    if score in [3, 4]:
        return "success"
    elif score in [1, 2]:
        return "failure"
    else:
        return None

df_preobs_2024["valence"] = df_preobs_2024["sc"].apply(assign_valence_preobs).astype("category")
df_preobs_2025["valence"] = df_preobs_2025["sc"].apply(assign_valence_preobs).astype("category")

# Add valence category based on scoring in pilot dataset
def assign_valence_pilot(score):
    if score in ["Ganske godt", "Svært godt"]:
        return "success"
    elif score in ["Ganske dårleg", "Svært dårleg"]:
        return "failure"
    else:
        return None

df_pilot["valence"] = df_pilot["sc"].apply(assign_valence_pilot).astype("category")

# Drop rows with missing values in any of the attribution columns
df_preobs_2024.dropna(subset=list(latent_map.keys()), inplace=True)
df_preobs_2025.dropna(subset=list(latent_map.keys()), inplace=True)
df_pilot.dropna(subset=list(latent_map.keys()), inplace=True)

# Filter out patterned responses based on flags in both datasets
df_preobs_2024 = df_preobs_2024[~((df_preobs_2024["attr_patterned"] == 1) & (df_preobs_2024["total_patterned"] > 2))]
df_preobs_2025 = df_preobs_2025[~((df_preobs_2025["attr_patterned"] == 1) & (df_preobs_2025["total_patterned"] > 2))]
df_pilot = df_pilot[~((df_pilot["attr_patterned"] == 1) | (df_pilot["time_flag"] == 1))]

# Reorder columns: keep identifiers and valence first, then latent attribution columns
columns_order = ["id_unique", "valence"] + list(latent_map.keys())
preobs_2024_attr_data = df_preobs_2024[columns_order].copy()
preobs_2025_attr_data = df_preobs_2025[columns_order].copy()
pilot_attr_data = df_pilot[columns_order].copy()
```

```python
# Extract attribution item columns matching pattern *_[1-3]
item_cols = preobs_2024_attr_data.filter(regex=r'_([1-3])$', axis=1).columns.tolist()

# Convert attribution columns to nullable integer type for both datasets
preobs_2024_attr_data.loc[:, item_cols] = preobs_2024_attr_data.loc[:, item_cols].astype('Int64')
preobs_2025_attr_data.loc[:, item_cols] = preobs_2025_attr_data.loc[:, item_cols].astype('Int64')
pilot_attr_data.loc[:, item_cols] = pilot_attr_data.loc[:, item_cols].astype('Int64')

# Separate success and failure subsets from preobs dataset
preobs_2024_success = preobs_2024_attr_data.loc[preobs_2024_attr_data['valence'] == 'success', item_cols]
preobs_2024_failure = preobs_2024_attr_data.loc[preobs_2024_attr_data['valence'] == 'failure', item_cols]
preobs_2025_success = preobs_2025_attr_data.loc[preobs_2025_attr_data['valence'] == 'success', item_cols]
preobs_2025_failure = preobs_2025_attr_data.loc[preobs_2025_attr_data['valence'] == 'failure', item_cols]

# Separate failure subset from pilot dataset
pilot_success = pilot_attr_data.loc[pilot_attr_data['valence'] == 'success', item_cols]
pilot_failure = pilot_attr_data.loc[pilot_attr_data['valence'] == 'failure', item_cols]

# Define ordered categorical dtype with categories 1 through 6
ordered_cat_type = pd.CategoricalDtype(categories=list(range(1, 7)), ordered=True)

# Convert attribution columns in main datasets to ordered categorical
preobs_2024_attr_data.loc[:, item_cols] = preobs_2024_attr_data.loc[:, item_cols].astype(ordered_cat_type)
preobs_2025_attr_data.loc[:, item_cols] = preobs_2025_attr_data.loc[:, item_cols].astype(ordered_cat_type)
pilot_attr_data.loc[:, item_cols] = pilot_attr_data.loc[:, item_cols].astype(ordered_cat_type)

# Convert separated success/failure subsets to ordered categorical dtype
preobs_2024_success = preobs_2024_success.astype(ordered_cat_type)
preobs_2024_failure = preobs_2024_failure.astype(ordered_cat_type)
preobs_2025_success = preobs_2025_success.astype(ordered_cat_type)
preobs_2025_failure = preobs_2025_failure.astype(ordered_cat_type)
pilot_success = pilot_success.astype(ordered_cat_type)
pilot_failure = pilot_failure.astype(ordered_cat_type)
```

```python
# Convert preobs attribution data to JAX arrays (zero-indexed)
preobs_2024_data_array = jnp.asarray(preobs_2024_attr_data[item_cols].to_numpy(), dtype=jnp.int32) - 1
preobs_2025_data_array = jnp.asarray(preobs_2025_attr_data[item_cols].to_numpy(), dtype=jnp.int32) - 1

# Convert preobs success/failure subsets to JAX arrays (zero-indexed)
preobs_2024_success_array = jnp.asarray(preobs_2024_success.to_numpy(), dtype=jnp.int32) - 1
preobs_2024_failure_array = jnp.asarray(preobs_2024_failure.to_numpy(), dtype=jnp.int32) - 1
preobs_2025_success_array = jnp.asarray(preobs_2025_success.to_numpy(), dtype=jnp.int32) - 1
preobs_2025_failure_array = jnp.asarray(preobs_2025_failure.to_numpy(), dtype=jnp.int32) - 1

# Convert pilot attribution data to JAX arrays (zero-indexed)
pilot_data_array = jnp.asarray(pilot_attr_data[item_cols].to_numpy(), dtype=jnp.int32) - 1

# Convert pilot success/failure subsets to JAX arrays (zero-indexed)
pilot_success_array = jnp.asarray(pilot_success.to_numpy(), dtype=jnp.int32) - 1
pilot_failure_array = jnp.asarray(pilot_failure.to_numpy(), dtype=jnp.int32) - 1
```

```python
# Initialize LabelEncoder to convert text labels into numeric codes
le = LabelEncoder()

# Encode 'valence' in preobs dataset and store as int32
preobs_2024_valence = le.fit_transform(preobs_2024_attr_data['valence']).astype(np.int32)
preobs_2024_attr_data.loc[:, 'valence_encoded'] = preobs_2024_valence
preobs_2025_valence = le.fit_transform(preobs_2025_attr_data['valence']).astype(np.int32)
preobs_2025_attr_data.loc[:, 'valence_encoded'] = preobs_2025_valence

# Encode 'valence' in pilot dataset using the same encoder to keep labels consistent
pilot_valence = le.transform(pilot_attr_data['valence']).astype(np.int32)
pilot_attr_data.loc[:, 'valence_encoded'] = pilot_valence

def create_group_indices(group_labels: np.ndarray) -> list[np.ndarray]:
    """Create list of index arrays, one per unique group label."""
    unique_groups = np.unique(group_labels)
    return [np.where(group_labels == g)[0] for g in unique_groups]

# Create index lists for preobs and pilot valence groups
preobs_2024_valence_indices = create_group_indices(preobs_2024_valence)
preobs_2025_valence_indices = create_group_indices(preobs_2025_valence)
pilot_valence_indices = create_group_indices(pilot_valence)

# Print mapping of encoded groups to original labels
print("Group labels:", le.classes_)
```

### Descriptive statistics: Preobs 2024

```python
# Categories (e.g., 1 to 6)
categories = range(1, 7)

# Set seaborn style
sns.set(style="whitegrid")

# 1. Frequency distribution barplots per item
for col in item_cols:
    counts = preobs_2024_attr_data[col].value_counts().reindex(categories, fill_value=0)
    plt.figure(figsize=(8, 4))
    sns.barplot(x=list(categories), y=counts.values, palette="muted")
    plt.title(f"Frequency Distribution: {col}")
    plt.xlabel("Category")
    plt.ylabel("Count")
    plt.xticks(ticks=list(categories), labels=list(categories))
    plt.show()

# 2. Stacked bar plot of category proportions across all items
prop_df = pd.DataFrame({
    col: preobs_2024_attr_data[col].value_counts(normalize=True).reindex(categories, fill_value=0)
    for col in item_cols
}).T

prop_df.plot(kind='bar', stacked=True, colormap='Spectral', figsize=(12,6))
plt.title("Stacked Bar Plot of Category Proportions by Item")
plt.xlabel("Items")
plt.ylabel("Proportion")
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 3. Cumulative distribution function (CDF) plots per item
for col in item_cols:
    data = preobs_2024_attr_data[col].dropna().astype(int)
    cdf = data.value_counts(normalize=True).sort_index().cumsum()
    plt.figure(figsize=(8, 4))
    sns.lineplot(x=cdf.index, y=cdf.values, marker="o")
    plt.title(f"Cumulative Distribution (CDF) of {col}")
    plt.xlabel("Category")
    plt.ylabel("Cumulative Proportion")
    plt.xticks(ticks=list(categories), labels=list(categories))
    plt.ylim(0, 1)
    plt.grid(True)
    plt.show()

# 4. Median and mode per item (descriptive summary)
summary_stats = []
for col in item_cols:
    data = preobs_2024_attr_data[col].dropna().astype(int)
    median_val = data.median()
    mode_val = data.mode().iloc[0] if not data.mode().empty else np.nan
    summary_stats.append({
        "Item": col,
        "Median": median_val,
        "Mode": mode_val,
        "Count": len(data)
    })

summary_df = pd.DataFrame(summary_stats)
print(summary_df)

# 5. Nonparametric group comparisons (example: valence groups)
import scipy.stats as stats

# Assuming 'valence' is categorical with two groups: 'success' and 'failure'
group1 = preobs_2024_attr_data[preobs_2024_attr_data['valence'] == 'success']
group2 = preobs_2024_attr_data[preobs_2024_attr_data['valence'] == 'failure']

group_comp_results = []
for col in item_cols:
    data1 = group1[col].dropna().astype(int)
    data2 = group2[col].dropna().astype(int)
    # Mann-Whitney U test (nonparametric test for independent samples)
    if len(data1) > 0 and len(data2) > 0:
        stat, p_val = stats.mannwhitneyu(data1, data2, alternative='two-sided')
    else:
        stat, p_val = np.nan, np.nan
    group_comp_results.append({
        "Item": col,
        "Mann-Whitney U stat": stat,
        "p-value": p_val
    })

group_comp_df = pd.DataFrame(group_comp_results)
print(group_comp_df)

# Optional: adjust p-values for multiple testing (e.g., Benjamini-Hochberg FDR)
from statsmodels.stats.multitest import multipletests
pvals = group_comp_df["p-value"].values
_, pvals_adj, _, _ = multipletests(pvals, method='fdr_bh')
group_comp_df["p-value_adj"] = pvals_adj

print(group_comp_df)
```

# Bayesian Ordinal EFA/CFA (NumPyro/JAX)

```python
import os
from pathlib import Path
from typing import Optional, Sequence, Tuple, List

import numpy as np
import jax
import jax.numpy as jnp
from jax.nn import softplus
from jax.scipy.special import log_ndtr

import numpyro
import numpyro.distributions as dist
from numpyro.distributions import Distribution, constraints
from numpyro.infer import MCMC, NUTS
from numpyro.infer.initialization import init_to_value
from numpyro.infer.util import log_likelihood

import arviz as az

import json

# ----------------------------
# Numerics helpers
# ----------------------------
def _log1mexp(x: jnp.ndarray) -> jnp.ndarray:
    log_half = jnp.asarray(-0.6931471805599453)
    return jnp.where(x < log_half, jnp.log1p(-jnp.exp(x)), jnp.log(-jnp.expm1(x)))


def _sample_truncnorm_pos(name: str, loc: float, scale: float, shape: int | tuple[int, ...]):
    shape_tup = (shape,) if isinstance(shape, int) else tuple(shape)
    try:
        d = dist.TruncatedNormal(loc=loc, scale=scale, low=0.0)
    except TypeError:
        try:
            d = dist.TruncatedNormal(loc, scale, 0.0, jnp.inf)
        except TypeError as e:
            raise TypeError(
                "NumPyro TruncatedNormal API mismatch. Pin/upgrade numpyro or adjust the call."
            ) from e
    return numpyro.sample(name, d.expand(shape_tup))


# ----------------------------
# Ordered Probit distribution
# ----------------------------
class OrderedProbit(Distribution):
    arg_constraints = {"mu": constraints.real, "thresholds": constraints.ordered_vector}

    def __init__(self, mu, thresholds, validate_args=None):
        self.mu = jnp.asarray(mu)
        thr = jnp.asarray(thresholds)

        if thr.ndim < 1:
            raise ValueError("thresholds must have at least 1 dimension (last dim = n_thr).")

        self.n_thr = int(thr.shape[-1])
        self.n_cats = self.n_thr + 1

        if thr.ndim == 1:
            thr = thr.reshape((1,) * self.mu.ndim + (self.n_thr,))
        elif thr.ndim == 2:
            if self.mu.ndim < 1:
                raise ValueError("mu must have at least 1 dimension.")
            n_items = self.mu.shape[-1]
            if thr.shape[0] != n_items:
                raise ValueError(f"thresholds shape mismatch, got {thr.shape}, expected [{n_items},{self.n_thr}]")
            thr = thr.reshape((1,) * (self.mu.ndim - 1) + (n_items, self.n_thr))

        self.thresholds = jnp.broadcast_to(thr, self.mu.shape + (self.n_thr,))
        super().__init__(batch_shape=self.mu.shape, event_shape=(), validate_args=validate_args)

    @property
    def support(self):
        return constraints.integer_interval(0, self.n_cats - 1)

    def log_prob(self, value):
        y = jnp.asarray(value).astype(jnp.int32)
        in_range = (y >= 0) & (y < self.n_cats)

        y_idx = jnp.clip(y, 0, self.n_cats - 1)
        y_exp = y_idx[..., None]

        t = self.thresholds - self.mu[..., None]

        neg_inf = jnp.full(self.mu.shape + (1,), -jnp.inf)
        pos_inf = jnp.full(self.mu.shape + (1,),  jnp.inf)
        t_full = jnp.concatenate([neg_inf, t, pos_inf], axis=-1)

        lower = jnp.take_along_axis(t_full[..., :-1], y_exp, axis=-1)[..., 0]
        upper = jnp.take_along_axis(t_full[...,  1:], y_exp, axis=-1)[..., 0]

        ordered = upper > lower

        log_cdf_u = log_ndtr(upper)
        log_cdf_l = log_ndtr(lower)

        d = jnp.minimum(log_cdf_l - log_cdf_u, 0.0)

        eps = jnp.asarray(1e-12)
        d_safe = jnp.where(d > -eps, -eps, d)

        log_p_ordered = log_cdf_u + _log1mexp(d_safe)

        log_p = jnp.where(ordered, log_p_ordered, -jnp.inf)
        log_p = jnp.where(in_range, log_p, -jnp.inf)
        return log_p

    def sample(self, key, sample_shape=()):
        shape = sample_shape + self.batch_shape
        z = self.mu + jax.random.normal(key, shape)
        y = jnp.sum(z[..., None] > self.thresholds, axis=-1)
        return y.astype(jnp.int32)


# ----------------------------
# Bayesian ordinal EFA/CFA
# ----------------------------
class BayesianOrdinalFactorModel:
    def __init__(
        self,
        n_factors: int,
        n_cats: int,
        mode: str = "efa",
        item_to_factor_indices: Optional[Sequence[Sequence[int]]] = None,
        identification: str = "unit_variance",

        loadings_scale: float = 0.5,
        cfa_loading_loc: float = 0.7,

        factor_scale_sd: float = 0.5,
        nu_sd: float = 1.0,

        delta_sd: float = 0.3,
        min_delta: float = 1e-3,

        lkj_eta: float = 2.0,
        efa_oblique: bool = True,

        allow_negative_cfa_loadings: bool = False,

        validate_cfa_coverage: bool = True,
        allow_cross_loadings: bool = False,

        efa_diag_jitter: float = 1e-4,
        efa_diag_loc: float = 1.0,
        efa_diag_sd: float = 0.5,
    ):
        self.n_factors = int(n_factors)
        self.n_cats = int(n_cats)
        self.mode = mode.lower()
        self.item_to_factor_indices = item_to_factor_indices
        self.identification = identification.lower()

        self.loadings_scale = float(loadings_scale)
        self.cfa_loading_loc = float(cfa_loading_loc)
        self.factor_scale_sd = float(factor_scale_sd)
        self.nu_sd = float(nu_sd)

        self.delta_sd = float(delta_sd)
        self.min_delta = float(min_delta)

        self.lkj_eta = float(lkj_eta)
        self.efa_oblique = bool(efa_oblique)

        self.allow_negative_cfa_loadings = bool(allow_negative_cfa_loadings)
        self.validate_cfa_coverage = bool(validate_cfa_coverage)
        self.allow_cross_loadings = bool(allow_cross_loadings)

        self.efa_diag_jitter = float(efa_diag_jitter)
        self.efa_diag_loc = float(efa_diag_loc)
        self.efa_diag_sd = float(efa_diag_sd)

        if self.mode not in ("efa", "cfa"):
            raise ValueError("mode must be 'efa' or 'cfa'")
        if self.identification not in ("unit_variance", "marker"):
            raise ValueError("identification must be 'unit_variance' or 'marker'")
        if self.mode == "cfa" and self.item_to_factor_indices is None:
            raise ValueError("CFA mode requires item_to_factor_indices")
        if self.mode == "efa" and self.identification == "marker":
            raise ValueError("marker identification supported only in CFA.")
        if self.n_cats < 2:
            raise ValueError("n_cats must be >= 2.")

        # EFA lower-triangular indices
        r_lt, c_lt = np.tril_indices(self.n_factors, k=-1)
        self._efa_r_lt = jnp.asarray(r_lt, dtype=jnp.int32)
        self._efa_c_lt = jnp.asarray(c_lt, dtype=jnp.int32)
        self._efa_n_free_lt = int(r_lt.shape[0])

        # CFA prepared structure
        self._cfa_prepared = False
        self._cfa_item_ix = None
        self._cfa_fact_ix = None
        self._cfa_marker_pos = None
        self._cfa_nonmarker_pos = None
        self._cfa_n_free = None
        self._cfa_n_nonmarker = None

    def _record_factor_moments(self, R: jnp.ndarray, scales: Optional[jnp.ndarray]):
        numpyro.deterministic("factor_corr", R)
        if scales is None:
            numpyro.deterministic("factor_cov", R)
        else:
            S = jnp.diag(scales)
            numpyro.deterministic("factor_cov", S @ R @ S)

    def _sample_eta(self, n_obs: int, k: int):
        if self.mode == "efa" and (not self.efa_oblique):
            R = jnp.eye(k)
            self._record_factor_moments(R, scales=None)
            return numpyro.sample("eta", dist.Normal(0.0, 1.0).expand([n_obs, k]))

        L_omega = numpyro.sample("L_omega", dist.LKJCholesky(k, self.lkj_eta))
        R = L_omega @ L_omega.T
        z = numpyro.sample("z_eta", dist.Normal(0.0, 1.0).expand([n_obs, k]))

        if self.mode == "efa" or self.identification == "unit_variance":
            self._record_factor_moments(R, scales=None)
            return numpyro.deterministic("eta", z @ L_omega.T)

        scales = numpyro.sample("factor_scales", dist.HalfNormal(self.factor_scale_sd).expand([k]))
        self._record_factor_moments(R, scales=scales)
        return numpyro.deterministic("eta", (z * scales[None, :]) @ L_omega.T)

    def _validate_cfa_indices(self, n_items: int) -> List[List[int]]:
        lists = self.item_to_factor_indices
        cleaned: List[List[int]] = []

        for f, idx_list in enumerate(lists):
            idx_py = [int(i) for i in idx_list]
            if len(idx_py) == 0:
                raise ValueError(f"CFA: factor {f} has no items assigned.")
            bad = [i for i in idx_py if (i < 0 or i >= n_items)]
            if bad:
                raise ValueError(f"CFA: factor {f} out-of-range indices: {sorted(set(bad))}")
            if len(set(idx_py)) != len(idx_py):
                raise ValueError(f"CFA: factor {f} has duplicate indices.")
            cleaned.append(idx_py)

        if self.validate_cfa_coverage:
            all_assigned = set(i for idx_py in cleaned for i in idx_py)
            missing = set(range(n_items)) - all_assigned
            if missing:
                raise ValueError(f"CFA: items not assigned to any factor: {sorted(missing)}")

        if not self.allow_cross_loadings:
            counts = np.zeros(n_items, dtype=int)
            for idx_py in cleaned:
                for i in idx_py:
                    counts[i] += 1
            overlaps = np.where(counts > 1)[0].tolist()
            if overlaps:
                raise ValueError(f"CFA: overlaps present but allow_cross_loadings=False: {overlaps}")

        return cleaned

    def prepare_cfa(self, n_items: int) -> List[Tuple[int, int]]:
        if self.mode != "cfa":
            raise ValueError("prepare_cfa is only for CFA mode.")
        cleaned = self._validate_cfa_indices(n_items=n_items)

        pairs = sorted({(int(i), int(f)) for f, idxs in enumerate(cleaned) for i in idxs})
        item_ix = np.asarray([p[0] for p in pairs], dtype=np.int32)
        fact_ix = np.asarray([p[1] for p in pairs], dtype=np.int32)
        n_free = int(len(pairs))

        if self.identification == "marker":
            marker_pairs = {(int(cleaned[f][0]), int(f)) for f in range(len(cleaned))}
            is_marker = np.asarray([(p in marker_pairs) for p in pairs], dtype=bool)
            marker_pos = np.where(is_marker)[0].astype(np.int32)
            nonmarker_pos = np.where(~is_marker)[0].astype(np.int32)
        else:
            marker_pos = np.asarray([], dtype=np.int32)
            nonmarker_pos = np.arange(n_free, dtype=np.int32)

        self._cfa_item_ix = jnp.asarray(item_ix, dtype=jnp.int32)
        self._cfa_fact_ix = jnp.asarray(fact_ix, dtype=jnp.int32)
        self._cfa_marker_pos = jnp.asarray(marker_pos, dtype=jnp.int32)
        self._cfa_nonmarker_pos = jnp.asarray(nonmarker_pos, dtype=jnp.int32)

        self._cfa_n_free = n_free
        self._cfa_n_nonmarker = int(nonmarker_pos.shape[0])
        self._cfa_prepared = True
        return pairs

    def model(self, data):
        y = jnp.asarray(data).astype(jnp.int32)
        n_obs, n_items = y.shape
        k = self.n_factors

        if self.mode == "efa" and n_items < k:
            raise ValueError(f"EFA requires n_items ({n_items}) >= n_factors ({k}).")
        if self.mode == "cfa" and not self._cfa_prepared:
            raise ValueError("CFA requires prepare_cfa(n_items) called on host before MCMC.")

        eta = self._sample_eta(n_obs=n_obs, k=k)

        # Loadings
        if self.mode == "efa":
            lam = jnp.zeros((n_items, k))

            if self._efa_n_free_lt > 0:
                lam_lt = numpyro.sample("lam_lt", dist.Normal(0.0, self.loadings_scale).expand([self._efa_n_free_lt]))
                lam = lam.at[self._efa_r_lt, self._efa_c_lt].set(lam_lt)

            lam_diag_raw = numpyro.sample(
                "lam_diag_raw",
                dist.Normal(self.efa_diag_loc, self.efa_diag_sd).expand([k]),
            )
            lam = lam.at[jnp.arange(k), jnp.arange(k)].set(softplus(lam_diag_raw) + self.efa_diag_jitter)

            if n_items > k:
                lam_rest = numpyro.sample(
                    "lam_rest",
                    dist.Normal(0.0, self.loadings_scale).expand([n_items - k, k]),
                )
                lam = lam.at[jnp.arange(k, n_items), :].set(lam_rest)

        else:
            lam = jnp.zeros((n_items, k))
            n_nonmarker = self._cfa_n_nonmarker

            if n_nonmarker > 0:
                if self.allow_negative_cfa_loadings:
                    vals_nm = numpyro.sample(
                        "lam_pairs_nonmarker",
                        dist.Normal(self.cfa_loading_loc, self.loadings_scale).expand([n_nonmarker]),
                    )
                else:
                    vals_nm = _sample_truncnorm_pos(
                        "lam_pairs_nonmarker",
                        loc=self.cfa_loading_loc,
                        scale=self.loadings_scale,
                        shape=n_nonmarker,
                    )
            else:
                vals_nm = jnp.zeros((0,))

            vals_full = jnp.zeros((self._cfa_n_free,))
            if self.identification == "marker" and self._cfa_marker_pos.size > 0:
                vals_full = vals_full.at[self._cfa_marker_pos].set(1.0)
            vals_full = vals_full.at[self._cfa_nonmarker_pos].set(vals_nm)

            lam = lam.at[self._cfa_item_ix, self._cfa_fact_ix].set(vals_full)

        numpyro.deterministic("loadings", lam)

        nu = numpyro.sample("nu", dist.Normal(0.0, self.nu_sd).expand([n_items]))
        numpyro.deterministic("item_intercepts", nu)

        mu = nu[None, :] + eta @ lam.T

        # Thresholds: first fixed 0, monotone deltas, no centering
        n_thr = self.n_cats - 1
        if self.n_cats == 2:
            thresholds = jnp.zeros((n_items, 1))
        else:
            raw_delta = numpyro.sample("raw_delta", dist.Normal(0.0, self.delta_sd).expand([n_items, n_thr - 1]))
            delta = softplus(raw_delta) + self.min_delta
            thresholds = jnp.concatenate([jnp.zeros((n_items, 1)), jnp.cumsum(delta, axis=1)], axis=1)

        numpyro.deterministic("thresholds", thresholds)

        numpyro.sample("obs", OrderedProbit(mu, thresholds, validate_args=False), obs=y)

# ----------------------------
# Diagnostic helpers
# ----------------------------
def max_rhat_all(idata: az.InferenceData) -> float:
    """
    Robust: compute R-hat for all posterior vars and return global max.
    """
    rhat_ds = az.rhat(idata)  # xarray Dataset
    max_vals = []
    for v in rhat_ds.data_vars.values():
        arr = v.to_numpy()
        if arr.size:
            max_vals.append(float(np.nanmax(arr)))
    return float(np.nanmax(max_vals)) if max_vals else float("nan")


def min_ess_bulk_all(idata: az.InferenceData) -> float:
    ess_ds = az.ess(idata, method="bulk")
    min_vals = []
    for v in ess_ds.data_vars.values():
        arr = v.to_numpy()
        if arr.size:
            min_vals.append(float(np.nanmin(arr)))
    return float(np.nanmin(min_vals)) if min_vals else float("nan")
```

## EFA: Preobs 2024

```python
from pathlib import Path
import numpy as np
import arviz as az
import jax
import jax.numpy as jnp
from numpyro.infer import NUTS, MCMC, init_to_value, log_likelihood

mcmc_results_success, idata_results_success = {}, {}
mcmc_results_failure, idata_results_failure = {}, {}

valence_data = {
    "success": preobs_2024_success_array,
    "failure": preobs_2024_failure_array,
}

factor_range = range(2, 7)
n_cats = 6
chain_method = "sequential"

base_save_dir = Path("/mnt/c/Users/rodan/OneDrive - Høgskulen på Vestlandet/Dokumenter/PhD/RQ2/results/efa_results/mcmc_idata")

for valence_name, data_array in valence_data.items():
    print(f"\nRunning UPDATED EFA models for {valence_name} valence...")

    Y_np = np.asarray(data_array, dtype=np.int32)

    if Y_np.min() < 0:
        raise ValueError("Negative category codes found (e.g., missing coded as -1). Handle missing first.")
    if Y_np.max() >= n_cats:
        raise ValueError(f"Data exceeds n_cats ({n_cats}). Max found: {Y_np.max()}")

    Y = jnp.asarray(Y_np)
    n_obs, n_items = Y_np.shape

    mcmc_results = mcmc_results_success if valence_name == "success" else mcmc_results_failure
    idata_results = idata_results_success if valence_name == "success" else idata_results_failure

    save_dir = base_save_dir / valence_name / "updated_v3"
    save_dir.mkdir(parents=True, exist_ok=True)

    for k in factor_range:
        print(f"\n--- {valence_name}: k={k} ---")

        if n_items < k:
            raise ValueError(f"EFA requires n_items ({n_items}) >= n_factors ({k}).")

        efa_oblique = True
        n_free_lt = k * (k - 1) // 2

        efa_diag_loc = 1.0
        efa_diag_sd = 0.5

        init_values = {
            "lam_lt":       jnp.zeros((n_free_lt,)),
            "lam_diag_raw": jnp.full((k,), efa_diag_loc),
            "nu":           jnp.zeros((n_items,)),
        }
        if efa_oblique:
            init_values["z_eta"] = jnp.zeros((n_obs, k))
            init_values["L_omega"] = jnp.eye(k)
        else:
            init_values["eta"] = jnp.zeros((n_obs, k))
        if n_items > k:
            init_values["lam_rest"] = jnp.zeros((n_items - k, k))
        if n_cats > 2:
            init_values["raw_delta"] = jnp.zeros((n_items, n_cats - 2))

        init_strategy = init_to_value(values=init_values)

        model = BayesianOrdinalFactorModel(
            n_factors=k,
            n_cats=n_cats,
            mode="efa",
            efa_oblique=efa_oblique,
            lkj_eta=2.0,
            efa_diag_loc=efa_diag_loc,
            efa_diag_sd=efa_diag_sd,
            efa_diag_jitter=1e-4,
        )

        max_tree_depth = 14
        target_accept = 0.99

        nuts = NUTS(
            model.model,
            init_strategy=init_strategy,
            target_accept_prob=target_accept,
            max_tree_depth=max_tree_depth,
        )

        mcmc = MCMC(
            nuts,
            num_warmup=2000,
            num_samples=4000,
            num_chains=4,
            chain_method=chain_method,
            progress_bar=True,
        )

        seed = 42 + (0 if valence_name == "success" else 10_000) + k * 100

        # ---- Run ----
        if chain_method == "sequential":
            mcmc.run(jax.random.PRNGKey(seed), data=Y, extra_fields=("diverging", "num_steps"))
        else:
            rng_keys = jax.random.split(jax.random.PRNGKey(seed), mcmc.num_chains)
            mcmc.run(rng_keys, data=Y, extra_fields=("diverging", "num_steps"))

        mcmc_results[k] = mcmc

        # ---- Minimal sampler diagnostics ----
        extra = mcmc.get_extra_fields()
        div = np.asarray(extra.get("diverging", 0))
        num_steps = np.asarray(extra.get("num_steps", np.nan))

        n_div = int(np.sum(div)) if np.ndim(div) else int(div)
        mean_steps = float(np.nanmean(num_steps))
        p95_steps = float(np.nanquantile(num_steps, 0.95))
        max_steps = int(np.nanmax(num_steps))

        max_allowed_steps = 2 ** max_tree_depth
        near_max = int(np.sum(num_steps >= (max_allowed_steps - 1)))

        print(
            f"Sampler: divergences={n_div}, num_steps mean={mean_steps:.1f}, p95={p95_steps:.1f}, "
            f"max={max_steps}, near_max_tree_depth={near_max}"
        )

        # ---- Attach pointwise log_likelihood ----
        try:
            posterior_samples = mcmc.get_samples(group_by_chain=False)
            ll_dict = log_likelihood(model.model, posterior_samples, data=Y)
            obs_key = "obs" if "obs" in ll_dict else next(iter(ll_dict.keys()))
            ll = ll_dict[obs_key]
            idata = az.from_numpyro(mcmc, log_likelihood={obs_key: ll})
        except Exception as e:
            print(f"WARNING: log_likelihood failed for {valence_name}, k={k}: {repr(e)}")
            idata = az.from_numpyro(mcmc, log_likelihood=False)

        idata_results[k] = idata
        az.to_netcdf(idata, save_dir / f"{valence_name}_{k}.nc")

print("\nDone.")
```

# k-selection + diagnostics for Bayesian (ordinal) EFA

```python
from __future__ import annotations

from pathlib import Path
import re
import json
import numpy as np
import pandas as pd
import arviz as az
import matplotlib.pyplot as plt

# ----------------------------
# 0) CONFIG
# ----------------------------
BASE_DIR = Path(
    "/mnt/c/Users/rodan/OneDrive - Høgskulen på Vestlandet/Dokumenter/PhD/RQ2/results/efa_results/mcmc_idata"
)
VALENCES = ("success", "failure")

# If files are: success_2.nc ... success_6.nc inside:
#   BASE_DIR / "success" / "updated_v3" / "success_2.nc"
SUBFOLDER = "updated_v3"

# If you know your k range, set it; otherwise loader will parse filenames
FACTOR_RANGE = range(2, 7)

# Convergence gate thresholds (tune to your tolerance)
RHAT_CUT = 1.01
ESS_MIN_CUT = 200  # adjust depending on your draws and complexity

# Variables to check for convergence (edit to match your model variable names)
VARS_PREFIX = ("loadings", "thresholds", "factor_corr", "factor_cov", "psi", "psi_raw")


# ----------------------------
# 1) LOADING
# ----------------------------
def load_idata(valence: str, k: int, base_dir: Path = BASE_DIR, subfolder: str = SUBFOLDER):
    path = base_dir / valence / subfolder / f"{valence}_{k}.nc"
    if path.exists():
        return az.from_netcdf(path)
    return None


def load_all_idatas(
    base_dir: Path = BASE_DIR,
    valences: tuple[str, ...] = VALENCES,
    subfolder: str = SUBFOLDER,
    factor_range=FACTOR_RANGE,
    parse_from_filenames: bool = True,
):
    """
    Returns dict: {valence: {k: idata}}
    If parse_from_filenames=True, it scans *.nc files and extracts k from filenames.
    Otherwise it loads using factor_range.
    """
    out: dict[str, dict[int, az.InferenceData]] = {}

    for val in valences:
        folder = base_dir / val / subfolder
        idata_by_k: dict[int, az.InferenceData] = {}

        if parse_from_filenames and folder.exists():
            # Match e.g. success_2.nc / failure_6.nc
            pat = re.compile(rf"^{re.escape(val)}_(\d+)\.nc$")
            for f in folder.glob(f"{val}_*.nc"):
                m = pat.match(f.name)
                if m:
                    k = int(m.group(1))
                    idata_by_k[k] = az.from_netcdf(f)
        else:
            for k in factor_range:
                idata = load_idata(val, k, base_dir=base_dir, subfolder=subfolder)
                if idata is not None:
                    idata_by_k[k] = idata

        out[val] = dict(sorted(idata_by_k.items(), key=lambda kv: kv[0]))

    return out


# ----------------------------
# 2) BASIC CHECKS / DIAGNOSTICS
# ----------------------------
def has_pointwise_loglik(idata: az.InferenceData) -> bool:
    return ("log_likelihood" in idata.groups()) and (len(idata.log_likelihood.data_vars) > 0)


def select_var_names(idata: az.InferenceData, prefixes=VARS_PREFIX):
    """
    Choose a focused set of posterior variables for convergence checks.
    Falls back to all posterior vars if none match.
    """
    posterior_vars = list(idata.posterior.data_vars)
    vars_sel = [v for v in posterior_vars if any(v.startswith(p) for p in prefixes)]
    return vars_sel if vars_sel else posterior_vars


def convergence_gate(
    idata: az.InferenceData,
    rhat_cut: float = RHAT_CUT,
    ess_min_cut: float = ESS_MIN_CUT,
    prefixes=VARS_PREFIX,
):
    """
    Returns (pass_gate: bool, summary: dict)
    summary includes max_rhat and min_ess_bulk over selected variables.
    """
    var_names = select_var_names(idata, prefixes=prefixes)

    # One call each (efficient) rather than per-parameter loops
    rh = az.rhat(idata, var_names=var_names, method="rank")
    ess = az.ess(idata, var_names=var_names, method="bulk")

    max_rhat = -np.inf
    min_ess = np.inf

    for v in var_names:
        r = np.asarray(rh[v].values).ravel()
        e = np.asarray(ess[v].values).ravel()
        r = r[np.isfinite(r)]
        e = e[np.isfinite(e)]
        if r.size:
            max_rhat = max(max_rhat, float(np.max(r)))
        if e.size:
            min_ess = min(min_ess, float(np.min(e)))

    if max_rhat == -np.inf:
        max_rhat = np.nan
    if min_ess == np.inf:
        min_ess = np.nan

    pass_gate = (np.isnan(max_rhat) or max_rhat <= rhat_cut) and (np.isnan(min_ess) or min_ess >= ess_min_cut)

    return pass_gate, {
        "max_rhat": max_rhat,
        "min_ess_bulk": min_ess,
        "n_vars_checked": len(var_names),
    }


# ----------------------------
# 3) LOO TABLE + SELECTION
# ----------------------------
def loo_table_with_pareto(
    idata_by_k: dict[int, az.InferenceData],
    scale: str = "log",          # per your ArviZ version: "log" | "negative_log" | "deviance"
    pointwise: bool = True,      # needed for pareto_k diagnostics
    rhat_cut: float = RHAT_CUT,
    ess_min_cut: float = ESS_MIN_CUT,
    prefixes=VARS_PREFIX,
):
    """
    Computes PSIS-LOO per k plus:
      - convergence gate summary
      - pareto-k diagnostics
      - flags for loglik presence and convergence

    Returns:
      df (sorted by elpd_loo desc),
      loos (dict {k: loo_obj} for reuse if needed)
    """
    rows = []
    loos: dict[int, object] = {}

    for k, idata in idata_by_k.items():
        ok_ll = has_pointwise_loglik(idata)
        ok_conv, conv = convergence_gate(idata, rhat_cut=rhat_cut, ess_min_cut=ess_min_cut, prefixes=prefixes)

        if ok_ll:
            loo = az.loo(idata, pointwise=pointwise, scale=scale)
            loos[k] = loo
            pk = np.asarray(loo.pareto_k) if pointwise else np.array([])
            pk = pk[np.isfinite(pk)] if pk.size else pk

            rows.append({
                "k": k,
                "ok_loglik": True,
                "ok_convergence": bool(ok_conv),
                "max_rhat": conv["max_rhat"],
                "min_ess_bulk": conv["min_ess_bulk"],
                "elpd_loo": float(loo.elpd_loo),
                "se": float(loo.se),
                "p_loo": float(loo.p_loo),
                "pareto_k_max": float(np.max(pk)) if pk.size else np.nan,
                "pareto_k_gt_0_7": int(np.sum(pk > 0.7)) if pk.size else np.nan,
                "pareto_k_gt_1_0": int(np.sum(pk > 1.0)) if pk.size else np.nan,
            })
        else:
            rows.append({
                "k": k,
                "ok_loglik": False,
                "ok_convergence": bool(ok_conv),
                "max_rhat": conv["max_rhat"],
                "min_ess_bulk": conv["min_ess_bulk"],
                "elpd_loo": np.nan,
                "se": np.nan,
                "p_loo": np.nan,
                "pareto_k_max": np.nan,
                "pareto_k_gt_0_7": np.nan,
                "pareto_k_gt_1_0": np.nan,
            })

    df = pd.DataFrame(rows)

    # Sort by elpd_loo descending, but keep missing at bottom
    df = df.sort_values(["ok_loglik", "elpd_loo"], ascending=[False, False]).reset_index(drop=True)

    # delta elpd vs best (among models with elpd)
    if df["elpd_loo"].notna().any():
        best = df.loc[df["elpd_loo"].notna(), "elpd_loo"].max()
        df["delta_elpd"] = df["elpd_loo"] - best
    else:
        df["delta_elpd"] = np.nan

    return df, loos


def choose_k_1se(df: pd.DataFrame, se_rule: float = 1.0, require_convergence: bool = True):
    """
    1-SE rule:
      - find best elpd_loo
      - accept models with elpd_loo >= best - se_rule * se_best
      - choose smallest k among accepted
    Optionally require ok_convergence==True.
    """
    use = df.copy()
    use = use[use["ok_loglik"] == True]
    use = use[use["elpd_loo"].notna()]

    if require_convergence and "ok_convergence" in use.columns:
        use = use[use["ok_convergence"] == True]

    if use.empty:
        raise RuntimeError("No models available after filtering (loglik/convergence).")

    best_row = use.sort_values("elpd_loo", ascending=False).iloc[0]
    best_elpd = float(best_row["elpd_loo"])
    best_se = float(best_row["se"])
    cutoff = best_elpd - se_rule * best_se

    candidates = use[use["elpd_loo"] >= cutoff].copy()
    k_choice = int(candidates.sort_values("k").iloc[0]["k"])

    info = {
        "k_choice": k_choice,
        "best_k": int(best_row["k"]),
        "best_elpd": best_elpd,
        "best_se": best_se,
        "cutoff_elpd": cutoff,
        "n_candidates": int(len(candidates)),
    }
    return k_choice, info


# ----------------------------
# 4) PLOTS (minimal but informative)
# ----------------------------
def plot_elpd_by_k(df: pd.DataFrame, title: str):
    d = df[df["ok_loglik"] == True].copy()
    d = d[d["elpd_loo"].notna()].sort_values("k")
    if d.empty:
        print(f"{title}: nothing to plot (no elpd_loo).")
        return

    plt.figure()
    plt.errorbar(d["k"], d["elpd_loo"], yerr=d["se"], fmt="o-")
    plt.xlabel("k")
    plt.ylabel("elpd_loo (±1 SE)")
    plt.title(title)
    plt.tight_layout()
    plt.show()


def plot_pareto_by_k(df: pd.DataFrame, title: str):
    d = df[df["ok_loglik"] == True].copy()
    d = d[d["elpd_loo"].notna()].sort_values("k")
    if d.empty:
        print(f"{title}: nothing to plot (no elpd_loo).")
        return

    plt.figure()
    plt.plot(d["k"], d["pareto_k_gt_0_7"], marker="o", label="count Pareto-k > 0.7")
    plt.plot(d["k"], d["pareto_k_gt_1_0"], marker="o", label="count Pareto-k > 1.0")
    plt.xlabel("k")
    plt.ylabel("count")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.show()


def plot_convergence_by_k(df: pd.DataFrame, title: str):
    d = df.copy().sort_values("k")
    if d.empty:
        print(f"{title}: nothing to plot.")
        return

    plt.figure()
    plt.plot(d["k"], d["max_rhat"], marker="o")
    plt.axhline(RHAT_CUT, linestyle="--")
    plt.xlabel("k")
    plt.ylabel("max rank-Rhat (selected vars)")
    plt.title(title)
    plt.tight_layout()
    plt.show()

    plt.figure()
    plt.plot(d["k"], d["min_ess_bulk"], marker="o")
    plt.axhline(ESS_MIN_CUT, linestyle="--")
    plt.xlabel("k")
    plt.ylabel("min bulk ESS (selected vars)")
    plt.title(title.replace("Rhat", "ESS"))
    plt.tight_layout()
    plt.show()


# ----------------------------
# 5) RUN EVERYTHING + SAVE OUTPUTS
# ----------------------------
def run_k_selection_pipeline(
    base_dir: Path = BASE_DIR,
    valences: tuple[str, ...] = VALENCES,
    subfolder: str = SUBFOLDER,
    factor_range=FACTOR_RANGE,
    scale: str = "log",
    se_rule: float = 1.0,
    require_convergence: bool = True,
    save: bool = True,
):
    idatas_all = load_all_idatas(
        base_dir=base_dir,
        valences=valences,
        subfolder=subfolder,
        factor_range=factor_range,
        parse_from_filenames=True,
    )

    results = {}

    for val in valences:
        idata_by_k = idatas_all.get(val, {})
        if not idata_by_k:
            print(f"\n{val}: no idata files found.")
            continue

        df, loos = loo_table_with_pareto(idata_by_k, scale=scale, pointwise=True)
        print(f"\n{val}: LOO table\n", df)

        try:
            k_choice, info = choose_k_1se(df, se_rule=se_rule, require_convergence=require_convergence)
            print(f"\n{val}: chosen k = {k_choice} (info: {info})")
        except Exception as e:
            k_choice, info = None, {"error": repr(e)}
            print(f"\n{val}: could not choose k. {info['error']}")

        # Plots
        plot_elpd_by_k(df, f"{val}: PSIS-LOO ELPD by k")
        plot_pareto_by_k(df, f"{val}: Pareto-k diagnostics by k")
        plot_convergence_by_k(df, f"{val}: Convergence (Rhat/ESS) by k")

        # Save outputs
        out_dir = base_dir / val / subfolder
        if save:
            out_dir.mkdir(parents=True, exist_ok=True)
            df.to_csv(out_dir / "k_selection_diagnostics.csv", index=False)

            payload = {
                "valence": val,
                "chosen_k": k_choice,
                "choice_info": info,
                "note": "Selection uses PSIS-LOO (elpd_loo) with 1-SE rule; requires pointwise log_likelihood.",
            }
            with open(out_dir / "k_selection_choice.json", "w", encoding="utf-8") as f:
                json.dump(payload, f, indent=2)

        results[val] = {
            "idata_by_k": idata_by_k,
            "loo_df": df,
            "loos": loos,
            "chosen_k": k_choice,
            "choice_info": info,
        }

    return results


# ----------------------------
# 6) EXECUTE
# ----------------------------
# Run this cell to produce tables, plots, and saved outputs.
results = run_k_selection_pipeline(
    base_dir=BASE_DIR,
    valences=VALENCES,
    subfolder=SUBFOLDER,
    factor_range=FACTOR_RANGE,
    scale="log",               # matches your ArviZ version's accepted values
    se_rule=1.0,               # 1-SE rule (conservative); use 0.0 to pick absolute best
    require_convergence=True,  # gate on rhat/ess before selecting
    save=True,
)

# After running, your chosen k values (if available) are:
# results["success"]["chosen_k"]
# results["failure"]["chosen_k"]
```

# Helper diagnostic functions

```python
# ============================================================
# Bayesian-congruent-ish EFA post-processing workflow (NumPyro/ArviZ)
# - Select k via PSIS-LOO (predictive; requires pointwise log_likelihood)
# - Align posterior loadings across groups/chains via:
#     (1) column permutation + sign
#     (2) rotation (orthogonal Procrustes or oblique least-squares)
# - Optionally rotate factor covariance/correlation consistently
# - Tucker's phi:
#     (A) phi on posterior mean loadings (recommended default)
#     (B) phi from cross-paired draws (independent posterior uncertainty)
#     (C) phi from paired-by-index draws (kept, but interpretation is limited)
# - Summaries for loadings/thresholds and basic diagnostics
# ============================================================

import warnings
import itertools
import numpy as np
import pandas as pd
import xarray as xr
import arviz as az

# ----------------------------
# Utilities
# ----------------------------
def _to_samples_matrix(da, item_dim=None, factor_dim=None):
    """
    da: xarray DataArray with dims ('chain','draw', item_dim, factor_dim)
    Returns:
      X: np.ndarray (S, n_items, k) where S = chain*draw
      item_dim, factor_dim
    """
    if "chain" not in da.dims or "draw" not in da.dims:
        raise ValueError(f"Expected dims include ('chain','draw',...). Got {da.dims}")

    other = [d for d in da.dims if d not in ("chain", "draw")]
    if len(other) != 2:
        raise ValueError(f"Expected exactly two non-(chain,draw) dims. Got {other}")

    if item_dim is None or factor_dim is None:
        # Heuristic: if 'item'/'factor' appear, prefer them; else accept order
        if item_dim is None and "item" in other:
            item_dim = "item"
        if factor_dim is None and "factor" in other:
            factor_dim = "factor"
        if item_dim is None or factor_dim is None:
            item_dim, factor_dim = other

    X = (
        da.transpose("chain", "draw", item_dim, factor_dim)
          .stack(sample=("chain", "draw"))
          .transpose("sample", item_dim, factor_dim)
          .values
    )
    return X, item_dim, factor_dim


def _choose_anchor_items(L_ref_draws, mode="maxabs"):
    """
    Choose anchor items per factor from reference group draws.
    mode:
      - 'maxabs': pick item with max |loading| in posterior mean for each factor
      - 'diag': anchor item = factor index if n_items>=k else fallback to maxabs
    """
    L_mean = L_ref_draws.mean(axis=0)  # (n_items, k)
    n_items, k = L_mean.shape

    if mode == "diag" and n_items >= k:
        return np.arange(k)

    return np.argmax(np.abs(L_mean), axis=0)


def _apply_signs(L, signs):
    """L: (..., n_items, k), signs: (..., k) or (k,)"""
    signs = np.asarray(signs)
    return L * signs[..., None, :]


def _best_perm_and_sign(L, T, max_k_bruteforce=8):
    """
    Find column permutation p and sign vector s that best aligns L to T
    using brute force (k<=max_k_bruteforce) or a greedy fallback.

    Objective: maximize sum_j | <L[:, p_j], T[:, j]> | (column-wise absolute dot).
    Returns:
      perm: (k,) indices
      signs: (k,) in {+1,-1}
    """
    n_items, k = L.shape
    # column dot products matrix: D[a,b] = <L[:,a], T[:,b]>
    D = L.T @ T  # (k,k)

    if k <= max_k_bruteforce:
        best_score = -np.inf
        best_perm = None
        best_signs = None
        for perm in itertools.permutations(range(k)):
            perm = np.asarray(perm, dtype=int)
            diag = D[perm, np.arange(k)]
            score = np.sum(np.abs(diag))
            if score > best_score:
                best_score = score
                best_perm = perm
                best_signs = np.sign(diag)
        signs = np.where(best_signs == 0.0, 1.0, best_signs).astype(float)
        return best_perm, signs

    # Greedy fallback (not optimal, but avoids factorial blow-up)
    D_abs = np.abs(D).copy()
    perm = -np.ones(k, dtype=int)
    used_cols = set()
    for j in range(k):
        # pick row a that best matches target column j among unused
        a = int(np.argmax(D_abs[:, j]))
        while a in used_cols:
            D_abs[a, j] = -1.0
            a = int(np.argmax(D_abs[:, j]))
        perm[j] = a
        used_cols.add(a)

    # perm currently maps target j -> source column perm[j]
    # we want an array of source indices ordered by target j
    perm = perm.astype(int)
    diag = D[perm, np.arange(k)]
    signs = np.where(np.sign(diag) == 0.0, 1.0, np.sign(diag)).astype(float)
    return perm, signs


def _orthogonal_procrustes(A, B, enforce_det_pos=True):
    """
    Solve R = argmin ||A R - B|| with R orthogonal (k×k).
    Returns R.
    """
    M = A.T @ B
    U, _, Vt = np.linalg.svd(M, full_matrices=False)
    R = U @ Vt
    if enforce_det_pos and np.linalg.det(R) < 0:
        U[:, -1] *= -1.0
        R = U @ Vt
    return R


def _oblique_ls(A, B, ridge=1e-8):
    """
    Solve R = argmin ||A R - B|| with unconstrained R (k×k).
    Uses ridge-stabilized normal equations.
    """
    k = A.shape[1]
    AtA = A.T @ A + ridge * np.eye(k)
    AtB = A.T @ B
    R = np.linalg.solve(AtA, AtB)
    return R


def _rotate_factor_cov(C, R, rotation_type):
    """
    If loadings are rotated as L' = L R, the latent factors rotate as:
      eta' = R^{-1} eta   (to keep y* invariant)
    Therefore factor covariance transforms as:
      Cov(eta') = R^{-1} Cov(eta) R^{-T}

    For orthogonal R, R^{-1} = R^T, so Cov' = R^T Cov R.

    C: (k,k)
    R: (k,k)
    rotation_type: "orthogonal" or "oblique"
    """
    if rotation_type == "orthogonal":
        return R.T @ C @ R
    else:
        Ri = np.linalg.inv(R)
        return Ri @ C @ Ri.T


# ----------------------------
# 2) Alignment + Tucker phi
# ----------------------------
def align_groups_loadings(
    idata_a,
    idata_b,
    loadings_var="loadings",
    factor_cov_var=None,        # "factor_cov" or "factor_corr" if you want to rotate it too
    ref_group="a",              # "a" or "b"
    anchor_mode="maxabs",
    rotation="auto",            # "auto" | "orthogonal" | "oblique"
    enforce_det_pos=True,       # only for orthogonal
    ridge=1e-8,                 # only for oblique LS
    max_k_bruteforce=8,
):
    """
    Align posterior loadings between two independent fits (e.g., success vs failure).

    Steps:
      1) Choose reference matrix T from reference group posterior mean loadings.
      2) For each draw:
         (a) find best column permutation + sign to match T (using brute-force if k<=max_k_bruteforce)
         (b) estimate rotation R (orthogonal Procrustes or oblique LS)
         (c) apply L' = L_perm_sign R
         (d) optionally rotate factor covariance/correlation accordingly

    rotation="auto":
      - chooses "orthogonal" if factor_cov_var is None OR if factor covariance is ~I on average;
      - otherwise chooses "oblique" (safer when factors correlated).

    Returns dict with aligned loadings DataArrays and rotation matrices per draw.
    """
    da_a = idata_a.posterior[loadings_var]
    da_b = idata_b.posterior[loadings_var]

    La_draws, item_dim_a, factor_dim_a = _to_samples_matrix(da_a)
    Lb_draws, item_dim_b, factor_dim_b = _to_samples_matrix(da_b)

    if La_draws.shape[1:] != Lb_draws.shape[1:]:
        raise ValueError(f"Loadings shape mismatch: {La_draws.shape} vs {Lb_draws.shape}")

    S_a, n_items, k = La_draws.shape
    S_b = Lb_draws.shape[0]

    # Choose reference group
    Lref_draws = La_draws if ref_group == "a" else Lb_draws
    anchor_items = _choose_anchor_items(Lref_draws, mode=anchor_mode)

    # Reference target matrix T: sign-align reference group by anchors, then mean
    def sign_align_by_anchors(L_draws):
        anchor_vals = L_draws[:, anchor_items, np.arange(k)]
        signs = np.sign(anchor_vals)
        signs = np.where(signs == 0.0, 1.0, signs)
        return L_draws * signs[:, None, :]

    Lref_signed = sign_align_by_anchors(Lref_draws)
    T = Lref_signed.mean(axis=0)  # (n_items, k)

    # Decide rotation type
    rot_type = rotation
    if rot_type == "auto":
        # If we have factor covariance/correlation, inspect average off-diagonal magnitude
        if factor_cov_var is None or factor_cov_var not in idata_a.posterior:
            rot_type = "orthogonal"
        else:
            Cbar = idata_a.posterior[factor_cov_var].mean(("chain", "draw")).values
            off = Cbar - np.diag(np.diag(Cbar))
            if np.max(np.abs(off)) < 1e-3:
                rot_type = "orthogonal"
            else:
                rot_type = "oblique"

    if rot_type == "orthogonal":
        pass
    elif rot_type == "oblique":
        warnings.warn(
            "Using oblique LS alignment. This changes factor covariance under the implied inverse rotation. "
            "If you report factor correlations/covariances, rotate them consistently (enabled if factor_cov_var is set)."
        )
    else:
        raise ValueError("rotation must be 'auto', 'orthogonal', or 'oblique'")

    # Align one group draws to T
    def align_draws(L_draws):
        S = L_draws.shape[0]
        L_aligned = np.empty_like(L_draws)
        R_all = np.empty((S, k, k), dtype=L_draws.dtype)
        perm_all = np.empty((S, k), dtype=int)
        sign_all = np.empty((S, k), dtype=L_draws.dtype)

        for s in range(S):
            # start from sign alignment by shared anchors (stabilizes sign a bit)
            Ls = L_draws[s].copy()
            # anchor sign convention
            anchor_vals = Ls[anchor_items, np.arange(k)]
            sgn = np.sign(anchor_vals)
            sgn = np.where(sgn == 0.0, 1.0, sgn).astype(float)
            Ls = _apply_signs(Ls[None, ...], sgn[None, :])[0]

            # column permutation + additional signs to best match target T
            perm, signs = _best_perm_and_sign(Ls, T, max_k_bruteforce=max_k_bruteforce)
            Lp = Ls[:, perm]
            Lp = _apply_signs(Lp[None, ...], signs[None, :])[0]

            # rotation
            if rot_type == "orthogonal":
                R = _orthogonal_procrustes(Lp, T, enforce_det_pos=enforce_det_pos)
            else:
                R = _oblique_ls(Lp, T, ridge=ridge)

            L_aligned[s] = Lp @ R
            R_all[s] = R
            perm_all[s] = perm
            sign_all[s] = signs

        return L_aligned, R_all, perm_all, sign_all

    La_aligned, Ra, perma, signa = align_draws(La_draws)
    Lb_aligned, Rb, permb, signb = align_draws(Lb_draws)

    # Rebuild DataArrays with original chain/draw dims
    n_chain_a = da_a.sizes["chain"]; n_draw_a = da_a.sizes["draw"]
    n_chain_b = da_b.sizes["chain"]; n_draw_b = da_b.sizes["draw"]

    La_aligned_4d = La_aligned.reshape(n_chain_a, n_draw_a, n_items, k)
    Lb_aligned_4d = Lb_aligned.reshape(n_chain_b, n_draw_b, n_items, k)

    da_a_aligned = xr.DataArray(
        La_aligned_4d,
        dims=("chain", "draw", item_dim_a, factor_dim_a),
        coords={
            "chain": idata_a.posterior["chain"].values,
            "draw": idata_a.posterior["draw"].values,
            item_dim_a: da_a.coords[item_dim_a].values,
            factor_dim_a: da_a.coords[factor_dim_a].values,
        },
        name=f"{loadings_var}_aligned",
    )

    da_b_aligned = xr.DataArray(
        Lb_aligned_4d,
        dims=("chain", "draw", item_dim_b, factor_dim_b),
        coords={
            "chain": idata_b.posterior["chain"].values,
            "draw": idata_b.posterior["draw"].values,
            item_dim_b: da_b.coords[item_dim_b].values,
            factor_dim_b: da_b.coords[factor_dim_b].values,
        },
        name=f"{loadings_var}_aligned",
    )

    out = {
        "target_T": T,
        "anchor_items": anchor_items,
        "rotation_type": rot_type,
        "a_loadings_aligned": da_a_aligned,
        "b_loadings_aligned": da_b_aligned,
        "a_rotations": Ra,
        "b_rotations": Rb,
        "a_permutations": perma,
        "b_permutations": permb,
        "a_signs": signa,
        "b_signs": signb,
    }

    # Optionally rotate factor covariance/correlation
    if factor_cov_var is not None:
        for label, idata, R_all in (("a", idata_a, Ra), ("b", idata_b, Rb)):
            if factor_cov_var not in idata.posterior:
                warnings.warn(f"{factor_cov_var} not found in posterior for group '{label}'. Skipping rotation.")
                continue

            daC = idata.posterior[factor_cov_var]
            # stack samples
            C = (
                daC.transpose("chain", "draw", ...)
                   .stack(sample=("chain", "draw"))
                   .transpose("sample", ...)
                   .values
            )  # (S,k,k)
            if C.shape[0] != R_all.shape[0]:
                warnings.warn(f"Sample count mismatch for {factor_cov_var} in group '{label}'. Skipping rotation.")
                continue

            C_rot = np.empty_like(C)
            for s in range(C.shape[0]):
                C_rot[s] = _rotate_factor_cov(C[s], R_all[s], rot_type)

            C_rot_4d = C_rot.reshape(daC.sizes["chain"], daC.sizes["draw"], k, k)
            daC_rot = xr.DataArray(
                C_rot_4d,
                dims=daC.dims,
                coords=daC.coords,
                name=f"{factor_cov_var}_aligned",
            )
            out[f"{label}_{factor_cov_var}_aligned"] = daC_rot

    return out


def tucker_phi(L1, L2, eps=1e-12):
    """Tucker's phi per factor for two loading matrices (n_items,k)."""
    num = np.sum(L1 * L2, axis=0)  # (k,)
    den = np.sqrt(np.sum(L1 * L1, axis=0) * np.sum(L2 * L2, axis=0)) + eps
    return num / den


def tucker_phi_modes(
    da_a_aligned,
    da_b_aligned,
    mode="mean",            # "mean" | "paired" | "crosspair"
    crosspair_n=2000,
    seed=0,
    warn_on_trunc=True,
):
    """
    Compute Tucker's phi with different interpretations.

    mode="mean":
      - phi computed on posterior mean loadings (recommended default).
      - also returns an uncertainty distribution via crosspair subsampling.

    mode="paired":
      - pairs draws by index (arbitrary if independent fits). Kept for continuity, but interpret cautiously.

    mode="crosspair":
      - sample random cross-pairs (i,j) from independent posterior draws; this is well-defined as
        uncertainty under independent posteriors.

    Returns dict with:
      - phi_draws: (S,k) if applicable
      - summary: median, 90% interval, P(phi>0.95)
      - phi_mean: (k,) for mean mode
    """
    La, item_dim_a, factor_dim_a = _to_samples_matrix(da_a_aligned)
    Lb, item_dim_b, factor_dim_b = _to_samples_matrix(da_b_aligned)

    if La.shape[1:] != Lb.shape[1:]:
        raise ValueError("Aligned loadings have mismatched shapes.")

    Sa, n_items, k = La.shape
    Sb = Lb.shape[0]
    rng = np.random.default_rng(seed)

    def summarize(phi_draws):
        phi_median = np.median(phi_draws, axis=0)
        phi_lo = np.quantile(phi_draws, 0.05, axis=0)
        phi_hi = np.quantile(phi_draws, 0.95, axis=0)
        p_gt_095 = np.mean(phi_draws > 0.95, axis=0)
        return {
            "phi_median": phi_median,
            "phi_hdi90": (phi_lo, phi_hi),
            "p_gt_095": p_gt_095,
        }

    if mode == "mean":
        L1m = La.mean(axis=0)
        L2m = Lb.mean(axis=0)
        phi_mean = tucker_phi(L1m, L2m)

        # Also provide a crosspair distribution for uncertainty
        n = min(crosspair_n, Sa * Sb)
        idx_a = rng.integers(0, Sa, size=n)
        idx_b = rng.integers(0, Sb, size=n)
        phi_draws = np.empty((n, k), dtype=float)
        for t in range(n):
            phi_draws[t] = tucker_phi(La[idx_a[t]], Lb[idx_b[t]])
        out = {"phi_mean": phi_mean, "phi_draws": phi_draws}
        out.update(summarize(phi_draws))
        return out

    if mode == "paired":
        S = min(Sa, Sb)
        if warn_on_trunc and Sa != Sb:
            warnings.warn(
                f"Unequal draw counts: a={Sa}, b={Sb}. "
                f"Truncating to {S} and pairing draws by index (arbitrary pairing)."
            )
        phi_draws = np.empty((S, k), dtype=float)
        for s in range(S):
            phi_draws[s] = tucker_phi(La[s], Lb[s])
        out = {"phi_draws": phi_draws}
        out.update(summarize(phi_draws))
        return out

    if mode == "crosspair":
        n = min(crosspair_n, Sa * Sb)
        idx_a = rng.integers(0, Sa, size=n)
        idx_b = rng.integers(0, Sb, size=n)
        phi_draws = np.empty((n, k), dtype=float)
        for t in range(n):
            phi_draws[t] = tucker_phi(La[idx_a[t]], Lb[idx_b[t]])
        out = {"phi_draws": phi_draws}
        out.update(summarize(phi_draws))
        return out

    raise ValueError("mode must be 'mean', 'paired', or 'crosspair'")


# ----------------------------
# 3) Bayesian summaries for reporting
# ----------------------------
def summarize_loadings(da_loadings, ci=0.90, item_names=None, factor_names=None, stat="mean"):
    """
    da_loadings: xarray DataArray (chain, draw, item, factor)
    Returns DataFrame with center and credible intervals per loading.
    """
    if ci > 1:
        ci = ci / 100.0
    alpha = (1 - ci) / 2

    other = [d for d in da_loadings.dims if d not in ("chain", "draw")]
    if len(other) != 2:
        raise ValueError(f"Expected (chain,draw,item,factor). Got {da_loadings.dims}")
    item_dim, factor_dim = other

    if stat == "mean":
        center = da_loadings.mean(("chain", "draw"))
    elif stat == "median":
        center = da_loadings.median(("chain", "draw"))
    else:
        raise ValueError("stat must be 'mean' or 'median'")

    lo = da_loadings.quantile(alpha, dim=("chain", "draw"))
    hi = da_loadings.quantile(1 - alpha, dim=("chain", "draw"))

    n_items = da_loadings.sizes[item_dim]
    k = da_loadings.sizes[factor_dim]
    if item_names is None:
        item_names = [f"Item_{i+1}" for i in range(n_items)]
    if factor_names is None:
        factor_names = [f"F{j+1}" for j in range(k)]

    out = pd.DataFrame(index=item_names)
    for j, f in enumerate(factor_names):
        out[f"{f}_{stat}"] = center.isel({factor_dim: j}).values
        out[f"{f}_lo"] = lo.isel({factor_dim: j}).values
        out[f"{f}_hi"] = hi.isel({factor_dim: j}).values
    return out


def summarize_thresholds(idata, thresholds_var="thresholds", ci=0.90, item_names=None, stat="mean"):
    """
    Expects idata.posterior[thresholds_var] dims: (chain, draw, item, threshold)
    Returns long DataFrame.
    """
    da = idata.posterior[thresholds_var]
    if ci > 1:
        ci = ci / 100.0
    alpha = (1 - ci) / 2

    other = [d for d in da.dims if d not in ("chain", "draw")]
    if len(other) != 2:
        raise ValueError(f"Expected thresholds dims (chain,draw,item,threshold). Got {da.dims}")
    item_dim, thr_dim = other

    if stat == "mean":
        center = da.mean(("chain", "draw"))
    elif stat == "median":
        center = da.median(("chain", "draw"))
    else:
        raise ValueError("stat must be 'mean' or 'median'")

    lo = da.quantile(alpha, dim=("chain", "draw"))
    hi = da.quantile(1 - alpha, dim=("chain", "draw"))

    n_items = da.sizes[item_dim]
    n_thr = da.sizes[thr_dim]
    if item_names is None:
        item_names = [f"Item_{i+1}" for i in range(n_items)]

    records = []
    for i in range(n_items):
        for t in range(n_thr):
            records.append({
                "Item": item_names[i],
                "Threshold": f"t{t+1}",
                stat: float(center.isel({item_dim: i, thr_dim: t}).values),
                "lo": float(lo.isel({item_dim: i, thr_dim: t}).values),
                "hi": float(hi.isel({item_dim: i, thr_dim: t}).values),
            })
    return pd.DataFrame.from_records(records)


def convergence_summary(idata, vars_prefix=("lam", "loadings", "thresholds", "factor_corr", "factor_cov"), rhat_cut=1.01):
    """
    Basic diagnostics focusing on key variables.
    Returns DataFrame: var, max_rhat, min_ess_bulk.
    """
    # Restrict to variables you care about (avoids deterministics/noise)
    vars_sel = [v for v in idata.posterior.data_vars if any(v.startswith(p) for p in vars_prefix)]
    if not vars_sel:
        vars_sel = list(idata.posterior.data_vars)

    rh = az.rhat(idata, var_names=vars_sel, method="rank")
    ess = az.ess(idata, var_names=vars_sel, method="bulk")

    rows = []
    for v in vars_sel:
        r = np.asarray(rh[v].values)
        e = np.asarray(ess[v].values)
        r = r[np.isfinite(r)]
        e = e[np.isfinite(e)]
        rows.append({
            "var": v,
            "max_rhat": float(np.max(r)) if r.size else np.nan,
            "min_ess_bulk": float(np.min(e)) if e.size else np.nan,
            "rhat_ok": bool(np.max(r) <= rhat_cut) if r.size else True,
        })
    return pd.DataFrame(rows).sort_values(["rhat_ok", "max_rhat"], ascending=[False, True])
```

```python
import jax
import jax.numpy as jnp
from jax.scipy.special import log_ndtr
from numpyro.distributions import Distribution, constraints

def _log1mexp(x: jnp.ndarray) -> jnp.ndarray:
    """
    Numerically stable log(1 - exp(x)) for x <= 0.
    Piecewise:
      if x < log(0.5): log1p(-exp(x))
      else:           log(-expm1(x))
    """
    log_half = -0.6931471805599453  # log(0.5)
    return jnp.where(x < log_half, jnp.log1p(-jnp.exp(x)), jnp.log(-jnp.expm1(x)))


def _sample_truncnorm_pos(name: str, loc: float, scale: float, shape: int | tuple[int, ...]):
    """
    Version-tolerant helper for NumPyro TruncatedNormal with low=0.
    Some NumPyro versions differ in signature; this fails loudly if incompatible.
    """
    shape_tup = (shape,) if isinstance(shape, int) else tuple(shape)
    try:
        d = dist.TruncatedNormal(loc=loc, scale=scale, low=0.0)
    except TypeError:
        # Older signature variants sometimes accept (loc, scale, low, high) positionally.
        try:
            d = dist.TruncatedNormal(loc, scale, 0.0, jnp.inf)
        except TypeError as e:
            raise TypeError(
                "NumPyro TruncatedNormal API mismatch. "
                "Please pin/upgrade numpyro or adjust the TruncatedNormal call."
            ) from e
    return numpyro.sample(name, d.expand(shape_tup))


class OrderedProbit(Distribution):
    """
    Ordered probit for integer categories {0, ..., K-1}, K = n_thr + 1.

    mu:         batch_shape, typically [n_obs, n_items]
    thresholds: one of:
      - [n_thr]                      shared thresholds
      - [n_items, n_thr]             item-specific thresholds
      - otherwise broadcastable to mu.shape + (n_thr,)
    """

    arg_constraints = {"mu": constraints.real, "thresholds": constraints.ordered_vector}

    def __init__(self, mu, thresholds, validate_args=None):
        self.mu = jnp.asarray(mu)
        thr = jnp.asarray(thresholds)

        if thr.ndim < 1:
            raise ValueError("thresholds must have at least 1 dimension (last dim = n_thr).")

        self.n_thr = int(thr.shape[-1])
        self.n_cats = self.n_thr + 1

        # Explicit, stricter broadcasting rules
        if thr.ndim == 1:
            # [n_thr] -> [1, 1, ..., 1, n_thr]
            thr = thr.reshape((1,) * self.mu.ndim + (self.n_thr,))
        elif thr.ndim == 2:
            # Only accept [n_items, n_thr], where n_items matches last dim of mu.
            if self.mu.ndim < 1:
                raise ValueError("mu must have at least 1 dimension.")
            n_items = self.mu.shape[-1]
            if thr.shape[0] != n_items:
                raise ValueError(
                    f"2D thresholds must have shape [n_items, n_thr] = [{n_items}, {self.n_thr}], "
                    f"but got {thr.shape}."
                )
            thr = thr.reshape((1,) * (self.mu.ndim - 1) + (n_items, self.n_thr))
        else:
            # For higher-rank thresholds, require broadcastability to mu.shape+(n_thr,)
            pass

        # This will raise if not broadcastable
        self.thresholds = jnp.broadcast_to(thr, self.mu.shape + (self.n_thr,))

        super().__init__(batch_shape=self.mu.shape, event_shape=(), validate_args=validate_args)

    @property
    def support(self):
        # Correct bounded support: {0, ..., n_cats-1}
        return constraints.integer_interval(0, self.n_cats - 1)

    def log_prob(self, value):
        y = jnp.asarray(value).astype(jnp.int32)
        in_range = (y >= 0) & (y < self.n_cats)

        # Safe indexing for gather; invalid values masked to -inf at end.
        y_idx = jnp.clip(y, 0, self.n_cats - 1)
        y_exp = y_idx[..., None]

        # Bounds in probit scale are (tau - mu)
        t = self.thresholds - self.mu[..., None]

        neg_inf = jnp.full(self.mu.shape + (1,), -jnp.inf)
        pos_inf = jnp.full(self.mu.shape + (1,),  jnp.inf)
        t_full = jnp.concatenate([neg_inf, t, pos_inf], axis=-1)  # [..., n_cats+1]

        lower = jnp.take_along_axis(t_full[..., :-1], y_exp, axis=-1)[..., 0]
        upper = jnp.take_along_axis(t_full[...,  1:], y_exp, axis=-1)[..., 0]

        ordered = upper >= lower

        log_cdf_u = log_ndtr(upper)
        log_cdf_l = log_ndtr(lower)

        # Compute log interval probability only if ordered; otherwise -inf.
        d = log_cdf_l - log_cdf_u                      # should be <= 0 when ordered
        d_safe = jnp.minimum(d, -1e-12)                # prevent log(0) when interval ~ 0
        log_p_ordered = log_cdf_u + _log1mexp(d_safe)

        log_p = jnp.where(ordered, log_p_ordered, -jnp.inf)
        log_p = jnp.where(in_range, log_p, -jnp.inf)
        return log_p

    def sample(self, key, sample_shape=()):
        shape = sample_shape + self.batch_shape
        z = self.mu + jax.random.normal(key, shape)
        y = jnp.sum(z[..., None] > self.thresholds, axis=-1)
        return y.astype(jnp.int32)
```

```python
# ============================================================
# Bayesian Ordinal Multi-Group CFA (MG-CFA) v3 (FULL UPDATED)
#
# Fixes / changes vs v2:
#  1) Fixed syntax bug in OrderedProbit.sample (duplicate return removed).
#  2) Added _unique_preserve_order helper (order-preserving de-duplication).
#  3) Removed remaining int(...) casts on potentially traced JAX values inside model().
#     All index sets and their sizes are made static in __init__ using NumPy.
#  4) Strong ID guard: if invariance="scalar" AND latent means are estimated
#     (this model does), then threshold_location must be "fixed0". Otherwise raise.
#  5) Partial invariance: marker pairs in loadings_free_pairs are dropped with warning
#     (marker loadings are fixed to 1 under marker identification).
#  6) Kept stable OrderedProbit log_prob (log_ndtr + log1mexp) with strict interval gate.
#
# Defaults remain: identification="marker", corr_invariance=True,
# threshold_location="fixed0", model_type="cfa".
# ============================================================

from __future__ import annotations

from typing import Optional, Sequence, Dict, List, Tuple
import warnings

import numpy as np
import jax
import jax.numpy as jnp
from jax.nn import softplus
from jax.scipy.special import log_ndtr

import numpyro
import numpyro.distributions as dist
from numpyro.distributions import Distribution, constraints


# ----------------------------
# utilities
# ----------------------------
def _unique_preserve_order(x: np.ndarray) -> np.ndarray:
    """Order-preserving unique for 1D integer array."""
    seen = set()
    out = []
    for v in x.tolist():
        if v not in seen:
            seen.add(v)
            out.append(v)
    return np.asarray(out, dtype=x.dtype)


def _log1mexp(x: jnp.ndarray) -> jnp.ndarray:
    """
    Numerically stable log(1 - exp(x)) for x <= 0.
    Piecewise:
      if x < log(0.5): log1p(-exp(x))
      else:           log(-expm1(x))
    """
    log_half = -0.6931471805599453
    return jnp.where(x < log_half, jnp.log1p(-jnp.exp(x)), jnp.log(-jnp.expm1(x)))


def _sample_truncnorm_pos(name: str, loc: float, scale: float, shape: int | tuple[int, ...]):
    """Version-tolerant helper for NumPyro TruncatedNormal with low=0."""
    shape_tup = (shape,) if isinstance(shape, int) else tuple(shape)
    try:
        d = dist.TruncatedNormal(loc=loc, scale=scale, low=0.0)
    except TypeError:
        try:
            d = dist.TruncatedNormal(loc, scale, 0.0, jnp.inf)
        except TypeError as e:
            raise TypeError(
                "NumPyro TruncatedNormal API mismatch. Pin/upgrade numpyro or adjust TruncatedNormal call."
            ) from e
    return numpyro.sample(name, d.expand(shape_tup))


# ----------------------------
# Likelihood
# ----------------------------
class OrderedProbit(Distribution):
    """
    Ordered probit for integer categories {0, ..., K-1}, K = n_thr + 1.

    mu:         batch_shape
    thresholds: [n_thr] shared OR [n_items, n_thr] item thresholds OR broadcastable to mu.shape+(n_thr,)
    """

    # Note: constraints.ordered_vector is 1D; validate_args typically False in usage.
    arg_constraints = {"mu": constraints.real, "thresholds": constraints.ordered_vector}

    def __init__(self, mu, thresholds, validate_args=None):
        self.mu = jnp.asarray(mu)
        thr = jnp.asarray(thresholds)

        if thr.ndim < 1:
            raise ValueError("thresholds must have at least 1 dimension (last dim = n_thr).")

        self.n_thr = int(thr.shape[-1])
        self.n_cats = self.n_thr + 1

        # Explicit broadcasting rules
        if thr.ndim == 1:
            thr = thr.reshape((1,) * self.mu.ndim + (self.n_thr,))
        elif thr.ndim == 2:
            if self.mu.ndim < 1:
                raise ValueError("mu must have at least 1 dimension.")
            n_items = self.mu.shape[-1]
            if thr.shape[0] != n_items:
                raise ValueError(
                    f"2D thresholds must be [n_items, n_thr]=[{n_items},{self.n_thr}], got {thr.shape}."
                )
            thr = thr.reshape((1,) * (self.mu.ndim - 1) + (n_items, self.n_thr))
        else:
            # Higher-rank thresholds: must be broadcastable to mu.shape+(n_thr,)
            pass

        self.thresholds = jnp.broadcast_to(thr, self.mu.shape + (self.n_thr,))
        super().__init__(batch_shape=self.mu.shape, event_shape=(), validate_args=validate_args)

    @property
    def support(self):
        return constraints.integer_interval(0, self.n_cats - 1)

    def log_prob(self, value):
        y = jnp.asarray(value).astype(jnp.int32)
        in_range = (y >= 0) & (y < self.n_cats)

        # Safe gather indices
        y_idx = jnp.clip(y, 0, self.n_cats - 1)
        y_exp = y_idx[..., None]

        # Bounds are (tau - mu)
        t = self.thresholds - self.mu[..., None]

        neg_inf = jnp.full(self.mu.shape + (1,), -jnp.inf)
        pos_inf = jnp.full(self.mu.shape + (1,),  jnp.inf)
        t_full = jnp.concatenate([neg_inf, t, pos_inf], axis=-1)  # [..., n_cats+1]

        lower = jnp.take_along_axis(t_full[..., :-1], y_exp, axis=-1)[..., 0]
        upper = jnp.take_along_axis(t_full[...,  1:], y_exp, axis=-1)[..., 0]

        # Strict ordering: upper == lower => probability 0 => -inf
        ordered = upper > lower

        log_cdf_u = log_ndtr(upper)
        log_cdf_l = log_ndtr(lower)

        # log(Φ(u) - Φ(l)) = logΦ(u) + log(1 - exp(logΦ(l) - logΦ(u)))
        d = log_cdf_l - log_cdf_u  # should be < 0 when ordered is True
        eps = jnp.asarray(1e-12)
        d_safe = jnp.where(d > -eps, -eps, d)

        log_p_ordered = log_cdf_u + _log1mexp(d_safe)

        log_p = jnp.where(ordered, log_p_ordered, -jnp.inf)
        log_p = jnp.where(in_range, log_p, -jnp.inf)
        return log_p

    def sample(self, key, sample_shape=()):
        shape = sample_shape + self.batch_shape
        z = self.mu + jax.random.normal(key, shape)
        y = jnp.sum(z[..., None] > self.thresholds, axis=-1)
        return y.astype(jnp.int32)


# ----------------------------
# MG-CFA model
# ----------------------------
class BayesianOrdinalMGCFA:
    """
    Bayesian ordinal multi-group CFA with optional invariance.

    identification: "marker" (default) or "unit_variance"
    invariance: "configural" | "metric" | "scalar"
      - metric: loadings invariant (except partial)
      - scalar (ordinal): thresholds invariant (except partial) AND loadings invariant
    corr_invariance: shared vs group-specific factor correlations
    threshold_location: "fixed0" | "free"
      - this model estimates latent means for groups 1..G-1; for scalar invariance,
        threshold_location must be fixed0 (enforced) to avoid location confounding.
    model_type: "cfa" | "null"

    Partial invariance keys:
      - loadings_free_factors: list[int]
      - thresholds_free_factors: list[int]
      - variances_free_factors: list[int]  (marker only)
      - thresholds_free_items: list[int]
      - loadings_free_pairs: list[(item,factor)]  # only primary pairs are respected
    """

    def __init__(
        self,
        n_factors: int,
        n_cats: int,
        group: jnp.ndarray,
        n_groups: int,
        item_to_factor_indices: Sequence[Sequence[int]],

        invariance: str = "metric",
        partial_invariance: Optional[Dict] = None,
        identification: str = "marker",       # "marker" | "unit_variance"
        corr_invariance: bool = True,
        threshold_location: str = "fixed0",   # "fixed0" | "free"
        model_type: str = "cfa",              # "cfa" | "null"

        # priors for loadings
        loading_loc_prior: float = 0.6,
        loading_scale_prior: float = 0.2,
        loading_prior: str = "truncnorm",     # "truncnorm" | "lognormal"
        cross_loadings: bool = False,
        cross_loading_scale: float = 0.05,

        # thresholds
        min_threshold_spacing: float = 0.05,
        spacing_base_sd: float = 0.5,
        use_hierarchical_spacing: bool = False,
        spacing_scale_prior: float = 0.5,     # HalfNormal

        thr_dev_scale_prior: float = 0.15,
        load_dev_scale_prior: float = 0.10,

        # latent distribution
        df_latent: float = 7.0,
        latent_mean_scale: float = 0.5,
        lkj_concentration: float = 4.0,

        # factor SDs (marker only)
        var_dev_scale_prior: float = 0.10,
        equal_latent_sds_by_default: bool = True,

        store_deterministics: bool = True,
        debug_shapes: bool = False,
        validate_primary_structure: bool = True,
    ):
        self.n_factors = int(n_factors)
        self.n_cats = int(n_cats)
        self.group = jnp.asarray(group).astype(jnp.int32)
        self.n_groups = int(n_groups)

        self.invariance = str(invariance).lower()
        self.identification = str(identification).lower()
        self.corr_invariance = bool(corr_invariance)
        self.threshold_location = str(threshold_location).lower()
        self.model_type = str(model_type).lower()

        self.loading_loc_prior = float(loading_loc_prior)
        self.loading_scale_prior = float(loading_scale_prior)
        self.loading_prior = str(loading_prior).lower()

        self.cross_loadings = bool(cross_loadings)
        self.cross_loading_scale = float(cross_loading_scale)

        self.min_threshold_spacing = float(min_threshold_spacing)
        self.spacing_base_sd = float(spacing_base_sd)
        self.use_hierarchical_spacing = bool(use_hierarchical_spacing)
        self.spacing_scale_prior = float(spacing_scale_prior)

        self.thr_dev_scale_prior = float(thr_dev_scale_prior)
        self.load_dev_scale_prior = float(load_dev_scale_prior)

        self.df_latent = float(df_latent)
        self.latent_mean_scale = float(latent_mean_scale)
        self.lkj_concentration = float(lkj_concentration)

        self.var_dev_scale_prior = float(var_dev_scale_prior)
        self.equal_latent_sds_by_default = bool(equal_latent_sds_by_default)

        self.store_deterministics = bool(store_deterministics)
        self.debug_shapes = bool(debug_shapes)
        self.validate_primary_structure = bool(validate_primary_structure)

        # ---- validation
        if self.n_cats < 2:
            raise ValueError("n_cats must be >= 2.")
        if self.invariance not in ("configural", "metric", "scalar"):
            raise ValueError("invariance must be one of: configural, metric, scalar.")
        if self.identification not in ("marker", "unit_variance"):
            raise ValueError("identification must be 'marker' or 'unit_variance'.")
        if self.threshold_location not in ("fixed0", "free"):
            raise ValueError("threshold_location must be 'fixed0' or 'free'.")
        if self.model_type not in ("cfa", "null"):
            raise ValueError("model_type must be 'cfa' or 'null'.")
        if self.loading_prior not in ("truncnorm", "lognormal"):
            raise ValueError("loading_prior must be 'truncnorm' or 'lognormal'.")
        if self.lkj_concentration <= 0:
            raise ValueError("lkj_concentration must be > 0.")
        if self.n_groups < 1:
            raise ValueError("n_groups must be >= 1.")
        if self.spacing_base_sd <= 0:
            raise ValueError("spacing_base_sd must be > 0.")
        if self.spacing_scale_prior <= 0:
            raise ValueError("spacing_scale_prior must be > 0.")
        if self.min_threshold_spacing <= 0:
            raise ValueError("min_threshold_spacing must be > 0.")

        # Strong ID guard for scalar invariance + latent means:
        # This model estimates group latent means by default.
        if self.invariance == "scalar" and self.threshold_location != "fixed0":
            raise ValueError(
                "For scalar invariance with estimated latent means, threshold_location must be 'fixed0' "
                "to avoid location confounding."
            )

        # ---- sanitize item_to_factor_indices; infer n_items
        idx_lists_np: List[np.ndarray] = []
        for f, ix in enumerate(item_to_factor_indices):
            arr = np.asarray(ix, dtype=int)
            if arr.size == 0:
                raise ValueError(f"Factor {f} has no items.")
            if np.any(arr < 0):
                raise ValueError(f"Negative item index in factor {f}.")
            arr_u = _unique_preserve_order(arr)
            idx_lists_np.append(arr_u)

        all_idx = np.concatenate(idx_lists_np)
        self.n_items = int(all_idx.max()) + 1
        if self.n_items <= 0:
            raise ValueError("Could not infer n_items.")

        # ---- validate primary structure coverage/overlap
        if validate_primary_structure:
            counts = np.zeros(self.n_items, dtype=int)
            for arr in idx_lists_np:
                counts[arr] += 1
            overlap = np.where(counts > 1)[0]
            missing = np.where(counts == 0)[0]
            if overlap.size > 0 and not self.cross_loadings:
                raise ValueError(
                    f"Primary structure overlap: items assigned to multiple factors: {overlap.tolist()}. "
                    "Fix item_to_factor_indices or set cross_loadings=True."
                )
            if missing.size > 0:
                raise ValueError(f"Some items are not assigned to any factor: {missing.tolist()}.")

        # ---- store as JAX arrays (static)
        self.item_to_factor_indices = [jnp.asarray(ix, dtype=jnp.int32) for ix in idx_lists_np]
        F = self.n_factors

        # Markers: first item per factor list (user order preserved)
        self.marker_items = jnp.asarray([ix[0] for ix in idx_lists_np], dtype=jnp.int32)

        # primary mask [n_items, F]
        primary_mask = jnp.zeros((self.n_items, F), dtype=bool)
        for f, idx in enumerate(self.item_to_factor_indices):
            primary_mask = primary_mask.at[idx, f].set(True)
        self.primary_mask = primary_mask

        # mask of primary loadings excluding marker items (marker id)
        marker_mask = jnp.zeros((self.n_items, F), dtype=bool)
        for f in range(F):
            marker_mask = marker_mask.at[self.marker_items[f], f].set(True)
        self.nonmarker_primary_mask = self.primary_mask & (~marker_mask)

        # ---- partial invariance spec
        self.partial_invariance = partial_invariance or {}
        self.partial_invariance.setdefault("loadings_free_factors", [])
        self.partial_invariance.setdefault("thresholds_free_factors", [])
        self.partial_invariance.setdefault("variances_free_factors", [])
        self.partial_invariance.setdefault("thresholds_free_items", [])
        self.partial_invariance.setdefault("loadings_free_pairs", [])

        # Drop marker pairs in loadings_free_pairs under marker ID (they are fixed anyway)
        if self.identification == "marker":
            pairs = self.partial_invariance.get("loadings_free_pairs", [])
            if pairs:
                filtered = []
                marker_map = {(int(self.marker_items[f]), f) for f in range(F)}
                dropped = []
                for (i, f) in pairs:
                    tup = (int(i), int(f))
                    if tup in marker_map:
                        dropped.append(tup)
                    else:
                        filtered.append((int(i), int(f)))
                if dropped:
                    warnings.warn(
                        f"Dropping loadings_free_pairs for marker positions (fixed to 1 under marker ID): {dropped}",
                        stacklevel=2,
                    )
                self.partial_invariance["loadings_free_pairs"] = filtered

        # ---- build masks (JAX) + static indices (NumPy)
        self._free_threshold_items = self._build_free_threshold_item_mask()   # [n_items] bool
        self._free_loading_mask = self._build_free_loading_mask()            # [n_items,F] bool
        self._free_variance_mask = self._build_free_variance_mask()          # [F] bool

        # Threshold free item indices (static)
        thr_free_idx_np = np.where(np.asarray(self._free_threshold_items))[0].astype(np.int32)
        self._thr_free_item_idx = jnp.asarray(thr_free_idx_np, dtype=jnp.int32)
        self._n_thr_free_items = int(thr_free_idx_np.shape[0])

        # Loading deviation indices (static from masks)
        self._lam_dev_rows_all, self._lam_dev_cols_all = self._where_mask(self.primary_mask)
        self._lam_dev_rows_nonmarker, self._lam_dev_cols_nonmarker = self._where_mask(self.nonmarker_primary_mask)
        self._lam_dev_rows_partial, self._lam_dev_cols_partial = self._where_mask(self._free_loading_mask)

        # Variance free indices (static)
        var_free_idx_np = np.where(np.asarray(self._free_variance_mask))[0].astype(np.int32)
        self._var_free_idx = jnp.asarray(var_free_idx_np, dtype=jnp.int32)
        self._n_var_free = int(var_free_idx_np.shape[0])

        if self.cross_loadings:
            warnings.warn("cross_loadings=True can be unstable in small/unequal groups.", stacklevel=2)

    # ----------------------------
    # mask helpers
    # ----------------------------
    @staticmethod
    def _where_mask(mask_2d: jnp.ndarray) -> Tuple[jnp.ndarray, jnp.ndarray]:
        rows, cols = jnp.where(mask_2d)
        return rows.astype(jnp.int32), cols.astype(jnp.int32)

    def _build_free_threshold_item_mask(self) -> jnp.ndarray:
        free_factors: List[int] = list(self.partial_invariance.get("thresholds_free_factors", []))
        free_items: List[int] = list(self.partial_invariance.get("thresholds_free_items", []))

        m = jnp.zeros((self.n_items,), dtype=bool)

        for f in free_factors:
            if f < 0 or f >= self.n_factors:
                raise ValueError(f"thresholds_free_factors contains invalid factor index {f}.")
            m = m.at[self.item_to_factor_indices[f]].set(True)

        if len(free_items) > 0:
            free_items_np = np.asarray(free_items, dtype=int)
            if free_items_np.min() < 0 or free_items_np.max() >= self.n_items:
                raise ValueError("thresholds_free_items has out-of-range item index.")
            m = m.at[jnp.asarray(free_items_np, dtype=jnp.int32)].set(True)

        return m

    def _build_free_loading_mask(self) -> jnp.ndarray:
        free_factors: List[int] = list(self.partial_invariance.get("loadings_free_factors", []))
        free_pairs: List[Tuple[int, int]] = list(self.partial_invariance.get("loadings_free_pairs", []))

        m = jnp.zeros((self.n_items, self.n_factors), dtype=bool)

        for f in free_factors:
            if f < 0 or f >= self.n_factors:
                raise ValueError(f"loadings_free_factors contains invalid factor index {f}.")
            idx = self.item_to_factor_indices[f]
            if self.identification == "marker":
                if idx.shape[0] > 1:
                    m = m.at[idx[1:], f].set(True)
            else:
                m = m.at[idx, f].set(True)

        if len(free_pairs) > 0:
            pairs = np.asarray(free_pairs, dtype=int)
            if pairs.ndim != 2 or pairs.shape[1] != 2:
                raise ValueError("loadings_free_pairs must be list of (item,factor) pairs.")
            it = pairs[:, 0]
            fa = pairs[:, 1]
            if it.min() < 0 or it.max() >= self.n_items or fa.min() < 0 or fa.max() >= self.n_factors:
                raise ValueError("loadings_free_pairs contains out-of-range indices.")
            m = m.at[jnp.asarray(it, dtype=jnp.int32), jnp.asarray(fa, dtype=jnp.int32)].set(True)

        return m & self.primary_mask

    def _build_free_variance_mask(self) -> jnp.ndarray:
        free_factors: List[int] = list(self.partial_invariance.get("variances_free_factors", []))
        m = jnp.zeros((self.n_factors,), dtype=bool)
        for f in free_factors:
            if f < 0 or f >= self.n_factors:
                raise ValueError(f"variances_free_factors contains invalid factor index {f}.")
            m = m.at[f].set(True)
        return m

    # ----------------------------
    # priors
    # ----------------------------
    def _sample_positive_loadings(self, name: str, shape):
        if self.loading_prior == "truncnorm":
            return numpyro.sample(
                name,
                dist.TruncatedNormal(
                    loc=self.loading_loc_prior,
                    scale=self.loading_scale_prior,
                    low=0.0,
                ).expand(shape),
            )
        s = self.loading_scale_prior
        mu = jnp.log(jnp.clip(self.loading_loc_prior, 1e-6, 1e6))
        return numpyro.sample(name, dist.LogNormal(loc=mu, scale=s).expand(shape))

    # ----------------------------
    # model
    # ----------------------------
    def model(self, data: jnp.ndarray, predictive_group: Optional[jnp.ndarray] = None):
        # Host-side bounds check for predictive_group
        if predictive_group is not None:
            g_np = np.asarray(predictive_group)
            if g_np.size == 0:
                raise ValueError("predictive_group is empty.")
            if g_np.min() < 0 or g_np.max() >= self.n_groups:
                raise ValueError(
                    f"predictive_group must be in [0, {self.n_groups - 1}], got [{g_np.min()}, {g_np.max()}]."
                )

        y = jnp.asarray(data).astype(jnp.int32)
        n_obs, n_items = y.shape
        if n_items != self.n_items:
            raise ValueError(f"data has {n_items} items but model expects {self.n_items}.")

        g = self.group if predictive_group is None else jnp.asarray(predictive_group).astype(jnp.int32)
        G = self.n_groups
        F = self.n_factors
        K = self.n_cats
        n_thr = K - 1

        # ============================================================
        # THRESHOLDS
        # ============================================================
        if K == 2:
            if self.threshold_location == "fixed0":
                t0_base = jnp.zeros((n_items,))
            else:
                t0_base = numpyro.sample("t0_base", dist.Normal(0.0, 1.0).expand([n_items]))

            t0_g = jnp.broadcast_to(t0_base[None, :], (G, n_items))

            if G > 1 and (self.invariance != "scalar" or self._n_thr_free_items > 0):
                dev_scale = numpyro.sample("t0_dev_scale", dist.HalfNormal(self.thr_dev_scale_prior))
                if self.invariance == "scalar":
                    z_free = numpyro.sample(
                        "t0_z_free",
                        dist.Normal(0.0, 1.0).expand([G - 1, self._n_thr_free_items]),
                    )
                    t0_g = t0_g.at[1:, self._thr_free_item_idx].add(dev_scale * z_free)
                else:
                    z_all = numpyro.sample("t0_z", dist.Normal(0.0, 1.0).expand([G - 1, n_items]))
                    t0_g = t0_g.at[1:, :].add(dev_scale * z_all)

            thresholds = t0_g[..., None]  # [G,n_items,1]

        else:
            n_sp = n_thr - 1  # K-2

            raw_sp_base = numpyro.sample(
                "raw_spacings_base",
                dist.Normal(0.0, self.spacing_base_sd).expand([n_items, n_sp]),
            )

            if self.use_hierarchical_spacing:
                sp_scale = numpyro.sample("spacings_global_scale", dist.HalfNormal(self.spacing_scale_prior))
                raw_sp_base = raw_sp_base * sp_scale

            raw_sp_g = jnp.broadcast_to(raw_sp_base[None, :, :], (G, n_items, n_sp))

            if G > 1 and (self.invariance != "scalar" or self._n_thr_free_items > 0):
                sp_dev_scale = numpyro.sample("spacings_dev_scale", dist.HalfNormal(self.thr_dev_scale_prior))
                if self.invariance == "scalar":
                    z_free = numpyro.sample(
                        "spacings_z_free",
                        dist.Normal(0.0, 1.0).expand([G - 1, self._n_thr_free_items, n_sp]),
                    )
                    raw_sp_g = raw_sp_g.at[1:, self._thr_free_item_idx, :].add(sp_dev_scale * z_free)
                else:
                    z_all = numpyro.sample(
                        "spacings_z",
                        dist.Normal(0.0, 1.0).expand([G - 1, n_items, n_sp]),
                    )
                    raw_sp_g = raw_sp_g.at[1:, :, :].add(sp_dev_scale * z_all)

            delta_g = softplus(raw_sp_g) + self.min_threshold_spacing
            t_rest = jnp.cumsum(delta_g, axis=-1)

            if self.threshold_location == "fixed0":
                t0_base = jnp.zeros((n_items,))
            else:
                t0_base = numpyro.sample("t0_base", dist.Normal(0.0, 1.0).expand([n_items]))

            t0_g = jnp.broadcast_to(t0_base[None, :], (G, n_items))

            if G > 1 and (self.invariance != "scalar" or self._n_thr_free_items > 0):
                t0_dev_scale = numpyro.sample("t0_dev_scale", dist.HalfNormal(self.thr_dev_scale_prior))
                if self.invariance == "scalar":
                    z_free = numpyro.sample(
                        "t0_z_free",
                        dist.Normal(0.0, 1.0).expand([G - 1, self._n_thr_free_items]),
                    )
                    t0_g = t0_g.at[1:, self._thr_free_item_idx].add(t0_dev_scale * z_free)
                else:
                    z_all = numpyro.sample("t0_z", dist.Normal(0.0, 1.0).expand([G - 1, n_items]))
                    t0_g = t0_g.at[1:, :].add(t0_dev_scale * z_all)

            thresholds = jnp.concatenate([t0_g[..., None], t0_g[..., None] + t_rest], axis=-1)

        # ============================================================
        # NULL model (baseline)
        # ============================================================
        if self.model_type == "null":
            mu0 = jnp.zeros((n_obs, n_items))
            numpyro.sample("obs", OrderedProbit(mu0, thresholds[g], validate_args=False), obs=y)
            if self.store_deterministics:
                numpyro.deterministic("thresholds", thresholds)
            return

        # ============================================================
        # LATENT MEANS (group 0 reference)
        # ============================================================
        latent_means = jnp.zeros((G, F))
        if G > 1:
            z_means = numpyro.sample(
                "z_latent_means",
                dist.StudentT(self.df_latent, 0.0, self.latent_mean_scale).expand([G - 1, F]),
            )
            latent_means = latent_means.at[1:].set(z_means)

        # ============================================================
        # FACTOR CORRELATIONS
        # ============================================================
        if self.corr_invariance:
            L_corr = numpyro.sample("L_corr", dist.LKJCholesky(dimension=F, concentration=self.lkj_concentration))
            L_corr_g = jnp.broadcast_to(L_corr[None, :, :], (G, F, F))
        else:
            L_corr_g = numpyro.sample(
                "L_corr_g",
                dist.LKJCholesky(dimension=F, concentration=self.lkj_concentration).expand([G]),
            )

        # ============================================================
        # FACTOR SDs / covariance Cholesky
        # ============================================================
        if self.identification == "unit_variance":
            sigma_g = jnp.ones((G, F))
        else:
            log_sigma_base = numpyro.sample("log_sigma_base", dist.Normal(-0.5, 0.5).expand([F]))
            log_sigma_g = jnp.broadcast_to(log_sigma_base[None, :], (G, F))

            if G > 1:
                # Decide which factors have free group deviations in sigma
                if self.invariance == "configural" or (not self.equal_latent_sds_by_default):
                    # all factors can deviate
                    var_idx_np = np.arange(F, dtype=np.int32)
                    n_var = F
                    var_idx = jnp.asarray(var_idx_np, dtype=jnp.int32)
                else:
                    # only partial free factors
                    var_idx = self._var_free_idx
                    n_var = self._n_var_free

                if n_var > 0:
                    z_sig = numpyro.sample("log_sigma_z_free", dist.Normal(0.0, 1.0).expand([G - 1, n_var]))
                    sig_scale = numpyro.sample(
                        "log_sigma_dev_scale_free",
                        dist.HalfNormal(self.var_dev_scale_prior).expand([n_var]),
                    )
                    log_sigma_g = log_sigma_g.at[1:, var_idx].add(z_sig * sig_scale[None, :])

            sigma_g = jnp.exp(log_sigma_g)

        L_omega_g = sigma_g[:, :, None] * L_corr_g  # [G,F,F]

        # ============================================================
        # LOADINGS baseline (group 0) and group deviations
        # ============================================================
        lam_base = jnp.zeros((n_items, F))
        for f, idx in enumerate(self.item_to_factor_indices):
            n_idx = idx.shape[0]  # static
            if self.identification == "marker":
                lam_base = lam_base.at[idx[0], f].set(1.0)
                if n_idx > 1:
                    vals = self._sample_positive_loadings(f"lam_base_f{f}_nonmarker", shape=[n_idx - 1])
                    lam_base = lam_base.at[idx[1:], f].set(vals)
            else:
                vals = self._sample_positive_loadings(f"lam_base_f{f}_all", shape=[n_idx])
                lam_base = lam_base.at[idx, f].set(vals)

        lam = jnp.broadcast_to(lam_base[None, :, :], (G, n_items, F))

        # Decide free deviations for loadings
        if self.invariance == "configural":
            if self.identification == "marker":
                free_rows, free_cols = self._lam_dev_rows_nonmarker, self._lam_dev_cols_nonmarker
            else:
                free_rows, free_cols = self._lam_dev_rows_all, self._lam_dev_cols_all
        else:
            free_rows, free_cols = self._lam_dev_rows_partial, self._lam_dev_cols_partial

        n_free = free_rows.shape[0]  # no int-cast

        if G > 1 and n_free > 0:
            lam_dev_scale = numpyro.sample("lam_dev_scale", dist.HalfNormal(self.load_dev_scale_prior).expand([F]))
            lam_z = numpyro.sample("lam_z_free", dist.Normal(0.0, 1.0).expand([G - 1, n_free]))
            lam = lam.at[1:, free_rows, free_cols].add(lam_z * lam_dev_scale[None, free_cols])

        # Re-enforce marker loadings
        if self.identification == "marker":
            for f in range(F):
                lam = lam.at[:, self.marker_items[f], f].set(1.0)

        # Optional dense cross-loadings
        if self.cross_loadings:
            cross = numpyro.sample("cross_loadings", dist.Normal(0.0, self.cross_loading_scale).expand([n_items, F]))
            cross = jnp.where(self.primary_mask, 0.0, cross)
            lam = lam + cross[None, :, :]

        # ============================================================
        # LATENT SCORES + linear predictor
        # ============================================================
        z = numpyro.sample("z", dist.StudentT(self.df_latent, 0.0, 1.0).expand([n_obs, F]))
        L_obs = L_omega_g[g]  # [n_obs,F,F]
        eta = jnp.einsum("nij,nj->ni", L_obs, z) + latent_means[g]

        lam_obs = lam[g]      # [n_obs,n_items,F]
        mu = jnp.einsum("nf,nif->ni", eta, lam_obs)

        # ============================================================
        # Likelihood
        # ============================================================
        numpyro.sample("obs", OrderedProbit(mu, thresholds[g], validate_args=False), obs=y)

        # ============================================================
        # Deterministics
        # ============================================================
        if self.store_deterministics:
            numpyro.deterministic("loadings", lam)
            numpyro.deterministic("thresholds", thresholds)
            numpyro.deterministic("latent_means", latent_means)

            cov_g = jnp.einsum("gij,gkj->gik", L_omega_g, L_omega_g)
            numpyro.deterministic("factor_cov_group", cov_g)

            std_g = jnp.sqrt(jnp.clip(jnp.diagonal(cov_g, axis1=1, axis2=2), 1e-12, 1e12))
            corr_g = cov_g / (std_g[:, None, :] * std_g[:, :, None])
            numpyro.deterministic("factor_corr_group", corr_g)

        if self.debug_shapes:
            if thresholds.shape != (G, n_items, n_thr):
                raise ValueError(f"thresholds shape {thresholds.shape} expected {(G, n_items, n_thr)}")
            if lam.shape != (G, n_items, F):
                raise ValueError(f"lam shape {lam.shape} expected {(G, n_items, F)}")
            if L_omega_g.shape != (G, F, F):
                raise ValueError(f"L_omega_g shape {L_omega_g.shape} expected {(G, F, F)}")
```

```python
# ============================================================
# FULL RUN + DIAGNOSTICS LOOP (for BayesianOrdinalMGCFA v3)
# - Fixed n_cats handling (use instrument-defined n_cats)
# - v3 scalar-invariance guard respected (threshold_location fixed0)
# - Proper multi-chain rng_keys
# - Robust log_likelihood extraction for custom OrderedProbit
# - Avoids device sync for sanity checks
# ============================================================

from pathlib import Path
import numpy as np
import jax
import jax.numpy as jnp
import arviz as az
import numpyro
from numpyro.infer import NUTS, MCMC, init_to_median
from numpyro.infer.util import log_likelihood


# ----------------------------
# INPUTS (keep as NumPy first for checks)
# ----------------------------
Y_np = np.asarray(preobs_data_array, dtype=np.int32)
group_np = np.asarray(preobs_valence, dtype=np.int32)  # must be 0..G-1, length n_obs

# IMPORTANT: marker item is the FIRST element of each list (order preserved).
item_indices = [
    [0, 1, 2],        # Factor 1
    [3, 4, 5],        # Factor 2
    [6, 7, 8],        # Factor 3 ("luck")
    [9, 10, 11],      # Factor 4
    [12, 13, 14],     # Factor 5
]

partial_invariance = {
    "loadings_free_factors":   [2],
    "thresholds_free_factors": [2],
    "variances_free_factors":  [2],   # only used under identification="marker"
    # optional fine-grain:
    # "thresholds_free_items": [6, 7],
    # "loadings_free_pairs": [(6, 2), (7, 2), (8, 2)],
}

n_factors = 5
n_groups = 2

IDENTIFICATION = "marker"        # "marker" | "unit_variance"
CORR_INVARIANCE = True           # True shared corr; False group-specific corr
THRESHOLD_LOCATION = "fixed0"    # v3: must be fixed0 if running scalar invariance
MODEL_TYPE = "cfa"               # "cfa" | "null" (baseline)

# Instrument-defined category count (recommended)
N_CATS_FIXED = 6

save_dir = Path(
    r"C:/Users/rodan/OneDrive - Høgskulen på Vestlandet/Dokumenter/PhD/RQ2/analyses/mgcfa_results/v3"
)
save_dir.mkdir(parents=True, exist_ok=True)


# ----------------------------
# SANITY CHECKS (NumPy; no device sync)
# ----------------------------
if Y_np.ndim != 2:
    raise ValueError("Y must be 2D [n_obs, n_items].")

Y_min = int(Y_np.min())
Y_max = int(Y_np.max())
if Y_min < 0:
    raise ValueError("Y must be coded as integers 0..K-1 (no negatives).")

n_cats = int(N_CATS_FIXED)
if Y_max >= n_cats:
    raise ValueError(f"Data exceeds n_cats ({n_cats}). Max found: {Y_max}")

if group_np.ndim != 1 or group_np.shape[0] != Y_np.shape[0]:
    raise ValueError("group must be 1D and length n_obs (same rows as Y).")

if int(group_np.min()) < 0 or int(group_np.max()) >= n_groups:
    raise ValueError(f"group must be coded 0..{n_groups-1}.")

n_obs, n_items = Y_np.shape
print(f"n_obs={n_obs}, n_items={n_items}, n_groups={n_groups}, n_cats={n_cats}")

# Move to JAX once
Y = jnp.asarray(Y_np)
group = jnp.asarray(group_np)


# ----------------------------
# OPTIONAL: CATEGORY SPARSITY CHECK
# ----------------------------
def ordinal_freq_table_np(y_np: np.ndarray, g_np: np.ndarray, n_cats: int, n_groups: int):
    """Returns counts[g, item, cat] as np.ndarray."""
    n_obs_, n_items_ = y_np.shape
    counts = np.zeros((n_groups, n_items_, n_cats), dtype=int)
    for gg in range(n_groups):
        y_g = y_np[g_np == gg]
        for c in range(n_cats):
            counts[gg, :, c] = (y_g == c).sum(axis=0)
    return counts

counts = ordinal_freq_table_np(Y_np, group_np, n_cats, n_groups)
if n_groups > 1:
    zeros_by_item_g1 = (counts[1] == 0).sum(axis=1)
    print("\nZero-count categories per item in group 1:", zeros_by_item_g1.tolist())
    print("Items with ANY zero-count category in group 1:", np.where(zeros_by_item_g1 > 0)[0].tolist())


# ----------------------------
# RUN SETTINGS
# ----------------------------
CROSS_LOADINGS = False
LKJ_CONC = 4.0
INIT_STRATEGY = init_to_median()

SPACING_BASE_SD = 0.5
USE_HIER_SPACING = False
SPACING_SCALE_PRIOR = 0.5

NUM_WARMUP = 2000
NUM_SAMPLES = 4000
NUM_CHAINS = 4
TARGET_ACCEPT = 0.95
MAX_TREE_DEPTH = 12


# ----------------------------
# Helper: Pareto-k by group (robust to loo pointwise shape)
# ----------------------------
def pareto_k_by_group(loo_obj, group_vec: jnp.ndarray, n_groups: int, threshold: float = 0.7):
    k = loo_obj.pareto_k
    k_vals = np.asarray(k.values)
    if k_vals.ndim == 1:
        k_obs = k_vals
    else:
        k_obs = np.max(k_vals.reshape((k_vals.shape[0], -1)), axis=1)

    g_np_local = np.asarray(group_vec).astype(int)
    counts_by_group = {}
    for gg in range(n_groups):
        counts_by_group[gg] = int(((k_obs > threshold) & (g_np_local == gg)).sum())
    return counts_by_group, k_obs


# ----------------------------
# FIT + COMPARE
# ----------------------------
invariance_tests_idata = {}
loos = {}

for i, invariance in enumerate(["configural", "metric", "scalar"]):
    print(f"\n=== {invariance.upper()} invariance ===")

    # v3 guard: scalar + free threshold_location would raise
    thr_loc = THRESHOLD_LOCATION
    if invariance == "scalar":
        thr_loc = "fixed0"

    model = BayesianOrdinalMGCFA(
        n_factors=n_factors,
        n_cats=n_cats,
        group=group,
        n_groups=n_groups,
        item_to_factor_indices=item_indices,

        invariance=invariance,
        partial_invariance=partial_invariance,

        identification=IDENTIFICATION,
        corr_invariance=CORR_INVARIANCE,
        threshold_location=thr_loc,
        model_type=MODEL_TYPE,

        cross_loadings=CROSS_LOADINGS,
        cross_loading_scale=0.05,

        lkj_concentration=LKJ_CONC,
        loading_prior="truncnorm",
        equal_latent_sds_by_default=True,

        spacing_base_sd=SPACING_BASE_SD,
        use_hierarchical_spacing=USE_HIER_SPACING,
        spacing_scale_prior=SPACING_SCALE_PRIOR,

        store_deterministics=True,
        debug_shapes=False,
    )

    nuts = NUTS(
        model.model,
        init_strategy=INIT_STRATEGY,
        target_accept_prob=TARGET_ACCEPT,
        max_tree_depth=MAX_TREE_DEPTH,
    )

    mcmc = MCMC(
        nuts,
        num_warmup=NUM_WARMUP,
        num_samples=NUM_SAMPLES,
        num_chains=NUM_CHAINS,
        chain_method="sequential",  # safer on single GPU; change if desired
        progress_bar=True,
    )

    seed = 42 + i * 10
    rng_keys = jax.random.split(jax.random.PRNGKey(seed), NUM_CHAINS)
    mcmc.run(rng_keys, data=Y)

    # Robust log-likelihood extraction for custom OrderedProbit
    samples = mcmc.get_samples()
    try:
        ll = log_likelihood(model.model, samples, data=Y)["obs"]  # [draw, n_obs, n_items] (typically)
        idata = az.from_numpyro(mcmc, log_likelihood={"obs": ll})
    except Exception:
        idata = az.from_numpyro(mcmc, log_likelihood=False)

    invariance_tests_idata[invariance] = idata

    out_path = save_dir / (
        f"idata_{MODEL_TYPE}_{IDENTIFICATION}_"
        f"{'corrInv' if CORR_INVARIANCE else 'corrFree'}_{invariance}.nc"
    )
    az.to_netcdf(idata, out_path)
    print(f"Saved idata -> {out_path}")

    # Sampler diagnostics
    extra = mcmc.get_extra_fields()
    div = int(np.asarray(extra["diverging"]).sum())
    print(f"Divergences: {div}")

    summ = az.summary(idata, var_names=["~log_likelihood"])
    max_rhat = float(summ["r_hat"].max())
    min_ess = float(summ["ess_bulk"].min())
    print(f"Max R-hat: {max_rhat:.3f} | Min ESS bulk: {min_ess:.1f}")

    # LOO + Pareto-k (only if log_likelihood present)
    if "log_likelihood" in idata.groups():
        loo = az.loo(idata, pointwise=True)
        loos[invariance] = loo
        print(f"elpd_loo={float(loo.elpd_loo):.1f} | p_loo={float(loo.p_loo):.1f}")

        k_counts_07, k_obs = pareto_k_by_group(loo, group, n_groups, threshold=0.7)
        print(f"Pareto-k: #>0.7 total={int((k_obs > 0.7).sum())}, by_group={k_counts_07}")

        k_counts_10, k_obs10 = pareto_k_by_group(loo, group, n_groups, threshold=1.0)
        print(f"Pareto-k: #>1.0 total={int((k_obs10 > 1.0).sum())}, by_group={k_counts_10}")
    else:
        print("Skipping LOO: log_likelihood not available in idata (custom likelihood extraction failed).")


# Model comparison (LOO)
print("\n=== az.compare (LOO) ===")
cmp = az.compare(invariance_tests_idata, ic="loo")
print(cmp)
cmp.to_csv(save_dir / f"compare_{MODEL_TYPE}_{IDENTIFICATION}_{'corrInv' if CORR_INVARIANCE else 'corrFree'}.csv")


# Post-sampling polish (scalar model)
if "scalar" in invariance_tests_idata:
    scalar_idata = invariance_tests_idata["scalar"]

    # A) Factor correlations by group
    if "factor_corr_group" in scalar_idata.posterior:
        f_corr = scalar_idata.posterior["factor_corr_group"].mean(dim=["chain", "draw"])
        gdim = f_corr.dims[0]
        print("\nMean Factor Correlations (Group 0):")
        print(np.round(np.asarray(f_corr.sel(**{gdim: 0})), 2))
        if n_groups > 1:
            print("\nMean Factor Correlations (Group 1):")
            print(np.round(np.asarray(f_corr.sel(**{gdim: 1})), 2))
    else:
        print("\nNo factor_corr_group in posterior (store_deterministics may be False).")

    # B) Loading differences group1 - group0 for freed factor items
    if "loadings" in scalar_idata.posterior and n_groups > 1:
        lam = scalar_idata.posterior["loadings"]
        group_dim, item_dim, factor_dim = lam.dims[2], lam.dims[3], lam.dims[4]

        lam_diff_mean = (lam.sel(**{group_dim: 1}) - lam.sel(**{group_dim: 0})).mean(dim=["chain", "draw"])

        freed_factor = 2
        freed_items = np.array([6, 7, 8], dtype=int)

        diff_vals = lam_diff_mean.isel(**{item_dim: freed_items, factor_dim: freed_factor}).values
        print(
            f"\nPosterior mean loading differences (group1 - group0) "
            f"for factor {freed_factor}, items {freed_items.tolist()}:"
        )
        print(np.round(np.asarray(diff_vals), 3))
    else:
        print("\nNo loadings found in posterior (or only one group).")
```

```python
import numpy as np
import pandas as pd
import arviz as az

def extract_latent_means(
    idata,
    factor_names,
    group_names,
    hdi_prob=0.95,
):
    da = idata.posterior["latent_means"]

    # Convert to numpy: (chain, draw, group, factor)
    values = da.values

    C, D, G, F = values.shape

    if F != len(factor_names):
        raise ValueError(
            f"Model has {F} latent factors, but {len(factor_names)} names were provided."
        )

    if G != len(group_names):
        raise ValueError(
            f"Model has {G} groups, but {len(group_names)} names were provided."
        )

    rows = []

    for g in range(G):
        for f in range(F):
            samples = values[:, :, g, f].reshape(-1)

            hdi = az.hdi(samples, hdi_prob)

            rows.append({
                "Group": group_names[g],
                "Factor": factor_names[f],
                "Mean": float(samples.mean()),
                "Median": float(np.median(samples)),
                f"{int(hdi_prob*100)}% HDI Lower": float(hdi[0]),
                f"{int(hdi_prob*100)}% HDI Upper": float(hdi[1]),
            })

    return pd.DataFrame(rows)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_ppc_ordinal_item(obs, ppc_samples, model, item_idx, num_pp_samples=100):
    """
    Plot observed category proportions vs posterior predictive category proportions
    for one ordinal item with uncertainty bands.

    Parameters:
    - obs: (n_obs, n_items) numpy array of observed zero-based ordinal data
    - ppc_samples: (n_samples, n_obs, n_items) numpy array of posterior predictive draws
    - model: your model object with attribute n_cats (number of categories)
    - item_idx: int, index of the item to plot
    - num_pp_samples: int, number of posterior predictive samples to plot for uncertainty
    """

    n_cats = model.n_cats
    n_obs = obs.shape[0]

    # Observed category proportions for the selected item
    obs_counts = np.bincount(obs[:, item_idx], minlength=n_cats) / n_obs

    # Subsample posterior predictive samples (if there are more than requested)
    ppc_subsamples = ppc_samples[:num_pp_samples]

    # Compute category proportions for each posterior predictive sample
    ppc_cat_props = np.array([
        np.bincount(sample[:, item_idx], minlength=n_cats) / n_obs
        for sample in ppc_subsamples
    ])

    # Compute mean and 90% credible interval for the posterior predictive category proportions
    ppc_mean = ppc_cat_props.mean(axis=0)
    ppc_lower = np.percentile(ppc_cat_props, 5, axis=0)
    ppc_upper = np.percentile(ppc_cat_props, 95, axis=0)

    # Plot setup
    categories = np.arange(n_cats)
    plt.figure(figsize=(8, 5))

    # Plot posterior predictive samples as thin gray lines for uncertainty
    for sample_props in ppc_cat_props:
        plt.plot(categories, sample_props, color='gray', alpha=0.15, lw=1)

    # Posterior predictive mean
    plt.plot(categories, ppc_mean, color='blue', lw=3, label='Posterior predictive mean')

    # 90% credible interval bands
    plt.fill_between(categories, ppc_lower, ppc_upper, color='blue', alpha=0.2, label='90% credible interval')

    # Observed data proportions as red points with stem lines
    plt.stem(categories, obs_counts, linefmt='r-', markerfmt='ro', basefmt='k-', label='Observed data')

    plt.xlabel("Category")
    plt.ylabel("Proportion")
    plt.title(f"Posterior Predictive Check for Item {item_idx}")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import arviz as az
import pandas as pd

def inspect_loadings(
    idata,
    n_items,
    n_factors,
    n_groups,
    item_names,
    invariance_type="configural",
    partial_invariance=None,
    standardize=True,
    return_intervals=True,
    hdi_prob=0.95,
    compute_reliability=True,
    compact_ci=True,
    summary_stat="mean",
):
    """
    Inspect Bayesian ordinal CFA loadings with posterior standardization and omega reliability.

    Parameters:
    - idata: InferenceData from numpyro sampling
    - n_items, n_factors, n_groups: ints describing data structure
    - item_names: list of item names
    - invariance_type: "configural", "metric", or "scalar"
    - partial_invariance: dict with keys like 'loadings_free_factors' listing factors with free loadings
    - standardize: whether to standardize loadings
    - return_intervals: whether to return HDI intervals
    - hdi_prob: HDI probability
    - compute_reliability: whether to compute omega reliability
    - compact_ci: if True, combines mean and CI into one string in output table
    - summary_stat: "mean" or "median"
    """

    if partial_invariance is None:
        partial_invariance = {
            "loadings_free_factors": [],
            "intercepts_free_factors": [],
            "variances_free_factors": [],
            "residuals_free_factors": [],
        }

    # 1. Extract loadings posterior and stack chains & draws
    lam_xr = idata.posterior["loadings"].stack(sample=("chain", "draw"))
    dims_lam = lam_xr.dims
    sizes_lam = lam_xr.sizes

    group_dim_lam = next(d for d in dims_lam if sizes_lam[d] == n_groups)
    item_dim = next(d for d in dims_lam if sizes_lam[d] == n_items)
    factor_dim_lam = next(d for d in dims_lam if sizes_lam[d] == n_factors)
    sample_dim = next(d for d in dims_lam if d == "sample")

    lam_xr = lam_xr.transpose(sample_dim, group_dim_lam, item_dim, factor_dim_lam)
    lam = lam_xr.values  # shape (samples, groups, items, factors)

    # 2. Extract factor covariance posterior and stack chains & draws
    factor_cov_xr = idata.posterior["factor_cov_group"].stack(sample=("chain", "draw"))
    dims_cov = factor_cov_xr.dims
    sizes_cov = factor_cov_xr.sizes

    group_dim_cov = next(d for d in dims_cov if sizes_cov[d] == n_groups)
    factor_dim_cov1 = next(d for d in dims_cov if sizes_cov[d] == n_factors)
    factor_dim_cov2 = next(d for d in dims_cov if sizes_cov[d] == n_factors and d != factor_dim_cov1)
    sample_dim_cov = next(d for d in dims_cov if d == "sample")

    factor_cov_xr = factor_cov_xr.transpose(sample_dim_cov, group_dim_cov, factor_dim_cov1, factor_dim_cov2)
    factor_cov = factor_cov_xr.values  # shape (samples, groups, factors, factors)

    n_samples, G, I, F = lam.shape

    # 3. Compute factor variances and residual variances per sample, group, item
    factor_vars = np.diagonal(factor_cov, axis1=2, axis2=3)  # (samples, groups, factors)

    residual_vars = np.zeros((n_samples, G, I))
    for s in range(n_samples):
        for g in range(G):
            communality = np.sum(lam[s, g, :, :]**2 * factor_vars[s, g, :], axis=1)
            residual_vars[s, g, :] = np.clip(1 - communality, 1e-6, None)

    # 4. Standardize loadings if requested
    if standardize:
        lam_std = np.zeros_like(lam)
        for s in range(n_samples):
            for g in range(G):
                for i in range(I):
                    for f in range(F):
                        denom = np.sqrt(
                            factor_vars[s, g, f] *
                            (factor_vars[s, g, f] * lam[s, g, i, f]**2 + residual_vars[s, g, i])
                        )
                        lam_std[s, g, i, f] = lam[s, g, i, f] / denom if denom > 0 else 0
    else:
        lam_std = lam

    # 5. Posterior summary for loadings
    if summary_stat == "mean":
        lam_center = lam_std.mean(axis=0)
    elif summary_stat == "median":
        lam_center = np.median(lam_std, axis=0)
    else:
        raise ValueError("summary_stat must be 'mean' or 'median'")

    # 6. Compute HDI intervals if requested
    if return_intervals:
        hdi_lo = np.zeros((G, I, F))
        hdi_hi = np.zeros((G, I, F))
        for g in range(G):
            for i in range(I):
                for f in range(F):
                    hdi = az.hdi(lam_std[:, g, i, f], hdi_prob)
                    hdi_lo[g, i, f] = hdi[0]
                    hdi_hi[g, i, f] = hdi[1]

    def fmt(mu, lo, hi):
        return f"{mu:.3f} [{lo:.3f}, {hi:.3f}]"

    # 7. Build loadings tables
    shared = invariance_type in ("metric", "scalar")
    loadings_tables = {}

    if shared:
        # If partial invariance on loadings, show separate loadings for free factors
        free_factors = partial_invariance.get("loadings_free_factors", [])
        if free_factors:
            for g in range(G):
                data = {}
                for f in range(F):
                    col = f"Factor {f+1}"
                    if return_intervals and compact_ci:
                        col += f" ({int(hdi_prob*100)}% CI)"
                        data[col] = [
                            fmt(lam_center[g, i, f], hdi_lo[g, i, f], hdi_hi[g, i, f])
                            for i in range(I)
                        ]
                    else:
                        data[col] = lam_center[g, :, f]
                loadings_tables[f"Group {g} (partial invariance)"] = pd.DataFrame(data, index=item_names)
        else:
            # Fully shared loadings table
            data = {}
            for f in range(F):
                col = f"Factor {f+1}"
                if return_intervals and compact_ci:
                    col += f" ({int(hdi_prob*100)}% CI)"
                    data[col] = [
                        fmt(lam_center[0, i, f], hdi_lo[0, i, f], hdi_hi[0, i, f])
                        for i in range(I)
                    ]
                else:
                    data[col] = lam_center[0, :, f]
            loadings_tables["Shared loadings"] = pd.DataFrame(data, index=item_names)

    else:
        # Configural invariance: separate tables for each group
        for g in range(G):
            data = {}
            for f in range(F):
                col = f"Factor {f+1}"
                if return_intervals and compact_ci:
                    col += f" ({int(hdi_prob*100)}% CI)"
                    data[col] = [
                        fmt(lam_center[g, i, f], hdi_lo[g, i, f], hdi_hi[g, i, f])
                        for i in range(I)
                    ]
                else:
                    data[col] = lam_center[g, :, f]
            loadings_tables[f"Group {g}"] = pd.DataFrame(data, index=item_names)

    # 8. Compute omega reliability if requested
    reliability_tables = None
    if compute_reliability:

        def omega_func(lam_sample, factor_var_sample, resid_var_sample):
            omega_per_factor = []
            for f in range(F):
                lam_f = lam_sample[:, f]
                phi_f = factor_var_sample[f]
                psi = resid_var_sample
                numerator = (lam_f.sum()) ** 2 * phi_f
                denominator = numerator + np.sum(lam_f**2 * phi_f * psi + phi_f * psi)
                omega_val = numerator / denominator if denominator > 0 else np.nan
                omega_per_factor.append(omega_val)
            return omega_per_factor

        reliability_tables = {}
        if shared and free_factors:
            # Omega per group (partial invariance)
            for g in range(G):
                omegas = np.zeros((n_samples, F))
                for s in range(n_samples):
                    omegas[s, :] = omega_func(lam[s, g], factor_vars[s, g], residual_vars[s, g])
                omega_mean = np.nanmean(omegas, axis=0)
                omega_median = np.nanmedian(omegas, axis=0)
                omega_hdi = np.array([az.hdi(omegas[:, f], hdi_prob) for f in range(F)])
                rel_df = pd.DataFrame(
                    {
                        "Omega Mean": omega_mean,
                        "Omega Median": omega_median,
                        f"{int(hdi_prob*100)}% HDI Lower": omega_hdi[:, 0],
                        f"{int(hdi_prob*100)}% HDI Upper": omega_hdi[:, 1],
                    },
                    index=[f"Factor {f+1}" for f in range(F)],
                )
                reliability_tables[f"Group {g} (partial invariance)"] = rel_df
        elif shared:
            # Shared omega
            omegas = np.zeros((n_samples, F))
            for s in range(n_samples):
                omegas[s, :] = omega_func(lam[s, 0], factor_vars[s, 0], residual_vars[s, 0])
            omega_mean = np.nanmean(omegas, axis=0)
            omega_median = np.nanmedian(omegas, axis=0)
            omega_hdi = np.array([az.hdi(omegas[:, f], hdi_prob) for f in range(F)])
            rel_df = pd.DataFrame(
                {
                    "Omega Mean": omega_mean,
                    "Omega Median": omega_median,
                    f"{int(hdi_prob*100)}% HDI Lower": omega_hdi[:, 0],
                    f"{int(hdi_prob*100)}% HDI Upper": omega_hdi[:, 1],
                },
                index=[f"Factor {f+1}" for f in range(F)],
            )
            reliability_tables["Shared reliability"] = rel_df
        else:
            # Per group omega
            for g in range(G):
                omegas = np.zeros((n_samples, F))
                for s in range(n_samples):
                    omegas[s, :] = omega_func(lam[s, g], factor_vars[s, g], residual_vars[s, g])
                omega_mean = np.nanmean(omegas, axis=0)
                omega_median = np.nanmedian(omegas, axis=0)
                omega_hdi = np.array([az.hdi(omegas[:, f], hdi_prob) for f in range(F)])
                rel_df = pd.DataFrame(
                    {
                        "Omega Mean": omega_mean,
                        "Omega Median": omega_median,
                        f"{int(hdi_prob*100)}% HDI Lower": omega_hdi[:, 0],
                        f"{int(hdi_prob*100)}% HDI Upper": omega_hdi[:, 1],
                    },
                    index=[f"Factor {f+1}" for f in range(F)],
                )
                reliability_tables[f"Group {g}"] = rel_df

    return loadings_tables, reliability_tables
```

```python
def factor_correlation_table(
    idata,
    summary="mean",
    include_ci=True,
    ci_prob=0.95,
    factor_names=None,
    by_group=True,
    group_names=None,
):
    """
    Create tables of factor correlations from Bayesian (multi-group) CFA models.
    Supports configural, metric, scalar, and partial invariance models.

    Returns
    -------
    dict[str, pd.DataFrame] or pd.DataFrame
        One table per group if by_group=True, otherwise a pooled table.
    """

    da = idata.posterior["factor_corr_group"]
    dims = da.dims
    values = da.values

    # --------------------------------------------------
    # Identify dimensions
    # --------------------------------------------------
    has_group = any("group" in d for d in dims)
    has_chain = "chain" in dims and "draw" in dims
    has_sample = "sample" in dims

    # --------------------------------------------------
    # Reshape to (group, F, F, S)
    # --------------------------------------------------
    if has_chain:
        # Stack chain + draw → sample
        sample_shape = da.sizes["chain"] * da.sizes["draw"]
        values = values.reshape(sample_shape, *values.shape[2:])

        # Move sample to last axis
        values = np.moveaxis(values, 0, -1)

    elif has_sample:
        sample_axis = dims.index("sample")
        values = np.moveaxis(values, sample_axis, -1)

    else:
        raise ValueError(f"Unsupported dimensions: {dims}")

    # Now values is either:
    # (F, F, S) or (group, F, F, S)

    if values.ndim == 3:
        # Single-group → add dummy group axis
        values = values[None, ...]  # (1, F, F, S)

    G, F, _, S = values.shape

    # --------------------------------------------------
    # Naming
    # --------------------------------------------------
    if factor_names is None:
        factor_names = [f"Factor {i+1}" for i in range(F)]

    if group_names is None:
        group_names = [f"Group {g}" for g in range(G)]

    # --------------------------------------------------
    # Build tables
    # --------------------------------------------------
    tables = {}

    for g in range(G):
        rows = []

        for i in range(F):
            for j in range(i + 1, F):
                samples = values[g, i, j, :]

                if summary == "mean":
                    point = samples.mean()
                elif summary == "median":
                    point = np.median(samples)
                else:
                    raise ValueError("summary must be 'mean' or 'median'")

                row = {
                    "factor_1": factor_names[i],
                    "factor_2": factor_names[j],
                    "r": float(point),
                }

                if include_ci:
                    alpha = (1 - ci_prob) / 2
                    row["r_lower"] = float(np.quantile(samples, alpha))
                    row["r_upper"] = float(np.quantile(samples, 1 - alpha))

                rows.append(row)

        tables[group_names[g]] = pd.DataFrame(rows)

    # --------------------------------------------------
    # Return
    # --------------------------------------------------
    if by_group:
        return tables
    else:
        # Pool across groups
        pooled = pd.concat(
            tables.values(),
            keys=tables.keys(),
            names=["group"]
        ).reset_index(level="group")
        return pooled
```

```python
import jax.numpy as jnp
import numpyro
import numpyro.distributions as dist
from numpyro.distributions import constraints
from jax.nn import softplus
from typing import Optional, Sequence


class BayesianMultigroupMultivariateCFA:
    def __init__(
        self,
        n_factors: int,
        n_cats: int,
        group: Optional[jnp.ndarray] = None,
        n_groups: int = 1,
        model_type: str = "cfa",  # "cfa" or "null"
        item_to_factor_indices: Optional[Sequence[jnp.ndarray]] = None,
        invariance: Optional[str] = "metric",  # "configural", "metric", "scalar"
        cross_loadings: bool = False,
        loading_loc_prior: float = 0.6,
        loading_scale_prior: float = 0.2,
        cross_loading_scale: float = 0.05,
        min_threshold_spacing: float = 0.05,
        df_latent: float = 7.0,
        partial_invariance: Optional[dict] = None,
        allowed_crossloading_mask: Optional[jnp.ndarray] = None,  # NEW: mask for allowed cross-loadings
    ):
        self.n_factors = n_factors
        self.n_cats = n_cats
        self.item_to_factor_indices = item_to_factor_indices
        self.n_groups = n_groups
        self.model_type = model_type.lower()
        self.invariance = invariance.lower() if invariance is not None else None
        self.cross_loadings = cross_loadings
        self.loading_loc_prior = loading_loc_prior
        self.loading_scale_prior = loading_scale_prior
        self.cross_loading_scale = cross_loading_scale
        self.min_threshold_spacing = min_threshold_spacing
        self.df_latent = df_latent
        self.partial_invariance = partial_invariance or {
            "loadings_free_factors": [],
            "intercepts_free_factors": [],
            "variances_free_factors": [],
            "residuals_free_factors": [],  # Placeholder, not implemented
        }

        # Handle group vector for single-group case
        if self.n_groups == 1:
            if group is None:
                raise ValueError(
                    "For single group, you must provide a group vector of zeros."
                )
            else:
                # verify group is all zeros
                if not jnp.all(group == 0):
                    raise ValueError(
                        "For single group, group vector must be all zeros."
                    )
        else:
            if group is None:
                raise ValueError("Group vector must be provided for multigroup model.")

        self.group = group

        # NULL model shortcut
        if self.model_type == "null":
            self.primary_mask = None
            return

        assert self.model_type in ("null", "cfa")
        assert self.invariance in ("configural", "metric", "scalar")

        n_items = sum(len(idx) for idx in item_to_factor_indices)
        mask = jnp.zeros((n_items, n_factors), dtype=bool)
        for f, idx in enumerate(item_to_factor_indices):
            mask = mask.at[idx, f].set(True)
        self.primary_mask = mask
        self.n_items = n_items

        # If cross_loadings enabled, set allowed mask
        if self.cross_loadings:
            if allowed_crossloading_mask is None:
                # Default: allow cross-loadings on all non-primary positions
                self.allowed_crossloading_mask = ~self.primary_mask
            else:
                # Use provided mask, but ensure primary loadings are excluded
                self.allowed_crossloading_mask = allowed_crossloading_mask & (~self.primary_mask)
        else:
            self.allowed_crossloading_mask = None

    def model(self, data: jnp.ndarray):
        n_obs, n_items = data.shape
        y = data
        group = self.group
        G = self.n_groups
        F = self.n_factors

        # --------------------------
        # Thresholds (Intercepts)
        # --------------------------

        tau0 = numpyro.sample(
            "tau0",
            dist.Normal(0.0, 1.0).expand(
                [G, n_items] if self.invariance != "scalar" else [n_items]
            ),
        )
        if self.invariance == "scalar":
            tau0 = jnp.broadcast_to(tau0, (G, n_items))

        intercepts_free_factors = self.partial_invariance.get("intercepts_free_factors", [])
        free_intercept_items = jnp.zeros(n_items, dtype=bool)
        for f in intercepts_free_factors:
            free_intercept_items = free_intercept_items.at[self.item_to_factor_indices[f]].set(True)

        raw_delta = numpyro.sample(
            "raw_delta",
            dist.Normal(0.0, 1.0).expand([G, n_items, self.n_cats - 2]),
        )
        delta = softplus(raw_delta) + self.min_threshold_spacing

        thresholds = jnp.cumsum(
            jnp.concatenate([jnp.zeros((G, n_items, 1)), delta], axis=-1),
            axis=-1,
        )
        thresholds = thresholds + tau0[..., None]

        thresholds = thresholds.at[..., 0].set(0.0)

        for f in range(F):
            if f not in intercepts_free_factors:
                idxs = self.item_to_factor_indices[f]
                thresholds = thresholds.at[1:, idxs, :].set(thresholds[0, idxs, :])

        thresholds = jnp.sort(thresholds, axis=-1)

        if self.model_type == "null":
            mu = jnp.zeros((n_obs, n_items))
            lik = OrderedProbit(mu, thresholds[group])
            numpyro.sample("obs", lik, obs=y)
            numpyro.deterministic("log_likelihood", lik.log_prob(y).sum(axis=1))
            return

        latent_means = jnp.zeros((G, F))
        z_latent_means = numpyro.sample(
            "z_latent_means",
            dist.StudentT(self.df_latent, 0.0, 1.0).expand([G - 1, F]),
        )
        latent_means = latent_means.at[1:].set(z_latent_means)

        L_corr = numpyro.sample(
            "L_corr",
            dist.LKJCholesky(dimension=F, concentration=2.0),
        )

        sigma_g = numpyro.sample(
            "sigma_g",
            dist.HalfNormal(0.5).expand([G, F]),
        )

        variances_free_factors = self.partial_invariance.get("variances_free_factors", [])
        for f in range(F):
            if f not in variances_free_factors:
                sigma_g = sigma_g.at[1:, f].set(sigma_g[0, f])

        L_omega_g = L_corr * sigma_g[:, None, :]
        cov_g = jnp.einsum("gij,gkj->gik", L_omega_g, L_omega_g)

        lam_list = []
        for g in range(G):
            lam_g = jnp.zeros((n_items, F))
            for f, idx in enumerate(self.item_to_factor_indices):
                vals = numpyro.sample(
                    f"lam_g{g}_f{f}",
                    dist.TruncatedNormal(
                        loc=self.loading_loc_prior,
                        scale=self.loading_scale_prior,
                        low=0.0,
                    ).expand([len(idx)]),
                )
                if not (self.invariance == "configural" or f in self.partial_invariance.get("loadings_free_factors", [])) and g > 0:
                    vals = numpyro.deterministic(f"lam_g{g}_f{f}_constrained", lam_list[0][idx, f])
                vals = vals.at[0].set(1.0)
                lam_g = lam_g.at[idx, f].set(vals)
            lam_list.append(lam_g)
        lam = jnp.stack(lam_list)

        if self.cross_loadings:
            cross = numpyro.sample(
                "cross_loadings",
                dist.Normal(0.0, self.cross_loading_scale).expand([n_items, F]),
            )
            cross = cross * self.allowed_crossloading_mask
            lam = lam + cross

        z = numpyro.sample(
            "z",
            dist.StudentT(self.df_latent, 0.0, 1.0).expand([n_obs, F]),
        )

        any_var_free = len(self.partial_invariance.get("variances_free_factors", [])) > 0
        L_omega_used = (
            L_omega_g[group] if (self.invariance == "configural" or any_var_free) else jnp.broadcast_to(L_omega_g[0], (n_obs, F, F))
        )

        eta = jnp.einsum("nf,nfh->nh", z, L_omega_used)

        mu = jnp.einsum(
            "nf,nif->ni",
            eta + latent_means[group],
            lam[group],
        )

        lik = OrderedProbit(mu, thresholds[group])
        numpyro.sample("obs", lik, obs=y)

        numpyro.deterministic("log_likelihood", lik.log_prob(y).sum(axis=1))
        numpyro.deterministic("loadings", lam)
        numpyro.deterministic("thresholds", thresholds)
        numpyro.deterministic("latent_means", latent_means)
        numpyro.deterministic("factor_cov_group", cov_g)

        std_g = jnp.sqrt(jnp.diagonal(cov_g, axis1=1, axis2=2))
        corr_g = cov_g / (std_g[:, None, :] * std_g[:, :, None])
        numpyro.deterministic("factor_corr_group", corr_g)
```

```python
# Attribution construct: 5 factors, items per factor (indices)
attribution_item_indices = [
    jnp.array([0,1,2]),     # Factor 1 items
    jnp.array([3,4,5]),     # Factor 2 items
    jnp.array([6,7,8]),     # Factor 3 items
    jnp.array([9,10]),      # Factor 4 items
    jnp.array([11,12,13]),  # Factor 5 items
]

# Self-efficacy construct: 3 factors
selfefficacy_item_indices = [
    jnp.array([14,15,16]),  # Factor 1 items
    jnp.array([17,18,19]),  # Factor 2 items
    jnp.array([20,21]),     # Factor 3 items
]

# Combine them all for joint model
all_item_indices = attribution_item_indices + selfefficacy_item_indices
```

```python
# ============================================================
# FULL UPDATED SCRIPT (NumPyro / JAX) — v4
# Longitudinal Bayesian ordinal Trait–Occasion CFA with TRUE ordinal
# configural / metric / scalar invariance series + partial invariance.
#
# v4 changes (responding to latest feedback):
# - Row-level LOO: keep as default; clarify as person-wave unit (documented)
# - attach_masked_loglik_rowlevel: uses xarray.assign (no in-place mutation)
# - compare_models_by_loo_rowlevel: correct return type (pandas.DataFrame)
# - infer chain/draw sizes from mcmc.num_chains / mcmc.num_samples (robust)
# - add Pareto-k diagnostics helper before compare (recommended reporting)
# - optional cell-level LOO helper (leave-one-cell-out; sensitive to missingness)
#
# DATA EXPECTATIONS (long format)
# - y:        [n_rows, n_items] int32 in {0,...,K-1} (values for missing cells arbitrary; masked)
# - pid:      [n_rows] int32 person id in {0,...,P-1}
# - tid:      [n_rows] int32 time id   in {0,...,T-1}
# - obs_mask: [n_rows, n_items] bool (True observed, False missing)
# ============================================================

from __future__ import annotations
from dataclasses import dataclass
from typing import Sequence, Tuple, Optional, Dict, Any, List, Set

import numpy as np
import pandas as pd
import jax
import jax.numpy as jnp
from jax.nn import softplus
from jax.scipy.special import log_ndtr

import numpyro
import numpyro.distributions as dist
from numpyro.distributions import Distribution, constraints
from numpyro.handlers import mask as mask_handler
from numpyro.infer import NUTS, MCMC
from numpyro.infer.initialization import init_to_value
from numpyro.infer.util import log_likelihood

import arviz as az
import xarray as xr


# ----------------------------
# Numerics helpers
# ----------------------------
def _log1mexp(x: jnp.ndarray) -> jnp.ndarray:
    log_half = jnp.asarray(-0.6931471805599453)
    return jnp.where(x < log_half, jnp.log1p(-jnp.exp(x)), jnp.log(-jnp.expm1(x)))


def _sample_truncnorm_pos(name: str, loc: float, scale: float, shape: int | tuple[int, ...]):
    shape_tup = (shape,) if isinstance(shape, int) else tuple(shape)
    try:
        d = dist.TruncatedNormal(loc=loc, scale=scale, low=0.0)
    except TypeError:
        try:
            d = dist.TruncatedNormal(loc, scale, 0.0, jnp.inf)
        except TypeError as e:
            raise TypeError("NumPyro TruncatedNormal API mismatch; pin/upgrade numpyro.") from e
    return numpyro.sample(name, d.expand(shape_tup))


# ----------------------------
# Ordered Probit distribution
# ----------------------------
class OrderedProbit(Distribution):
    arg_constraints = {"mu": constraints.real, "thresholds": constraints.ordered_vector}

    def __init__(self, mu, thresholds, validate_args=None):
        self.mu = jnp.asarray(mu)
        thr = jnp.asarray(thresholds)

        if thr.ndim < 1:
            raise ValueError("thresholds must have at least 1 dimension (last dim = n_thr).")

        self.n_thr = int(thr.shape[-1])
        self.n_cats = self.n_thr + 1

        # Broadcast thresholds to mu.shape+(n_thr,)
        if thr.ndim == 1:
            thr = thr.reshape((1,) * self.mu.ndim + (self.n_thr,))
        elif thr.ndim == 2:
            n_items = self.mu.shape[-1]
            if thr.shape[0] != n_items:
                raise ValueError(f"thresholds shape mismatch: got {thr.shape}, expected [{n_items},{self.n_thr}]")
            thr = thr.reshape((1,) * (self.mu.ndim - 1) + (n_items, self.n_thr))
        else:
            # higher rank: rely on broadcast_to below
            pass

        self.thresholds = jnp.broadcast_to(thr, self.mu.shape + (self.n_thr,))
        super().__init__(batch_shape=self.mu.shape, event_shape=(), validate_args=validate_args)

    @property
    def support(self):
        return constraints.integer_interval(0, self.n_cats - 1)

    def log_prob(self, value):
        y = jnp.asarray(value).astype(jnp.int32)
        in_range = (y >= 0) & (y < self.n_cats)

        y_idx = jnp.clip(y, 0, self.n_cats - 1)
        y_exp = y_idx[..., None]

        t = self.thresholds - self.mu[..., None]
        neg_inf = jnp.full(self.mu.shape + (1,), -jnp.inf)
        pos_inf = jnp.full(self.mu.shape + (1,),  jnp.inf)
        t_full = jnp.concatenate([neg_inf, t, pos_inf], axis=-1)

        lower = jnp.take_along_axis(t_full[..., :-1], y_exp, axis=-1)[..., 0]
        upper = jnp.take_along_axis(t_full[...,  1:], y_exp, axis=-1)[..., 0]

        ordered = upper > lower

        log_cdf_u = log_ndtr(upper)
        log_cdf_l = log_ndtr(lower)

        # Critical guard: ensure d <= 0 (prevents _log1mexp on positive input)
        d = jnp.minimum(log_cdf_l - log_cdf_u, 0.0)

        eps = jnp.asarray(1e-12)
        d_safe = jnp.where(d > -eps, -eps, d)

        log_p_ordered = log_cdf_u + _log1mexp(d_safe)
        log_p = jnp.where(ordered, log_p_ordered, -jnp.inf)
        log_p = jnp.where(in_range, log_p, -jnp.inf)
        return log_p


# ----------------------------
# Long data container
# ----------------------------
@dataclass(frozen=True)
class LongData:
    y: jnp.ndarray
    pid: jnp.ndarray
    tid: jnp.ndarray
    obs_mask: jnp.ndarray


# ============================================================
# MODEL CLASS
# ============================================================
class BayesianOrdinalTraitOccasionCFA:
    """
    Ordinal trait–occasion CFA in long format.

    Invariance semantics (ordinal):
      - configural: loadings vary across time; thresholds vary across time
      - metric:     loadings fixed across time (except partial loadings); thresholds vary across time
      - scalar:     loadings fixed across time (except partial loadings); thresholds fixed across time
                   (except partial thresholds via free_threshold_items)

    Evaluation note (LOO):
      - "row-level" LOO treats each person-wave (row) as the left-out unit by summing item log-likelihoods
        within row: ll_row = sum_item ll_cell. This is correct under local independence and yields predictive
        accuracy for held-out person-waves (appropriate for invariance model comparison).
      - "cell-level" LOO treats each observed (row,item) cell as the left-out unit (optional helper included).
    """

    def __init__(
        self,
        n_factors: int,
        n_cats: int,
        n_times: int,
        n_person: int,
        item_to_factor_indices: Sequence[Sequence[int]],
        marker_items: Optional[Sequence[int]] = None,

        invariance: str = "metric",

        # Partial invariance
        free_loading_pairs: Optional[Sequence[Tuple[int, int]]] = None,
        free_threshold_items: Optional[Sequence[int]] = None,

        # Deviations
        loading_dev_scale: float = 0.15,
        threshold_loc_dev_scale: float = 0.10,

        # Loadings prior
        cfa_loading_loc: float = 0.7,
        loadings_scale: float = 0.5,
        allow_negative_loadings: bool = False,

        # Correlation priors
        lkj_eta_T: float = 2.0,
        lkj_eta_O: float = 2.0,

        # Time means
        alpha_sd: float = 1.0,

        # Trait/occasion score priors
        trait_df: float = 7.0,
        occas_df: float = 7.0,

        # Occasion scale prior (trait fixed)
        sigO_sd: float = 0.5,

        # Thresholds base (per item)
        delta_sd: float = 0.3,
        min_delta: float = 1e-3,
        threshold_location: str = "fixed0",  # fixed0 or free_t0
        t0_sd: float = 1.0,
    ):
        self.F = int(n_factors)
        self.K = int(n_cats)
        self.T = int(n_times)
        self.P = int(n_person)

        self.item_to_factor_indices = [list(map(int, lst)) for lst in item_to_factor_indices]

        self.invariance = invariance.lower()
        if self.invariance not in ("configural", "metric", "scalar"):
            raise ValueError("invariance must be one of: configural, metric, scalar")

        self.free_loading_pairs = None if free_loading_pairs is None else [tuple(map(int, p)) for p in free_loading_pairs]
        self.free_threshold_items = None if free_threshold_items is None else [int(i) for i in free_threshold_items]

        self.loading_dev_scale = float(loading_dev_scale)
        self.threshold_loc_dev_scale = float(threshold_loc_dev_scale)

        self.cfa_loading_loc = float(cfa_loading_loc)
        self.loadings_scale = float(loadings_scale)
        self.allow_negative_loadings = bool(allow_negative_loadings)

        self.lkj_eta_T = float(lkj_eta_T)
        self.lkj_eta_O = float(lkj_eta_O)

        self.alpha_sd = float(alpha_sd)

        self.trait_df = float(trait_df)
        self.occas_df = float(occas_df)

        self.sigO_sd = float(sigO_sd)

        self.delta_sd = float(delta_sd)
        self.min_delta = float(min_delta)
        self.threshold_location = threshold_location.lower()
        if self.threshold_location not in ("fixed0", "free_t0"):
            raise ValueError("threshold_location must be fixed0 or free_t0")
        self.t0_sd = float(t0_sd)

        # Marker items
        if marker_items is None:
            marker_items = [lst[0] for lst in self.item_to_factor_indices]
        if len(marker_items) != self.F:
            raise ValueError("marker_items must have length n_factors")
        self.marker_items = np.asarray([int(i) for i in marker_items], dtype=int)

        # CFA sparse pairs
        pairs = sorted({(i, f) for f, items in enumerate(self.item_to_factor_indices) for i in items})
        self._pairs = pairs
        self._pair_items = np.asarray([p[0] for p in pairs], dtype=np.int32)
        self._pair_facts = np.asarray([p[1] for p in pairs], dtype=np.int32)
        self._n_pairs = len(pairs)

        # Markers in pairs
        marker_pairs = {(int(self.marker_items[f]), int(f)) for f in range(self.F)}
        is_marker = np.asarray([(p in marker_pairs) for p in pairs], dtype=bool)
        self._marker_pos = np.where(is_marker)[0].astype(np.int32)
        self._nonmarker_pos = np.where(~is_marker)[0].astype(np.int32)

        # Free loading set
        self._free_loading_set: Optional[Set[Tuple[int, int]]] = None if self.free_loading_pairs is None else set(self.free_loading_pairs)

        # Free threshold items set (partial scalar)
        self._free_threshold_item_set: Optional[Set[int]] = None if self.free_threshold_items is None else set(self.free_threshold_items)

        # Precompute deviation positions for loadings (for init sizing + clarity)
        if self.invariance == "configural":
            dev_pos = self._nonmarker_pos
        else:
            if self._free_loading_set is None:
                dev_pos = np.asarray([], dtype=np.int32)
            else:
                dev_pos = np.asarray(
                    [i for i in self._nonmarker_pos.tolist() if self._pairs[i] in self._free_loading_set],
                    dtype=np.int32,
                )
        self._dev_pos = dev_pos

        # For thresholds: None means "all items" (configural/metric)
        if self.invariance in ("configural", "metric"):
            self._thr_free_items: Optional[np.ndarray] = None
        else:
            if self._free_threshold_item_set is None or len(self._free_threshold_item_set) == 0:
                self._thr_free_items = np.asarray([], dtype=np.int32)
            else:
                self._thr_free_items = np.asarray(sorted(self._free_threshold_item_set), dtype=np.int32)

    def _sample_thresholds(self, n_items: int, time_varying: bool):
        """
        Returns:
          - if not time_varying: thresholds [n_items, n_thr]
          - if time_varying: thresholds_t [T, n_items, n_thr] with location shifts applied at t>0
            (same shift applied to all cutpoints per item/time, preserving ordering).
        """
        n_thr = self.K - 1
        if self.K == 2:
            return jnp.zeros((self.T, n_items, 1)) if time_varying else jnp.zeros((n_items, 1))

        if self.threshold_location == "fixed0":
            t0 = jnp.zeros((n_items,))
        else:
            t0 = numpyro.sample("t0", dist.Normal(0.0, self.t0_sd).expand([n_items]))

        raw_delta = numpyro.sample("raw_delta", dist.Normal(0.0, self.delta_sd).expand([n_items, n_thr - 1]))
        delta = softplus(raw_delta) + self.min_delta
        t_rest = jnp.cumsum(delta, axis=1)
        thr0 = jnp.concatenate([t0[:, None], t0[:, None] + t_rest], axis=1)

        if not time_varying:
            return thr0

        if self.invariance in ("configural", "metric"):
            free_items = jnp.arange(n_items, dtype=jnp.int32)
        else:
            free_items = jnp.asarray(self._thr_free_items, dtype=jnp.int32)

        thr_t = jnp.broadcast_to(thr0[None, :, :], (self.T, n_items, n_thr))
        n_free = int(free_items.shape[0])

        if self.T > 1 and n_free > 0:
            thr_loc_dev = numpyro.sample(
                "thr_loc_dev",
                dist.Normal(0.0, self.threshold_loc_dev_scale).expand([self.T - 1, n_free]),
            )
            thr_t = thr_t.at[1:, free_items, :].add(thr_loc_dev[:, :, None])

        return thr_t

    def model(self, long_data: LongData):
        y = long_data.y.astype(jnp.int32)
        pid = long_data.pid.astype(jnp.int32)
        tid = long_data.tid.astype(jnp.int32)
        obs_mask = long_data.obs_mask.astype(jnp.bool_)

        n_rows, n_items = y.shape
        F, T = self.F, self.T

        # Separate correlation structures
        L_T = numpyro.sample("L_T", dist.LKJCholesky(F, self.lkj_eta_T))
        numpyro.deterministic("factor_corr_T", L_T @ L_T.T)

        L_O = numpyro.sample("L_O", dist.LKJCholesky(F, self.lkj_eta_O))
        numpyro.deterministic("factor_corr_O", L_O @ L_O.T)

        # Anchored time means: alpha[0]=0
        if T == 1:
            alpha = jnp.zeros((1, F))
        else:
            alpha_dev = numpyro.sample("alpha_dev", dist.Normal(0.0, self.alpha_sd).expand([T - 1, F]))
            alpha = jnp.concatenate([jnp.zeros((1, F)), alpha_dev], axis=0)
        numpyro.deterministic("alpha", alpha)

        # Trait scores: scale fixed
        zT = numpyro.sample("z_trait", dist.StudentT(self.trait_df, 0.0, 1.0).expand([self.P, F]))
        T_lat = zT @ L_T.T

        # Occasion scores: scale estimated
        sigO = numpyro.sample("sig_O", dist.HalfNormal(self.sigO_sd).expand([F]))
        numpyro.deterministic("ICC_marginal_per_factor", 1.0 / (1.0 + sigO**2))

        zO = numpyro.sample("z_occasion", dist.StudentT(self.occas_df, 0.0, 1.0).expand([n_rows, F]))
        O_lat = (zO * sigO[None, :]) @ L_O.T

        eta = alpha[tid, :] + T_lat[pid, :] + O_lat
        numpyro.deterministic("eta", eta)

        # Baseline loadings lam_pairs0 with markers fixed
        n_pairs = self._n_pairs
        n_nonmarker = int(self._nonmarker_pos.shape[0])

        if n_nonmarker > 0:
            if self.allow_negative_loadings:
                lam_nm = numpyro.sample(
                    "lam_pairs_nonmarker",
                    dist.Normal(self.cfa_loading_loc, self.loadings_scale).expand([n_nonmarker]),
                )
            else:
                lam_nm = _sample_truncnorm_pos(
                    "lam_pairs_nonmarker",
                    loc=self.cfa_loading_loc,
                    scale=self.loadings_scale,
                    shape=n_nonmarker,
                )
        else:
            lam_nm = jnp.zeros((0,))

        lam_pairs0 = jnp.zeros((n_pairs,))
        if self._marker_pos.size > 0:
            lam_pairs0 = lam_pairs0.at[jnp.asarray(self._marker_pos)].set(1.0)
        lam_pairs0 = lam_pairs0.at[jnp.asarray(self._nonmarker_pos)].set(lam_nm)

        # Time deviations for loadings
        dev_pos = self._dev_pos
        n_dev = int(dev_pos.shape[0])

        lam_pairs_t = jnp.broadcast_to(lam_pairs0[None, :], (T, n_pairs))
        if T > 1 and n_dev > 0:
            z_lam_dev = numpyro.sample("z_lam_dev", dist.Normal(0.0, 1.0).expand([T - 1, n_dev]))
            lam_pairs_t = lam_pairs_t.at[1:, jnp.asarray(dev_pos)].add(z_lam_dev * self.loading_dev_scale)

        lam_t = jnp.zeros((T, n_items, F))
        lam_t = lam_t.at[:, jnp.asarray(self._pair_items), jnp.asarray(self._pair_facts)].set(lam_pairs_t)

        # Defensive marker enforcement
        for f in range(F):
            lam_t = lam_t.at[:, self.marker_items[f], f].set(1.0)
        numpyro.deterministic("loadings_t", lam_t)

        # Thresholds per invariance
        if self.invariance in ("configural", "metric"):
            thresholds_t = self._sample_thresholds(n_items, time_varying=True)
            thresholds_row = thresholds_t[tid, :, :]
            numpyro.deterministic("thresholds_t", thresholds_t)
        else:
            time_varying = bool(self._free_threshold_item_set) if self._free_threshold_item_set is not None else False
            if time_varying:
                thresholds_t = self._sample_thresholds(n_items, time_varying=True)
                thresholds_row = thresholds_t[tid, :, :]
                numpyro.deterministic("thresholds_t", thresholds_t)
            else:
                thresholds = self._sample_thresholds(n_items, time_varying=False)
                thresholds_row = jnp.broadcast_to(thresholds[None, :, :], (n_rows, n_items, thresholds.shape[-1]))
                numpyro.deterministic("thresholds", thresholds)

        lam_row = lam_t[tid, :, :]
        mu = jnp.einsum("rf,rif->ri", eta, lam_row)

        with mask_handler(mask=obs_mask):
            numpyro.sample("obs", OrderedProbit(mu, thresholds_row, validate_args=False), obs=y)


# ============================================================
# FITTING / DIAGNOSTICS HELPERS
# ============================================================
def _run_with_extras_compat(mcmc: MCMC, key, *, model_kwargs: dict, extra_fields: Tuple[str, ...]):
    """
    API compatibility: try run(..., extra_fields=...), else run without extras.
    """
    try:
        return mcmc.run(key, **model_kwargs, extra_fields=extra_fields)
    except TypeError:
        return mcmc.run(key, **model_kwargs)


def _targeted_max_rhat_min_ess(idata: az.InferenceData, var_names: List[str]) -> Tuple[float, float]:
    rhat_ds = az.rhat(idata, var_names=var_names)
    ess_ds = az.ess(idata, var_names=var_names, method="bulk")

    max_rhat = float(np.nanmax([np.nanmax(v.to_numpy()) for v in rhat_ds.data_vars.values()]))  # type: ignore
    min_ess = float(np.nanmin([np.nanmin(v.to_numpy()) for v in ess_ds.data_vars.values()]))    # type: ignore
    return max_rhat, min_ess


def run_mcmc(
    model_obj: BayesianOrdinalTraitOccasionCFA,
    long_data: LongData,
    seed: int,
    num_warmup: int = 2000,
    num_samples: int = 4000,
    num_chains: int = 4,
    chain_method: str = "sequential",
    target_accept_prob: float = 0.95,
    max_tree_depth: int = 12,
    init_values: Optional[Dict[str, Any]] = None,
    print_diagnostics: bool = True,
    compute_rhat_ess: bool = True,
) -> Tuple[MCMC, NUTS, Optional[az.InferenceData]]:
    nuts = NUTS(
        model_obj.model,
        target_accept_prob=target_accept_prob,
        max_tree_depth=max_tree_depth,
        init_strategy=init_to_value(values=init_values) if init_values is not None else None,
    )

    mcmc = MCMC(
        nuts,
        num_warmup=num_warmup,
        num_samples=num_samples,
        num_chains=num_chains,
        chain_method=chain_method,
        progress_bar=True,
    )

    if chain_method == "sequential":
        key = jax.random.PRNGKey(seed)
    else:
        key = jax.random.split(jax.random.PRNGKey(seed), num_chains)

    _run_with_extras_compat(
        mcmc,
        key,
        model_kwargs={"long_data": long_data},
        extra_fields=("diverging", "num_steps"),
    )

    if print_diagnostics:
        try:
            extra = mcmc.get_extra_fields()
            div = np.asarray(extra.get("diverging", 0))
            steps = np.asarray(extra.get("num_steps", 0))
            n_div = int(div.sum()) if div.ndim else int(div)
            max_steps = int(steps.max()) if steps.ndim else int(steps)
            print(f"Diagnostics: divergences={n_div}, max_num_steps={max_steps}")
        except Exception as e:
            print(f"Diagnostics: could not read extra_fields ({repr(e)})")

    idata: Optional[az.InferenceData] = None
    if compute_rhat_ess:
        try:
            idata = az.from_numpyro(mcmc)
            var_names = [
                "L_T", "L_O", "alpha_dev", "sig_O",
                "lam_pairs_nonmarker", "z_lam_dev",
                "raw_delta", "thr_loc_dev", "t0",
            ]
            present = [v for v in var_names if v in idata.posterior.data_vars]
            if present:
                max_rhat, min_ess = _targeted_max_rhat_min_ess(idata, present)
                print(f"Targeted convergence: max_Rhat={max_rhat:.3f}, min_ESS_bulk={min_ess:.0f} (vars={present})")
        except Exception as e:
            print(f"Targeted convergence: failed ({repr(e)})")

    return mcmc, nuts, idata


# ============================================================
# LOG-LIKELIHOOD EXTRACTION (MASKED)
# ============================================================
def masked_loglik_cells(nuts_model, samples_flat: Dict[str, jnp.ndarray], long_data: LongData) -> np.ndarray:
    """
    Per-cell loglik with missing cells set to 0.
    Returns ll_masked [n_draws, n_rows, n_items].
    """
    ll_raw = log_likelihood(nuts_model, samples_flat, long_data=long_data)["obs"]
    ll_raw_np = np.asarray(ll_raw)
    mask_np = np.asarray(long_data.obs_mask).astype(bool)
    return np.where(mask_np[None, :, :], ll_raw_np, 0.0)


# ============================================================
# ATTACH LOG-LIKELIHOOD TO IDATA (ROW-LEVEL AND CELL-LEVEL)
# ============================================================
def attach_masked_loglik_rowlevel(
    idata: az.InferenceData,
    ll_masked_cells: np.ndarray,
    *,
    num_chains: int,
    num_samples: int,
) -> az.InferenceData:
    """
    Row-level loglik (person-wave unit):
      ll_row[draw,row] = sum_item ll_cell[draw,row,item]
    Attach as idata.log_likelihood['obs_row'] with dims [chain, draw, obs_id].
    """
    n_draws, n_rows, _ = ll_masked_cells.shape
    if n_draws != num_chains * num_samples:
        raise ValueError(f"ll first dim {n_draws} != num_chains*num_samples {num_chains*num_samples}")

    ll_row = ll_masked_cells.sum(axis=-1).reshape(num_chains, num_samples, n_rows)

    ll_xr = xr.DataArray(
        ll_row,
        dims=["chain", "draw", "obs_id"],
        name="obs_row",
    )

    # safer: build new log_likelihood dataset via assign
    if getattr(idata, "log_likelihood", None) is None:
        new_ll = xr.Dataset({"obs_row": ll_xr})
        return idata.assign(log_likelihood=new_ll)

    new_ll = idata.log_likelihood.assign(obs_row=ll_xr)
    return idata.assign(log_likelihood=new_ll)


def attach_masked_loglik_celllevel(
    idata: az.InferenceData,
    ll_masked_cells: np.ndarray,
    *,
    num_chains: int,
    num_samples: int,
) -> az.InferenceData:
    """
    Cell-level loglik (leave-one-cell-out unit).
    Flatten (row,item) to obs_id = row*n_items + item.

    Attach as idata.log_likelihood['obs_cell'] with dims [chain, draw, obs_id].
    Note: missing cells were already set to 0; they remain in obs_id unless you filter them out.
    """
    n_draws, n_rows, n_items = ll_masked_cells.shape
    if n_draws != num_chains * num_samples:
        raise ValueError(f"ll first dim {n_draws} != num_chains*num_samples {num_chains*num_samples}")

    ll_cell = ll_masked_cells.reshape(n_draws, n_rows * n_items).reshape(num_chains, num_samples, n_rows * n_items)

    ll_xr = xr.DataArray(
        ll_cell,
        dims=["chain", "draw", "obs_id"],
        name="obs_cell",
    )

    if getattr(idata, "log_likelihood", None) is None:
        new_ll = xr.Dataset({"obs_cell": ll_xr})
        return idata.assign(log_likelihood=new_ll)

    new_ll = idata.log_likelihood.assign(obs_cell=ll_xr)
    return idata.assign(log_likelihood=new_ll)


# ============================================================
# LOO DIAGNOSTICS + MODEL COMPARISON
# ============================================================
def loo_diagnostics(
    results: Dict[str, Dict[str, Any]],
    *,
    var_name: str = "obs_row",
    pointwise: bool = True,
) -> pd.DataFrame:
    """
    Compute PSIS-LOO per model (for diagnostics) and return a table with Pareto-k summaries.
    Requires idata.log_likelihood[var_name] to exist.

    Returns DataFrame with columns:
      elpd_loo, p_loo, se, pareto_k_max, pareto_k_p99, n_k_gt_0p7, warning
    """
    rows = []
    for name, res in results.items():
        idata = res.get("idata")
        ok = (idata is not None) and (getattr(idata, "log_likelihood", None) is not None) and (var_name in idata.log_likelihood)
        if not ok:
            rows.append({"model": name, "error": f"missing log_likelihood[{var_name}]"} )
            continue
        try:
            loo_res = az.loo(idata, var_name=var_name, pointwise=pointwise)
            pk = loo_res.pareto_k
            pk_np = np.asarray(pk)
            rows.append({
                "model": name,
                "elpd_loo": float(loo_res.elpd_loo),
                "p_loo": float(loo_res.p_loo),
                "se": float(loo_res.se),
                "pareto_k_max": float(np.nanmax(pk_np)),
                "pareto_k_p99": float(np.nanpercentile(pk_np, 99)),
                "n_k_gt_0p7": int(np.sum(pk_np > 0.7)),
                "warning": bool(getattr(loo_res, "warning", False)),
                "error": "",
            })
        except Exception as e:
            rows.append({"model": name, "error": repr(e)})
    return pd.DataFrame(rows).set_index("model")


def compare_models_by_loo(
    results: Dict[str, Dict[str, Any]],
    *,
    var_name: str = "obs_row",
    scale: str = "log",
) -> pd.DataFrame:
    """
    Compare models with az.compare using idata objects that contain log_likelihood[var_name].

    Returns a pandas DataFrame (ArviZ output).
    """
    compare_dict = {}
    for name, res in results.items():
        idata = res.get("idata")
        if idata is None:
            continue
        if getattr(idata, "log_likelihood", None) is None:
            continue
        if var_name not in idata.log_likelihood:
            continue
        compare_dict[name] = idata

    if len(compare_dict) < 2:
        raise ValueError(f"Need at least 2 models with log_likelihood[{var_name}] to compare.")

    return az.compare(compare_dict, ic="loo", scale=scale)


# ============================================================
# INVARIANCE SERIES FIT
# ============================================================
def fit_invariance_series(
    long_data: LongData,
    n_factors: int,
    n_cats: int,
    n_times: int,
    n_person: int,
    item_to_factor_indices: Sequence[Sequence[int]],
    marker_items: Optional[Sequence[int]] = None,
    free_loading_pairs: Optional[Sequence[Tuple[int, int]]] = None,
    free_threshold_items: Optional[Sequence[int]] = None,
    invariances: Sequence[str] = ("configural", "metric", "scalar"),
    seed: int = 42,
    chain_method: str = "sequential",
    num_warmup: int = 2000,
    num_samples: int = 4000,
    num_chains: int = 4,
    attach_celllevel_ll: bool = False,
) -> Dict[str, Dict[str, Any]]:
    # Host sanity checks
    y_np = np.asarray(long_data.y)
    if y_np.min() < 0:
        raise ValueError("Negative codes in y: use obs_mask for missing, do not code missing as -1.")
    if y_np.max() >= n_cats:
        raise ValueError(f"y has values >= n_cats ({n_cats}). max={y_np.max()}")

    results: Dict[str, Dict[str, Any]] = {}
    n_rows, n_items = map(int, long_data.y.shape)

    for i, inv in enumerate(invariances):
        model = BayesianOrdinalTraitOccasionCFA(
            n_factors=n_factors,
            n_cats=n_cats,
            n_times=n_times,
            n_person=n_person,
            item_to_factor_indices=item_to_factor_indices,
            marker_items=marker_items,
            invariance=inv,
            free_loading_pairs=free_loading_pairs,
            free_threshold_items=free_threshold_items,
        )

        # Explicit init values (robust)
        init_vals: Dict[str, Any] = {
            "L_T": jnp.eye(n_factors),
            "L_O": jnp.eye(n_factors),
            "sig_O": jnp.full((n_factors,), 0.3),
            "z_trait": jnp.zeros((n_person, n_factors)),
            "z_occasion": jnp.zeros((n_rows, n_factors)),
            "lam_pairs_nonmarker": jnp.full((len(model._nonmarker_pos),), model.cfa_loading_loc),
        }

        if n_times > 1:
            init_vals["alpha_dev"] = jnp.zeros((n_times - 1, n_factors))

        if n_cats > 2 and model.threshold_location == "free_t0":
            init_vals["t0"] = jnp.zeros((n_items,))
        if n_cats > 2:
            init_vals["raw_delta"] = jnp.zeros((n_items, n_cats - 2))

        # z_lam_dev init if deviations exist
        n_dev = int(model._dev_pos.shape[0])
        if n_times > 1 and n_dev > 0:
            init_vals["z_lam_dev"] = jnp.zeros((n_times - 1, n_dev))

        # thr_loc_dev init: only if sampled, robust sizing
        thr_loc_is_sampled = (inv in ("configural", "metric")) or (inv == "scalar" and (free_threshold_items or []))
        if n_times > 1 and thr_loc_is_sampled:
            n_free_thr = n_items if inv in ("configural", "metric") else len(free_threshold_items or [])
            init_vals["thr_loc_dev"] = jnp.zeros((n_times - 1, int(n_free_thr)))

        mcmc, nuts, idata = run_mcmc(
            model_obj=model,
            long_data=long_data,
            seed=seed + 10 * i,
            num_warmup=num_warmup,
            num_samples=num_samples,
            num_chains=num_chains,
            chain_method=chain_method,
            init_values=init_vals,
            print_diagnostics=True,
            compute_rhat_ess=True,
        )

        # masked per-cell loglik
        samples_flat = mcmc.get_samples(group_by_chain=False)
        ll_cells = masked_loglik_cells(nuts.model, samples_flat, long_data)

        # robust chain/draw sizes from MCMC object
        num_chains_eff = int(mcmc.num_chains)
        num_samples_eff = int(mcmc.num_samples)

        if idata is None:
            idata = az.from_numpyro(mcmc)

        # Attach row-level (default)
        idata = attach_masked_loglik_rowlevel(
            idata,
            ll_cells,
            num_chains=num_chains_eff,
            num_samples=num_samples_eff,
        )

        # Optionally attach cell-level too
        if attach_celllevel_ll:
            idata = attach_masked_loglik_celllevel(
                idata,
                ll_cells,
                num_chains=num_chains_eff,
                num_samples=num_samples_eff,
            )

        results[inv] = {
            "model": model,
            "mcmc": mcmc,
            "nuts": nuts,
            "idata": idata,
            "log_lik_masked_cell": ll_cells,  # retained for custom post-processing
        }

    return results


# ============================================================
# USAGE SKETCH
# ============================================================
# long_data = LongData(
#     y=jnp.asarray(y_long, dtype=jnp.int32),
#     pid=jnp.asarray(pid_long, dtype=jnp.int32),
#     tid=jnp.asarray(tid_long, dtype=jnp.int32),
#     obs_mask=jnp.asarray(mask_long, dtype=jnp.bool_),
# )
#
# results = fit_invariance_series(
#     long_data=long_data,
#     n_factors=F,
#     n_cats=K,
#     n_times=T,
#     n_person=P,
#     item_to_factor_indices=item_indices,
#     marker_items=None,
#     free_loading_pairs=None,
#     free_threshold_items=None,
#     invariances=("configural","metric","scalar"),
#     seed=42,
#     chain_method="sequential",
#     num_warmup=2000,
#     num_samples=4000,
#     num_chains=4,
#     attach_celllevel_ll=True,  # if you want both obs_row and obs_cell
# )
#
# # Pareto-k diagnostics (row-level)
# diag = loo_diagnostics(results, var_name="obs_row", pointwise=True)
# print(diag)
#
# # Compare (row-level)
# cmp = compare_models_by_loo(results, var_name="obs_row", scale="log")
# print(cmp)
#
# # If you attached cell-level, you can also compare by obs_cell:
# # diag_cell = loo_diagnostics(results, var_name="obs_cell", pointwise=True)
# # cmp_cell = compare_models_by_loo(results, var_name="obs_cell", scale="log")
# ============================================================
```
